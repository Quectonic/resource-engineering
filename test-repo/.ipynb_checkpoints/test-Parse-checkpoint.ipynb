```python
"""
    Utility Function
"""

def find_last_level(processed_elements, indent=0, verbose=False):
        
    compared_idx = len(processed_elements) - 1
    idx = 0
    
    while compared_idx >= 0:
        if verbose:
            print(f'{" " * indent}loop {idx}: {processed_elements[compared_idx]}')

        
        print(f'{" " * indent}type: {processed_elements[compared_idx].type}')
        level = 0
        nested_last_level = None 
        if type(processed_elements[compared_idx]) == HeadingDict:
            level = processed_elements[compared_idx].level
            
            if processed_elements[compared_idx].has_children():
                nested_last_level = find_last_level(processed_elements=processed_elements[compared_idx].children, indent=indent+4)
            
                if nested_last_level and nested_last_level > level:
                    level = nested_last_level
            break ## 
        
        idx += 1
        compared_idx -= 1
    return level
```

```python
def layer_(tag, layer_num=0):

    """
    Parameter(s)
    ------------
    tag: ElementDict
    layer_num: int
    """


    ### QUESTION: Is Placeholder really needed? Should it be removed and just encapsulate it by a list?

    ## Consider the heading cases, replace the layer_num with the layer number of HeadingDict

    if (not isinstance(tag, PlaceholderDict)):
        tag.set_level(layer_num)
        new_layer_num = layer_num + 1
    else:
        new_layer_num = layer_num

    if hasattr(tag, 'content'):
        for content_idx, content_ in enumerate(tag.content):
            print(content_idx)
            layer_(content_, new_layer_num)

    if hasattr(tag, 'children'):
        for child_idx, child_ in enumerate(tag.children):
            layer_(child_, new_layer_num)

    if hasattr(tag, 'row'):
        for row_idx, row_ in enumerate(tag.row):
            for cell_idx, cell_ in enumerate(row_):
                for cell_composition_idx, cell_composition in enumerate(cell_):
                    layer_(child_, new_layer_num)

    if hasattr(tag, 'item'):
        for unit_idx, unit_ in enumerate(tag.item):
            for unit_composition_idx, unit_composition in enumerate(unit_):
                layer_(child_, new_layer_num)

    if hasattr(tag, 'description'):
        for description_idx, description_ in enumerate(tag.description):
            layer_(description_, new_layer_num)
```

```python
def find_inner_elements(content, path=None, element_type=HeadingDict, cut_off=True):
    """
    Description(s)
    --------------
    This is a function that checks for the existence of inner HeadingDict 
    or other ElementDict type if specified. 

    **Best to be used on the outermost ElementDict layer of any object**
        --> the previous content indices are not recorded, so use it for inner layers 
            only if the indices is well known
    
    
    Parameter(s)
    ------------
    content: ElementDict[] --> EXCEPT HeadingDict
    path: int[] --> Recursive parameter to indicate previous path visited
    element_type: ElementDict --> the type to be found inside the nested ElementDicts
    cut_off: boolean --> whether or not the finding algorithm should stop upon reaching a target element

    Return(s)
    ---------
    int[] --> indices of the nested HeadingDict
            --> if empty, then empty list is returned
            --> e.g. [0, 1, 2] --> target HeadingDict is            \
                                    the THIRD (idx=2) element of    \
                                    the SECOND (idx=1) element of   \
                                    the FIRST (idx=0) element of    \
                                    the content list
    """  
    if path is None:
        path = []

    headings = []

    for i, item in enumerate(content):
        current_path = path + [i]
        
        if item and isinstance(item, element_type) and current_path:
            ## Base Case
            
            headings.append(current_path)
            if cut_off:
                continue

        if hasattr(item, 'content') and isinstance(item.content, list):
            headings.extend(find_inner_elements(item.content, path=current_path, element_type=element_type))

    return headings
```

```python
def organize_into_headings(elements):
    heading_only_elements = list()
    
    for idx, item in enumerate(elements):
        if item and isinstance(item, HeadingDict):
            ## appending HeadingDict, not crawling deeper into the nested hierarchy
            
            heading_only_elements.append(item)
            
        else:
                
            ## TODO:    set the level of the nested ElementDict before appending
            ##           -  new function might be needed

            ## NEED A FUNCTION TO CHECK FOR INNER HEADINGS 
            ## --> if it exists, return the index of the headings in the format of [0, 2, 1, 3, ...]
            ##                                                                   |--> depends on how deep the heading is nested into
            ## --> if it does not exist, return an empty list []
            try: 
                nested_headings_ = list()

                ## CHECKING EXISTENCE FOR INNER HEADINGS 
                ## APPEND TO heading_only_elements IF FOUND
                if item and hasattr(item, 'content'):
                    ## NOTE: Upon reaching HeadingDict, the crawl would stop
                    nested_headings_ = find_inner_elements(item.content)
                for index in nested_headings_:
                    idx = 0
                    child = item
                    while idx < len(index):
                        if idx == len(index) - 1:
                            child = child.content.pop(index[idx])
                            heading_only_elements.append(child)
                        else:
                            child = child.content[index[idx]]
                        idx += 1
                        
                heading_only_elements[-1].content.append(item)            
            except Exception as error:
                print(f'Appending as an item instead - - - {error}')
                heading_only_elements.append(item)
    return heading_only_elements
```

```python
def find_last_nested_headingdict(processed_elements):
        
    """
    Return a list of indexes referring to the index of content list of each child
    """
    
    compared_idx = len(processed_elements) - 1
    
    while compared_idx >= 0:
        if type(processed_elements[compared_idx]) == HeadingDict:
            if processed_elements[compared_idx].has_children():
                return [compared_idx] + find_last_nested_headingdict(processed_elements[compared_idx].children)
            else:
                return [compared_idx]
            break
        compared_idx -= 1
```

```python
practiceHtml = """<div class="mw-parser-output">
<figure class="mw-halign-right" typeof="mw:File/Thumb"><a href="/index.php/File:Heart-and-lungs.jpg" class="mw-file-description"><img src="/images/e/e6/Heart-and-lungs.jpg" decoding="async" width="230" height="224"></a><figcaption>The heart and lungs, from an older edition of <i> <a href="/index.php/Gray%27s_Anatomy" class="mw-redirect" title="Gray's Anatomy">Gray's Anatomy</a></i>.</figcaption></figure>
<table class="infobox bordered" style="width: 15em; text-align: left; font-size: 90%; background:AliceBlue">

<tbody><tr>
<td colspan="1" style="text-align:center; background:DarkGray">
<p><b>WikiDoc Resources for Heart</b>
</p>
</td></tr>
<tr bgcolor="LightGrey">
<th>
<p>Articles
</p>
</th></tr>
<tr>
<th>
<p><small>
<a rel="nofollow" class="external text" href="http://www.ncbi.nlm.nih.gov/entrez/query.fcgi?CMD=search&amp;db=pubmed&amp;term=Heart">Most recent articles on Heart</a>
</small></p><p><small><a rel="nofollow" class="external text" href="http://scholar.google.com/scholar?ie=UTF-8&amp;oe=utf-8&amp;q=Heart&amp;qt_s=Search&amp;sa=N&amp;tab=ps">Most cited articles on Heart</a>
</small></p><p><small><a rel="nofollow" class="external text" href="http://www.ncbi.nlm.nih.gov/sites/entrez?cmd=search&amp;db=pubmed&amp;term=Heart%20AND%20systematic%5Bsb%5D">Review articles on Heart</a>
</small></p><p><small><a rel="nofollow" class="external text" href="http://www.ncbi.nlm.nih.gov/entrez/query.fcgi?cmd=search&amp;db=pubmed&amp;term=Heart+AND+%28%28N+Engl+J+Med%5Bta%5D%29+OR+%28Lancet%5Bta%5D%29+OR+%28BMJ%5Bta%5D%29%29">Articles on Heart in N Eng J Med, Lancet, BMJ</a>
</small>
</p>
</th></tr>
<tr bgcolor="LightGrey">
<th>
<p>Media
</p>
</th></tr>
<tr>
<th>
<p><small>
<a rel="nofollow" class="external text" href="http://www.google.com/search?q=Heart+ppt&amp;ie=utf-8&amp;oe=utf-8&amp;aq=t&amp;rls=org.mozilla:en-US:official&amp;client=firefox-a">Powerpoint slides on Heart</a>
</small></p><p><small><a rel="nofollow" class="external text" href="http://images.google.com/images?q=Heart&amp;ie=UTF-8&amp;oe=utf-8&amp;rls=org.mozilla:en-US:official&amp;client=firefox-a&amp;um=1&amp;sa=N&amp;tab=wi">Images of Heart</a>
</small></p><p><small><a rel="nofollow" class="external text" href="http://picasaweb.google.com/lh/searchbrowse?q=Heart&amp;ie=UTF-8&amp;oe=utf-8&amp;rls=org.mozilla:en-US:official&amp;client=firefox-a&amp;um=1&amp;sa=N&amp;tab=wq#2+1">Photos of Heart</a>
</small></p><p><small><a rel="nofollow" class="external text" href="http://www.google.com/search?hl=en&amp;client=firefox-a&amp;rls=org.mozilla%3Aen-US%3Aofficial&amp;hs=hPo&amp;q=Heart+podcasts+OR+MP3&amp;btnG=Search">Podcasts &amp; MP3s on Heart</a>
</small></p><p><small><a rel="nofollow" class="external text" href="http://video.google.com/videosearch?q=Heart&amp;ie=UTF-8&amp;oe=utf-8&amp;rls=org.mozilla:en-US:official&amp;um=1&amp;sa=N&amp;tab=fv#">Videos on Heart</a></small>
</p>
</th></tr>
<tr bgcolor="LightGrey">
<th>
<p>Evidence Based Medicine
</p>
</th></tr>
<tr>
<th>
<p><small>
<a rel="nofollow" class="external text" href="http://www.ncbi.nlm.nih.gov/entrez/query.fcgi?cmd=search&amp;db=pubmed&amp;term=%28Heart%29+AND+%28Cochrane+Database+Syst+Rev%5Bta%5D%29">Cochrane Collaboration on Heart</a>
</small></p><p><small><a rel="nofollow" class="external text" href="http://www.jr2.ox.ac.uk/cgi-bin/htsearch?config=bando&amp;restrict=&amp;exclude=&amp;method=and&amp;format=builtin-long&amp;sort=score&amp;words=Heart">Bandolier on Heart</a>
</small></p><p><small><a rel="nofollow" class="external text" href="http://www.tripdatabase.com/SearchResults.html?s=1&amp;criteria=Heart&amp;btnG=Search">TRIP on Heart</a>
</small>
</p>
</th></tr>
<tr bgcolor="LightGrey">
<th>
<p>Clinical Trials
</p>
</th></tr>
<tr>
<th>
<p><small>
<a rel="nofollow" class="external text" href="http://clinicaltrials.gov/search/open/condition=Heart">Ongoing Trials on Heart at Clinical Trials.gov</a>
</small></p><p><small><a rel="nofollow" class="external text" href="http://www.ncbi.nlm.nih.gov/entrez/query.fcgi?cmd=search&amp;db=pubmed&amp;term=Heart+AND+%28randomized+controlled+trial%5BPublication+Type%5D+OR+%28randomized%5BTitle%2FAbstract%5D+AND+controlled%5BTitle%2FAbstract%5D+AND+trial%5BTitle%2FAbstract%5D%29%29">Trial results on Heart</a>
</small></p><p><small><a rel="nofollow" class="external text" href="http://www.google.com/search?hl=en&amp;client=firefox-a&amp;rls=org.mozilla:en-US:official&amp;hs=9X6&amp;q=Heart+more:clinical_trials&amp;cx=disease_for_health_professionals&amp;sa=N&amp;oi=coopctx&amp;resnum=0&amp;ct=col2&amp;cd=3">Clinical Trials on Heart at Google</a>
</small></p><p class="mw-empty-elt">
</p>
</th></tr>
<tr bgcolor="LightGrey">
<th>
<p>Guidelines / Policies / Govt
</p>
</th></tr>
<tr>
<th>
<p><small>
<a rel="nofollow" class="external text" href="http://www.guideline.gov/search/searchresults.aspx?Type=3&amp;txtSearch=Heart&amp;num=20">US National Guidelines Clearinghouse on Heart</a>
</small></p><p><small><a rel="nofollow" class="external text" href="http://www.nice.org.uk/search/guidancesearchresults.jsp?keywords=Heart&amp;searchType=guidance">NICE Guidance on Heart</a>
</small></p><p><small><a rel="nofollow" class="external text" href="http://www.prodigy.nhs.uk/search?&amp;page=1&amp;q=Heart&amp;site=0">NHS PRODIGY Guidance</a>
</small></p><p><small><a rel="nofollow" class="external text" href="http://google2.fda.gov/search?q=Heart&amp;x=0&amp;y=0&amp;client=FDA&amp;site=FDA&amp;lr=&amp;proxystylesheet=FDA&amp;output=xml_no_dtd&amp;getfields=*">FDA on Heart</a>
</small></p><p><small><a rel="nofollow" class="external text" href="http://www.cdc.gov/search.do?queryText=Heart&amp;searchButton.x=33&amp;searchButton.y=6&amp;action=search">CDC on Heart</a>
</small></p><p class="mw-empty-elt">
</p>
</th></tr>
<tr bgcolor="LightGrey">
<th>
<p>Books
</p>
</th></tr>
<tr>
<th>
<p><small>
<a rel="nofollow" class="external text" href="http://books.google.com/books?ie=UTF-8&amp;oe=utf-8&amp;q=Heart&amp;qt_s=Search&amp;sa=N&amp;tab=gp">Books on Heart</a></small>
</p>
</th></tr>
<tr bgcolor="LightGrey">
<th>
<p>News
</p>
</th></tr>
<tr>
<th>
<p><small>
<a rel="nofollow" class="external text" href="http://news.google.com/news?q=Heart&amp;ie=UTF-8&amp;oe=utf-8&amp;rls=org.mozilla:en-US:official&amp;client=firefox-a&amp;um=1&amp;sa=N&amp;tab=wn">Heart in the news</a>
</small></p><p><small><a rel="nofollow" class="external text" href="http://www.google.com/alerts/create?hl=en&amp;gl=">Be alerted to news on Heart</a>
</small></p><p><small><a rel="nofollow" class="external text" href="http://www.google.com/trends?q=Heart">News trends on Heart</a></small>
</p>
</th></tr>
<tr bgcolor="LightGrey">
<th>
<p>Commentary
</p>
</th></tr>
<tr>
<th>
<p><small>
<a rel="nofollow" class="external text" href="http://blogsearch.google.com/blogsearch?q=Heart&amp;ie=UTF-8&amp;oe=utf-8&amp;rls=org.mozilla:en-US:official&amp;client=firefox-a&amp;um=1&amp;sa=N&amp;tab=wb">Blogs on Heart</a></small>
</p>
</th></tr>
<tr bgcolor="LightGrey">
<th>
<p>Definitions
</p>
</th></tr>
<tr>
<th>
<p><small>
<a rel="nofollow" class="external text" href="http://www.google.com/search?hl=en&amp;q=define+Heart&amp;btnG=Search">Definitions of Heart</a></small>
</p>
</th></tr>
<tr bgcolor="LightGrey">
<th>
<p>Patient Resources / Community
</p>
</th></tr>
<tr>
<th>
<p><small>
<a rel="nofollow" class="external text" href="http://www.google.com/search?hl=en&amp;q=Heart+more:for_patients&amp;cx=disease_for_patients&amp;sa=N&amp;oi=cooptsr&amp;resnum=0&amp;ct=col3&amp;cd=1">Patient resources on Heart</a>
</small></p><p><small><a rel="nofollow" class="external text" href="http://groups.google.com/groups/search?ie=UTF-8&amp;oe=utf-8&amp;q=Heart&amp;qt_s=Search">Discussion groups on Heart</a>
</small></p><p><small><a rel="nofollow" class="external text" href="http://www.google.com/search?hl=en&amp;client=firefox-a&amp;rls=org.mozilla:en-US:official&amp;q=Heart+more:patient_handouts&amp;cx=disease_for_health_professionals&amp;sa=N&amp;oi=coopctx&amp;resnum=0&amp;ct=col1&amp;cd=3">Patient Handouts on Heart</a>
</small></p><p><small><a rel="nofollow" class="external text" href="http://maps.google.com/maps?f=q&amp;hl=en&amp;geocode=&amp;q=Heart&amp;sll=37.0625,-95.677068&amp;sspn=65.008093,112.148438&amp;ie=UTF8&amp;ll=37.0625,-95.677068&amp;spn=91.690419,149.414063&amp;z=2&amp;source=embed">Directions to Hospitals Treating Heart</a>
</small></p><p><small><a rel="nofollow" class="external text" href="http://www.google.com/search?hl=en&amp;client=firefox-a&amp;rls=org.mozilla%3Aen-US%3Aofficial&amp;hs=QWo&amp;q=Heart+AND+risk+score+OR+risk+calculator&amp;btnG=Search">Risk calculators and risk factors for Heart</a>
</small></p><p class="mw-empty-elt">
</p>
</th></tr>
<tr bgcolor="LightGrey">
<th>
<p>Healthcare Provider Resources
</p>
</th></tr>
<tr>
<th>
<p><small>
<a rel="nofollow" class="external text" href="http://www.google.com/search?hl=en&amp;q=Heart+more:condition_symptoms&amp;cx=disease_for_patients&amp;sa=N&amp;oi=coopctx&amp;resnum=0&amp;ct=col1&amp;cd=2">Symptoms of Heart</a>
</small></p><p><small><a rel="nofollow" class="external text" href="http://www.google.com/search?hl=en&amp;client=firefox-a&amp;rls=org.mozilla:en-US:official&amp;hs=qml&amp;q=Heart+more:causes_risk_factors&amp;cx=disease_for_patients&amp;sa=N&amp;oi=coopctx&amp;resnum=0&amp;ct=col2&amp;cd=1">Causes &amp; Risk Factors for Heart</a>
</small></p><p><small><a rel="nofollow" class="external text" href="http://www.google.com/search?hl=en&amp;client=firefox-a&amp;rls=org.mozilla:en-US:official&amp;hs=R7Q&amp;q=Heart+more:tests_diagnosis&amp;cx=disease_for_patients&amp;sa=N&amp;oi=cooptsr&amp;resnum=0&amp;ct=col2&amp;cd=1">Diagnostic studies for Heart</a>
</small></p><p><small><a rel="nofollow" class="external text" href="http://www.google.com/search?hl=en&amp;client=firefox-a&amp;rls=org.mozilla:en-US:official&amp;hs=XX6&amp;q=Heart+more:condition_treatment&amp;cx=disease_for_health_professionals&amp;sa=N&amp;oi=coopctx&amp;resnum=0&amp;ct=col1&amp;cd=1">Treatment of Heart</a>
</small>
</p>
</th></tr>
<tr bgcolor="LightGrey">
<th>
<p>Continuing Medical Education (CME)
</p>
</th></tr>
<tr>
<th>
<p><small>
<a rel="nofollow" class="external text" href="http://www.google.com/search?hl=en&amp;client=firefox-a&amp;rls=org.mozilla:en-US:official&amp;q=Heart+more:health_continuing_education&amp;cx=disease_for_health_professionals&amp;sa=N&amp;oi=coopctx&amp;resnum=0&amp;ct=col3&amp;cd=2">CME Programs on Heart</a>
</small>
</p>
</th></tr>
<tr bgcolor="LightGrey">
<th>
<p>International
</p>
</th></tr>
<tr>
<th>
<p><small>
<a rel="nofollow" class="external text" href="http://www.google.com/search?q=Heart+en+espanol&amp;ie=utf-8&amp;oe=utf-8&amp;aq=t&amp;rls=org.mozilla:en-US:official&amp;client=firefox-a">Heart en Espanol</a>
</small></p><p><small><a rel="nofollow" class="external text" href="http://www.google.com/search?q=Heart+en+francais&amp;ie=utf-8&amp;oe=utf-8&amp;aq=t&amp;rls=org.mozilla:en-US:official&amp;client=firefox-a">Heart en Francais</a>
</small></p><p class="mw-empty-elt">
</p>
</th></tr>
<tr bgcolor="LightGrey">
<th>
<p>Business
</p>
</th></tr>
<tr>
<th>
<p><small>
<a rel="nofollow" class="external text" href="http://finance.google.com/finance?ie=UTF-8&amp;oe=utf-8&amp;q=Heart&amp;qt_s=Search&amp;sa=N&amp;tab=te">Heart in the Marketplace</a>
</small></p><p><small><a rel="nofollow" class="external text" href="http://www.google.com/patents?tab=st&amp;sa=N&amp;ie=UTF-8&amp;oe=utf-8&amp;q=Heart&amp;qt_s=Search&amp;sa=N">Patents on Heart</a></small>
</p>
</th></tr>
<tr bgcolor="LightGrey">
<th>
<p>Experimental / Informatics
</p>
</th></tr>
<tr>
<th>
<p><small>
<a rel="nofollow" class="external text" href="http://labs.google.com/sets?hl=en&amp;q1=Heart&amp;q2=&amp;q3=&amp;q4=&amp;q5=&amp;btn=Large+Set">List of terms related to Heart</a>
</small>
</p>
</th></tr>
</tbody></table>
<p><b>Editor-In-Chief:</b> <a href="/index.php/User:C_Michael_Gibson" title="User:C Michael Gibson">C. Michael Gibson, M.S., M.D.</a> <a rel="nofollow" class="external autonumber" href="mailto:charlesmichaelgibson@gmail.com">[1]</a>
</p>
<h2><span class="mw-headline" id="Overview">Overview</span></h2>
<p>The <b>heart</b> is a <a href="/index.php/Muscle" title="Muscle">muscular</a> <a href="/index.php/Organ_(anatomy)" title="Organ (anatomy)">organ</a> responsible for pumping <a href="/index.php/Blood" title="Blood">blood</a> through the <a href="/index.php/Blood_vessel" title="Blood vessel">blood vessels</a> by repeated, rhythmic contractions, or a similar structure in the annelids, mollusks, and arthropods.<sup id="cite_ref-1" class="reference"><a href="#cite_note-1">[1]</a></sup> The term <i>cardiac</i> (as in <a href="/index.php/Cardiology" title="Cardiology">cardiology</a>) means "related to the heart" and comes from the <a href="/index.php?title=Greek_language&amp;action=edit&amp;redlink=1" class="new" title="Greek language (page does not exist)">Greek</a> καρδία, <i>kardia</i>, for "heart." The heart is composed of <a href="/index.php/Cardiac_muscle" title="Cardiac muscle">cardiac muscle</a>, an <a href="/index.php/Involuntary_muscle" title="Involuntary muscle">involuntary muscle</a> tissue which is found only within this organ.<sup id="cite_ref-2" class="reference"><a href="#cite_note-2">[2]</a></sup> The average human heart beating at 72 BPM, will beat approximately 2.5 billion times during a lifetime of 66 years.
</p>
<h2><span class="mw-headline" id="Early_development">Early development</span></h2>
<div role="note" class="hatnote navigation-not-searchable">Main article: <a href="/index.php/Heart_development" title="Heart development">Heart development</a></div>
<p>The human <a href="/index.php/Embryon" class="mw-redirect" title="Embryon">embryonic</a> heart begins beating about 21 days after conception, or five weeks after the last normal <a href="/index.php/Menstrual_period" class="mw-redirect" title="Menstrual period">menstrual period</a> (LMP), which is the date normally used to date pregnancy.  The human heart begins beating at a rate near the mother’s, about 75-80 beats per minute (BPM). The embryonic heart rate (EHR) then accelerates linearly for the first month of beating, peaking at 165-185 BPM during the early 7th week, (early 9th week after the LMP). This acceleration is approximately 3.3 BPM per day, or about 10 BPM every three days, an increase of 100 BPM in the first month.<sup id="cite_ref-3" class="reference"><a href="#cite_note-3">[3]</a></sup>
</p><p>After peaking at about 9.5 weeks after the LMP, it decelerates to about 152 BPM (+/-25 BPM) during the 15th week after the LMP. After the 15th week the deceleration slows reaching an average rate of about 145 (+/-25 BPM) BPM at term. The regression formula which describes this acceleration before the embryo reaches 25 mm in crown-rump length or 9.2 LMP weeks is Age in days = EHR(0.3)+6 
</p><p>There is no difference in male and female heart rates before birth.<sup id="cite_ref-4" class="reference"><a href="#cite_note-4">[4]</a></sup>
</p>
<h2><span class="mw-headline" id="Structure">Structure</span></h2>
<p>In the human body, the heart is usually situated in the middle of the <a href="/index.php/Thorax" class="mw-redirect" title="Thorax">thorax</a> with the largest part of the heart slightly offset to the left (although sometimes it is on the right, see <a href="/index.php/Dextrocardia" title="Dextrocardia">dextrocardia</a>), underneath the <a href="/index.php/Breastbone" class="mw-redirect" title="Breastbone">breastbone</a> (see <a rel="nofollow" class="external text" href="http://www.easyrashi.com/TEFILLIN/Position_of_heart.htm">diagrams</a>). The heart is usually felt to be on the left side because the <a href="/index.php/Left_heart" title="Left heart">left heart</a> (left ventricle) is stronger (it pumps to all body parts). The left <a href="/index.php/Lung" title="Lung">lung</a> is smaller than the right lung because the heart occupies more of the left hemithorax.  The heart is enclosed by a sac known as the <a href="/index.php/Pericardium" title="Pericardium">pericardium</a> and is surrounded by the <a href="/index.php/Lung" title="Lung">lungs</a>. The pericardium is a double membrane structure containing a <a href="/index.php/Serous" class="mw-redirect" title="Serous">serous</a> fluid to reduce friction during heart contractions. The <a href="/index.php/Mediastinum" title="Mediastinum">mediastinum</a>, a subdivision of the thoracic cavity, is the name of the heart cavity. 
</p><p>The <i>apex</i> is the blunt point situated in an inferior (pointing down and left) direction. A stethoscope can be placed directly over the apex so that the beats can be counted. It is located posterior to the 5th intercostal space in the left mid-clavicular line. In normal adults, the mass of the heart is 250-350 g (9-12 oz), or about three quarters the size of a clenched fist, but extremely diseased hearts can be up to 1000 g (2 lb) in mass due to <a href="/index.php/Organ_hypertrophy" class="mw-redirect" title="Organ hypertrophy">hypertrophy</a>. It consists of four chambers, the two upper <i>atria</i> (singular: atrium
) and the two lower <i>ventricles</i>.  Below, on the right, is a picture of a fresh human heart which was removed from a 64-year-old British male. 
</p>
<div align="left">
<ul class="gallery mw-gallery-traditional">
		<li class="gallerybox" style="width: 260px">
			<div class="thumb" style="width: 255px; height: 255px;"><span typeof="mw:File"><a href="/index.php/File:EHR-BBII.jpg" class="mw-file-description" title="At 21 days after conception, the human heart begins beating at 70 to 80 beats per minute and accelerates linearly for the first month of beating."><img alt="At 21 days after conception, the human heart begins beating at 70 to 80 beats per minute and accelerates linearly for the first month of beating." src="/images/c/ce/EHR-BBII.jpg" decoding="async" width="225" height="155"></a></span></div>
			<div class="gallerytext">
<p>At 21 days after <a href="/index.php/Fertilization" title="Fertilization">conception</a>, the human heart begins beating at 70 to 80 beats per minute and accelerates linearly for the first month of beating.
</p>
			</div>
		</li>
		<li class="gallerybox" style="width: 260px">
			<div class="thumb" style="width: 255px; height: 255px;"><span typeof="mw:File"><a href="/index.php/File:3DScience_Human_Heart.jpg" class="mw-file-description" title="Anterior (frontal) view of the opened heart. Arrows indicate normal blood flow. Image provided courtesy of www.3dscience.com."><img alt="Anterior (frontal) view of the opened heart. Arrows indicate normal blood flow. Image provided courtesy of www.3dscience.com." src="/images/c/c9/3DScience_Human_Heart.jpg" decoding="async" width="207" height="225"></a></span></div>
			<div class="gallerytext">
<p>Anterior (frontal) view of the opened heart. Arrows indicate normal blood flow. Image provided courtesy of www.3dscience.com.
</p>
			</div>
		</li>
		<li class="gallerybox" style="width: 260px">
			<div class="thumb" style="width: 255px; height: 255px;"><span typeof="mw:File"><a href="/index.php/File:Humhrt2.jpg" class="mw-file-description" title="Human heart"><img alt="Human heart" src="/images/b/b7/Humhrt2.jpg" decoding="async" width="174" height="225"></a></span></div>
			<div class="gallerytext">
<p>Human heart
</p>
			</div>
		</li>
</ul>
</div>
<h2><span class="mw-headline" id="Functioning">Functioning</span></h2>
<p>The function of the right side of the heart (see <a href="/index.php/Right_heart" title="Right heart">right heart</a>) is to collect de-oxygenated blood, in the <a href="/index.php/Right_atrium" title="Right atrium">right atrium</a>, from the body and pump it, via the <a href="/index.php/Right_ventricle" title="Right ventricle">right ventricle</a>, into the lungs (<a href="/index.php/Pulmonary_circulation" title="Pulmonary circulation">pulmonary circulation</a>) so that carbon dioxide can be dropped off and oxygen picked up (<a href="/index.php/Gas_exchange" title="Gas exchange">gas exchange</a>). This happens through a passive process called diffusion. The left side (see <a href="/index.php/Left_heart" title="Left heart">left heart</a>) collects oxygenated blood from the <a href="/index.php/Lung" title="Lung">lungs</a> into the <a href="/index.php/Left_atrium" title="Left atrium">left atrium</a>.  From the left atrium the blood moves to the <a href="/index.php/Left_ventricle" title="Left ventricle">left ventricle</a> which pumps it out to the body.  On both sides, the lower ventricles are thicker and stronger than the upper atria.
The muscle wall surrounding the left ventricle is thicker than the wall surrounding the right ventricle due to the higher force needed to pump the blood through the <a href="/index.php/Systemic_circulation" title="Systemic circulation">systemic circulation</a>.
</p><p>Starting in the <a href="/index.php/Right_atrium" title="Right atrium">right atrium</a>, the blood flows through the <a href="/index.php/Tricuspid_valve" title="Tricuspid valve">tricuspid valve</a> to the <a href="/index.php/Right_ventricle" title="Right ventricle">right ventricle</a>. Here it is pumped out the pulmonary semilunar valve and travels through the pulmonary <a href="/index.php/Artery" title="Artery">artery</a> to the lungs. From there, blood flows back through the pulmonary <a href="/index.php/Vein" title="Vein">vein</a> to the left atrium. It then travels through the bicuspid valve to the left ventricle and on to through the aortic semilunar valve to the aorta. The aorta forks, and the blood is divided between major arteries which supply the upper and lower body. The blood travels the arteries to the smaller arterioles, then finally to the tiny capillaries which feed each cell. The (relatively) deoxygenated blood then travels to the venules, which coalesce into veins, then to the inferior and superior vena cavae and finally back to the right atrium where the process began.
</p><p>The heart is effectively a <a href="/index.php/Syncytium" title="Syncytium">syncytium</a>, a meshwork of cardiac muscle cells interconnected by contiguous cytoplasmic bridges. This relates to electrical stimulation of one cell spreading to neighboring cells
</p>
<h2><span class="mw-headline" id="First_aid">First aid</span></h2>
<p><i>See <a href="/index.php/Cardiac_arrest" class="mw-redirect" title="Cardiac arrest">cardiac arrest</a> for emergencies involving the heart</i>
</p><p>If a person is encountered in cardiac arrest (no heartbeat), <a href="/index.php/Cardiopulmonary_resuscitation" title="Cardiopulmonary resuscitation">cardiopulmonary resuscitation</a> (CPR) should be started and help called. If an <a href="/index.php/Automated_external_defibrillator" title="Automated external defibrillator">automated external defibrillator</a> is available, this device may automatically administer <a href="/index.php/Defibrillation" title="Defibrillation">defibrillation</a> if this is indicated. Usually, if there is enough time, the person can be rushed to the hospital where he or she will be cared for by a cardiologist, a doctor who specializes in the heart and lungs.
</p>
<h2><span class="mw-headline" id="References">References</span></h2>
<div class="reflist columns references-column-width" style="-moz-column-width: 30em; -webkit-column-width: 30em; column-width: 30em; list-style-type: decimal;">
<ol class="references">
<li id="cite_note-1"><span class="mw-cite-backlink"><a href="#cite_ref-1" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="http://www.kmle.com">The American Heritage Stedman's Medical Dictionary</a>. <a rel="nofollow" class="external text" href="http://www.kmle.com/search.php?Search=heart">"<i>KMLE Medical Dictionary Definition of heart</i>"</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=KMLE+Medical+Dictionary+Definition+of+heart&amp;rft.au=%5Bhttp%3A%2F%2Fwww.kmle.com+The+American+Heritage+Stedman%27s+Medical+Dictionary%5D&amp;rft_id=http%3A%2F%2Fwww.kmle.com%2Fsearch.php%3FSearch%3Dheart&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3AHeart" class="Z3988"></span><style data-mw-deduplicate="TemplateStyles:r1594688">.mw-parser-output cite.citation{font-style:inherit}.mw-parser-output q{quotes:"\"""\"""'""'"}.mw-parser-output code.cs1-code{color:inherit;background:inherit;border:inherit;padding:inherit}.mw-parser-output .cs1-lock-free a{background:url("https://upload.wikimedia.org/wikipedia/commons/thumb/6/65/Lock-green.svg/9px-Lock-green.svg.png")no-repeat;background-position:right .1em center}.mw-parser-output .cs1-lock-limited a,.mw-parser-output .cs1-lock-registration a{background:url("https://upload.wikimedia.org/wikipedia/commons/thumb/d/d6/Lock-gray-alt-2.svg/9px-Lock-gray-alt-2.svg.png")no-repeat;background-position:right .1em center}.mw-parser-output .cs1-lock-subscription a{background:url("https://upload.wikimedia.org/wikipedia/commons/thumb/a/aa/Lock-red-alt-2.svg/9px-Lock-red-alt-2.svg.png")no-repeat;background-position:right .1em center}.mw-parser-output .cs1-subscription,.mw-parser-output .cs1-registration{color:#555}.mw-parser-output .cs1-subscription span,.mw-parser-output .cs1-registration span{border-bottom:1px dotted;cursor:help}.mw-parser-output .cs1-hidden-error{display:none;font-size:100%}.mw-parser-output .cs1-visible-error{display:none;font-size:100%}.mw-parser-output .cs1-subscription,.mw-parser-output .cs1-registration,.mw-parser-output .cs1-format{font-size:95%}.mw-parser-output .cs1-kern-left,.mw-parser-output .cs1-kern-wl-left{padding-left:0.2em}.mw-parser-output .cs1-kern-right,.mw-parser-output .cs1-kern-wl-right{padding-right:0.2em}</style>.</span>
</li>
<li id="cite_note-2"><span class="mw-cite-backlink"><a href="#cite_ref-2" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="http://www.kmle.com">The American Heritage Stedman's Medical Dictionary</a>. <a rel="nofollow" class="external text" href="http://www.kmle.com/search.php?Search=cardiac">"<i>KMLE Medical Dictionary Definition of cardiac</i>"</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=KMLE+Medical+Dictionary+Definition+of+cardiac&amp;rft.au=%5Bhttp%3A%2F%2Fwww.kmle.com+The+American+Heritage+Stedman%27s+Medical+Dictionary%5D&amp;rft_id=http%3A%2F%2Fwww.kmle.com%2Fsearch.php%3FSearch%3Dcardiac&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3AHeart" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-3"><span class="mw-cite-backlink"><a href="#cite_ref-3" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><a rel="nofollow" class="external free" href="http://www.obgyn.net/us/us.asp?page=/us/cotm/0001/ehr2000">http://www.obgyn.net/us/us.asp?page=/us/cotm/0001/ehr2000</a></span>
</li>
<li id="cite_note-4"><span class="mw-cite-backlink"><a href="#cite_ref-4" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text">Terry J. DuBose <a rel="nofollow" class="external free" href="http://www.obgyn.net/english/pubs/features/dubose/ehr-age.htm">http://www.obgyn.net/english/pubs/features/dubose/ehr-age.htm</a> Sex, Heart Rate and Age]</span>
</li>
</ol></div>
<h2><span class="mw-headline" id="See_also">See also</span></h2>
<ul><li><a href="/index.php/Artificial_heart" title="Artificial heart">Artificial heart</a></li>
<li><a href="/index.php/Atrium_(heart)" title="Atrium (heart)">Atrium</a></li>
<li><a href="/index.php/Blood_pressure" title="Blood pressure">Blood pressure</a></li>
<li><a href="/index.php/Cardiology" title="Cardiology">Cardiology</a></li>
<li><a href="/index.php/Cardiothoracic_Surgery" class="mw-redirect" title="Cardiothoracic Surgery">Cardiothoracic Surgery</a></li>
<li><a href="/index.php/Cardiovascular_pathology" title="Cardiovascular pathology">Cardiovascular pathology</a></li>
<li><a href="/index.php/Circulatory_system" title="Circulatory system">Circulatory system</a></li>
<li><a href="/index.php/Echocardiography" title="Echocardiography">Echocardiography</a></li>
<li><a href="/index.php/Electrical_conduction_system_of_the_heart" title="Electrical conduction system of the heart">Electrical conduction system of the heart</a></li>
<li><a href="/index.php/Haemodynamics" class="mw-redirect" title="Haemodynamics">Haemodynamics</a></li>
<li><a href="/index.php/Heart_cancer" class="mw-redirect" title="Heart cancer">Heart cancer</a></li>
<li><a href="/index.php/Congenital_heart_defect" class="mw-redirect" title="Congenital heart defect">Heart defects</a></li>
<li><a href="/index.php/Heart_rate" title="Heart rate">Heart rate</a></li>
<li><a href="/index.php/Heart_transplantation" title="Heart transplantation">Heart transplantation</a></li>
<li><a href="/index.php/Heart_valve" title="Heart valve">Heart valve</a></li>
<li><a href="/index.php/Human_anatomy" title="Human anatomy">Human anatomy</a></li>
<li><a href="/index.php/Pulse" title="Pulse">Pulse</a></li>
<li><a href="/index.php/Ventricle_(heart)" title="Ventricle (heart)">Ventricle</a></li>
<li><a href="/index.php/Aorta" title="Aorta">Aorta</a></li>
<li><a href="/index.php/Ventricular_hypertrophy" title="Ventricular hypertrophy">Ventricular hypertrophy</a></li>
<li><a href="/index.php/Holiday_heart_syndrome" title="Holiday heart syndrome">Holiday heart syndrome</a></li>
<li>Circle map — simplified mathematical model of the beating heart.</li>
<li><a href="/index.php/MUGA_scan" title="MUGA scan">MUGA scan</a></li>
<li><a href="/index.php/Cardiac_stress_test" title="Cardiac stress test">Cardiac stress test</a></li></ul>
<h2><span class="mw-headline" id="External_links">External links</span></h2>
<ul><li><a rel="nofollow" class="external text" href="http://www.emedicine.com/ped/topic2902.htm">eMedicine: Surgical anatomy of the heart</a></li>
<li><a rel="nofollow" class="external text" href="http://library.thinkquest.org/C003758/home.htm">Very Comprehensive Heart Site</a></li>
<li><a rel="nofollow" class="external text" href="http://www.abc.net.au/sydney/stories/s1572459.htm">Self Improvement Wednesday</a> - ABC 702 Drive audio</li>
<li><a rel="nofollow" class="external text" href="http://www.emc.maricopa.edu/faculty/farabee/BIOBK/BioBookcircSYS.html">The circulatory system</a></li>
<li><a rel="nofollow" class="external text" href="http://www.easyrashi.com/TEFILLIN/sternum.htm">The position of the heart</a></li>
<li><a rel="nofollow" class="external text" href="http://thevirtualheart.org/anatomyindex.html">Interactive 3D heart</a> This realistic heart can be rotated, and all its components can be studied from any angle.</li></ul>
<div role="navigation" class="navbox" aria-labelledby="Anatomy_of_the_heart" style="padding:3px"><table class="nowraplinks collapsible autocollapsed navbox-inner" style="border-spacing:0;background:transparent;color:inherit"><tbody><tr><th scope="col" class="navbox-title" colspan="2"><div class="plainlinks hlist navbar mini"><ul><li class="nv-view"><a href="/index.php?title=Template:Heart_anatomy&amp;action=edit&amp;redlink=1" class="new" title="Template:Heart anatomy (page does not exist)"><abbr title="View this template" style=";;background:none transparent;border:none;-moz-box-shadow:none;-webkit-box-shadow:none;box-shadow:none; padding:0;">v</abbr></a></li><li class="nv-talk"><a href="/index.php?title=Template_talk:Heart_anatomy&amp;action=edit&amp;redlink=1" class="new" title="Template talk:Heart anatomy (page does not exist)"><abbr title="Discuss this template" style=";;background:none transparent;border:none;-moz-box-shadow:none;-webkit-box-shadow:none;box-shadow:none; padding:0;">t</abbr></a></li><li class="nv-edit"><a rel="nofollow" class="external text" href="https://www.wikidoc.org/index.php?title=Template:Heart_anatomy&amp;action=edit"><abbr title="Edit this template" style=";;background:none transparent;border:none;-moz-box-shadow:none;-webkit-box-shadow:none;box-shadow:none; padding:0;">e</abbr></a></li></ul></div><div id="Anatomy_of_the_heart" style="font-size:114%;margin:0 4em">Anatomy of the <a class="mw-selflink selflink">heart</a></div></th></tr><tr><th scope="row" class="navbox-group" style="width:1%">General</th><td class="navbox-list navbox-odd hlist" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"></div><table class="nowraplinks navbox-subgroup" style="border-spacing:0"><tbody><tr><th scope="row" class="navbox-group" style="width:1%">Surface</th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em">
<ul><li><a href="/index.php/Base_of_the_heart" title="Base of the heart">base</a></li>
<li><a href="/index.php/Apex_of_the_heart" title="Apex of the heart">apex</a></li>
<li><a href="/index.php?title=Sulcus_(morphology)&amp;action=edit&amp;redlink=1" class="new" title="Sulcus (morphology) (page does not exist)">sulci</a>
<ul><li><a href="/index.php/Coronary_sulcus" title="Coronary sulcus">coronary</a></li>
<li><a href="/index.php?title=Interatrial_sulcus&amp;action=edit&amp;redlink=1" class="new" title="Interatrial sulcus (page does not exist)">interatrial</a></li>
<li><a href="/index.php/Anterior_interventricular_sulcus" title="Anterior interventricular sulcus">anterior interventricular</a></li>
<li><a href="/index.php/Posterior_interventricular_sulcus" title="Posterior interventricular sulcus">posterior interventricular</a></li></ul></li>
<li>borders
<ul><li><a href="/index.php/Right_border_of_heart" title="Right border of heart">right</a></li>
<li><a href="/index.php/Left_margin_of_heart" title="Left margin of heart">left</a></li></ul></li></ul>
</div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%">Internal</th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"></div><table class="nowraplinks navbox-subgroup" style="border-spacing:0"><tbody><tr><td colspan="2" class="navbox-list navbox-even" style="width:100%;padding:0px"><div style="padding:0em 0.25em">
<ul><li><a href="/index.php/Atrium_(heart)" title="Atrium (heart)">atria</a>
<ul><li><a href="/index.php/Interatrial_septum" title="Interatrial septum">interatrial septum</a></li>
<li><a href="/index.php/Pectinate_muscles" class="mw-redirect" title="Pectinate muscles">pectinate muscles</a></li>
<li><a href="/index.php/Terminal_sulcus_(heart)" title="Terminal sulcus (heart)">terminal sulcus</a></li></ul></li>
<li><a href="/index.php/Ventricle_(heart)" title="Ventricle (heart)">ventricles</a>
<ul><li><a href="/index.php/Interventricular_septum" title="Interventricular septum">interventricular septum</a></li>
<li><a href="/index.php/Trabeculae_carneae" title="Trabeculae carneae">trabeculae carneae</a></li>
<li><a href="/index.php/Chordae_tendineae" title="Chordae tendineae">chordae tendineae</a></li>
<li><a href="/index.php/Papillary_muscle" title="Papillary muscle">papillary muscle</a></li></ul></li>
<li><a href="/index.php/Heart_valve" title="Heart valve">valves</a></li>
<li><a href="/index.php/Cusps_of_heart_valves" title="Cusps of heart valves">cusps</a></li>
<li><a href="/index.php?title=Atrioventricular_septum&amp;action=edit&amp;redlink=1" class="new" title="Atrioventricular septum (page does not exist)">atrioventricular septum</a></li></ul>
</div></td></tr><tr><td colspan="2" class="navbox-list navbox-odd" style="width:100%;padding:0px"><div style="padding:0em 0.25em">
<ul><li><a href="/index.php/Cardiac_skeleton" title="Cardiac skeleton">cardiac skeleton</a></li>
<li><a href="/index.php/Intervenous_tubercle" title="Intervenous tubercle">intervenous tubercle</a></li></ul>
</div></td></tr></tbody></table><div></div></td></tr></tbody></table><div></div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%">Chambers</th><td class="navbox-list navbox-odd hlist" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"></div><table class="nowraplinks navbox-subgroup" style="border-spacing:0"><tbody><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/index.php/Right_heart" title="Right heart">Right heart</a></th><td class="navbox-list navbox-even" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em">
<ul><li><i>(<a href="/index.php/Venae_cavae" class="mw-redirect" title="Venae cavae">venae cavae</a>, <a href="/index.php/Coronary_sinus" title="Coronary sinus">coronary sinus</a>)</i> → <a href="/index.php/Right_atrium" title="Right atrium">right atrium</a> (<a href="/index.php?title=Right_atrial_appendage&amp;action=edit&amp;redlink=1" class="new" title="Right atrial appendage (page does not exist)">atrial appendage</a>, <a href="/index.php/Fossa_ovalis_(heart)" title="Fossa ovalis (heart)">fossa ovalis</a>, <a href="/index.php/Limbus_of_fossa_ovalis" title="Limbus of fossa ovalis">limbus of fossa ovalis</a>, <a href="/index.php/Crista_terminalis" title="Crista terminalis">crista terminalis</a>, <a href="/index.php?title=Valve_of_inferior_vena_cava&amp;action=edit&amp;redlink=1" class="new" title="Valve of inferior vena cava (page does not exist)">valve of inferior vena cava</a>, <a href="/index.php?title=Valve_of_coronary_sinus&amp;action=edit&amp;redlink=1" class="new" title="Valve of coronary sinus (page does not exist)">valve of coronary sinus</a>) → <a href="/index.php/Tricuspid_valve" title="Tricuspid valve">tricuspid valve</a> → <a href="/index.php/Right_ventricle" title="Right ventricle">right ventricle</a> (<a href="/index.php?title=Infundibulum_(heart)&amp;action=edit&amp;redlink=1" class="new" title="Infundibulum (heart) (page does not exist)">infundibulum</a>, <a href="/index.php?title=Moderator_band_(heart)&amp;action=edit&amp;redlink=1" class="new" title="Moderator band (heart) (page does not exist)">moderator band/septomarginal trabecula</a>, <a href="/index.php?title=Crista_supraventricularis&amp;action=edit&amp;redlink=1" class="new" title="Crista supraventricularis (page does not exist)">crista supraventricularis</a>),  → <a href="/index.php/Pulmonary_valve" title="Pulmonary valve">pulmonary valve</a> → <i>(<a href="/index.php/Pulmonary_artery" title="Pulmonary artery">pulmonary artery</a> and <a href="/index.php/Pulmonary_circulation" title="Pulmonary circulation">pulmonary circulation</a>)</i></li></ul>
</div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/index.php/Left_heart" title="Left heart">Left heart</a></th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em">
<ul><li><i>(<a href="/index.php/Pulmonary_vein" class="mw-redirect" title="Pulmonary vein">pulmonary veins</a>)</i> → <a href="/index.php/Left_atrium" title="Left atrium">left atrium</a> (<a href="/index.php/Left_atrial_appendage" class="mw-redirect" title="Left atrial appendage">atrial appendage</a>)  → <a href="/index.php/Mitral_valve" title="Mitral valve">mitral valve</a> → <a href="/index.php/Left_ventricle" title="Left ventricle">left ventricle</a> → <a href="/index.php/Aortic_valve" title="Aortic valve">aortic valve</a> (<a href="/index.php/Aortic_sinus" title="Aortic sinus">aortic sinus</a>) → <i>(<a href="/index.php/Aorta" title="Aorta">aorta</a> and <a href="/index.php/Systemic_circulation" title="Systemic circulation">systemic circulation</a>)</i></li></ul>
</div></td></tr></tbody></table><div></div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%">Layers</th><td class="navbox-list navbox-odd hlist" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"></div><table class="nowraplinks navbox-subgroup" style="border-spacing:0"><tbody><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/index.php/Endocardium" title="Endocardium">Endocardium</a></th><td class="navbox-list navbox-even" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em">
<ul><li><a href="/index.php/Heart_valve" title="Heart valve">heart valves</a></li></ul>
</div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/index.php/Cardiac_muscle" title="Cardiac muscle">Myocardium</a></th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em">
<ul><li><a href="/index.php/Electrical_conduction_system_of_the_heart" title="Electrical conduction system of the heart">Conduction system</a>
<ul><li><a href="/index.php/Cardiac_pacemaker" title="Cardiac pacemaker">cardiac pacemaker</a></li>
<li><a href="/index.php/Sinoatrial_node" title="Sinoatrial node">SA node</a></li>
<li><a href="/index.php/Bachmann%27s_bundle" title="Bachmann's bundle">Bachmann's bundle</a></li>
<li><a href="/index.php/Atrioventricular_node" title="Atrioventricular node">AV node</a></li>
<li><a href="/index.php/Bundle_of_His" title="Bundle of His">bundle of His</a></li>
<li><a href="/index.php?title=Bundle_branches&amp;action=edit&amp;redlink=1" class="new" title="Bundle branches (page does not exist)">bundle branches</a></li>
<li><a href="/index.php/Purkinje_fibers" title="Purkinje fibers">Purkinje fibers</a></li></ul></li></ul>
</div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/index.php/Pericardium" title="Pericardium">Pericardial cavity</a></th><td class="navbox-list navbox-even" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em">
<ul><li><a href="/index.php/Pericardial_sinus" title="Pericardial sinus">pericardial sinus</a></li></ul>
</div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/index.php/Pericardium" title="Pericardium">Pericardium</a></th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em">
<ul><li><a href="/index.php/Fibrous_pericardium" title="Fibrous pericardium">fibrous pericardium</a>
<ul><li><a href="/index.php?title=Sternopericardial_ligaments&amp;action=edit&amp;redlink=1" class="new" title="Sternopericardial ligaments (page does not exist)">sternopericardial ligaments</a></li></ul></li>
<li><a href="/index.php/Serous_pericardium" title="Serous pericardium">serous pericardium</a>
<ul><li><a href="/index.php/Epicardium" title="Epicardium">epicardium/visceral layer</a></li></ul></li>
<li><a href="/index.php?title=Fold_of_left_vena_cava&amp;action=edit&amp;redlink=1" class="new" title="Fold of left vena cava (page does not exist)">fold of left vena cava</a></li></ul>
</div></td></tr></tbody></table><div></div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%">Blood supply</th><td class="navbox-list navbox-even hlist" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em">
<ul><li><a href="/index.php/Circulatory_system" title="Circulatory system">Circulatory system</a></li>
<li><a href="/index.php/Coronary_circulation" title="Coronary circulation">Coronary circulation</a></li>
<li><a href="/index.php/Coronary_arteries" title="Coronary arteries">Coronary arteries</a></li></ul>
</div></td></tr></tbody></table></div>
<div role="navigation" class="navbox" aria-labelledby="Arteries_and_veins" style="padding:3px"><table class="nowraplinks collapsible autocollapse navbox-inner" style="border-spacing:0;background:transparent;color:inherit"><tbody><tr><th scope="col" class="navbox-title" colspan="2"><div class="plainlinks hlist navbar mini"><ul><li class="nv-view"><a href="/index.php/Template:Arteries_and_veins" title="Template:Arteries and veins"><abbr title="View this template" style=";;background:none transparent;border:none;-moz-box-shadow:none;-webkit-box-shadow:none;box-shadow:none; padding:0;">v</abbr></a></li><li class="nv-talk"><a href="/index.php?title=Template_talk:Arteries_and_veins&amp;action=edit&amp;redlink=1" class="new" title="Template talk:Arteries and veins (page does not exist)"><abbr title="Discuss this template" style=";;background:none transparent;border:none;-moz-box-shadow:none;-webkit-box-shadow:none;box-shadow:none; padding:0;">t</abbr></a></li><li class="nv-edit"><a rel="nofollow" class="external text" href="https://www.wikidoc.org/index.php?title=Template:Arteries_and_veins&amp;action=edit"><abbr title="Edit this template" style=";;background:none transparent;border:none;-moz-box-shadow:none;-webkit-box-shadow:none;box-shadow:none; padding:0;">e</abbr></a></li></ul></div><div id="Arteries_and_veins" style="font-size:114%;margin:0 4em"><a href="/index.php/Artery" title="Artery">Arteries</a> and <a href="/index.php/Vein" title="Vein">veins</a></div></th></tr><tr><th scope="row" class="navbox-group" style="width:1%">Vessels</th><td class="navbox-list navbox-odd hlist" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"></div><table class="nowraplinks navbox-subgroup" style="border-spacing:0"><tbody><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/index.php/Artery" title="Artery">Arteries</a></th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em">
<ul><li><a href="/index.php/Nutrient_artery" title="Nutrient artery">Nutrient artery</a></li>
<li><a href="/index.php/Arteriole" title="Arteriole">Arteriole</a>
<ul><li><a href="/index.php/Metarteriole" title="Metarteriole">Metarteriole</a></li></ul></li>
<li><a href="/index.php/Elastic_artery" title="Elastic artery">Elastic artery</a></li></ul>
</div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/index.php/Capillary" title="Capillary">Capillaries</a></th><td class="navbox-list navbox-even" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em">
<ul><li>Types
<ul><li><a href="/index.php/Capillary#Types" title="Capillary">Continuous</a></li>
<li><a href="/index.php/Capillary#Types" title="Capillary">Fenestrated</a></li>
<li><a href="/index.php/Sinusoid_(blood_vessel)" title="Sinusoid (blood vessel)">Sinusoidal</a></li></ul></li>
<li><a href="/index.php/Precapillary_sphincter" title="Precapillary sphincter">Precapillary sphincter</a></li>
<li><a href="/index.php?title=Precapillary_resistance&amp;action=edit&amp;redlink=1" class="new" title="Precapillary resistance (page does not exist)">Precapillary resistance</a></li></ul>
</div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/index.php/Vein" title="Vein">Veins</a></th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em">
<ul><li><a href="/index.php/Vena_comitans" title="Vena comitans">Vena comitans</a></li>
<li><a href="/index.php/Superficial_vein" title="Superficial vein">Superficial vein</a></li>
<li><a href="/index.php/Deep_vein" title="Deep vein">Deep vein</a></li>
<li><a href="/index.php/Emissary_veins" title="Emissary veins">Emissary veins</a></li>
<li><a href="/index.php/Venous_plexus" title="Venous plexus">Venous plexus</a></li>
<li><a href="/index.php/Venule" title="Venule">Venule</a></li></ul>
</div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/index.php/Lymph" class="mw-redirect" title="Lymph">Lymph</a></th><td class="navbox-list navbox-even" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em">
<ul><li><a href="/index.php/Lymphatic_vessel" class="mw-redirect" title="Lymphatic vessel">Lymphatic vessel</a></li>
<li><a href="/index.php/Lymph" class="mw-redirect" title="Lymph">Lymph</a></li>
<li><a href="/index.php/Lymph_capillary" title="Lymph capillary">Lymph capillary</a></li></ul>
</div></td></tr></tbody></table><div></div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/index.php/Circulatory_system" title="Circulatory system">Circulatory system</a></th><td class="navbox-list navbox-odd hlist" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"></div><table class="nowraplinks navbox-subgroup" style="border-spacing:0"><tbody><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/index.php/Circulatory_system#Systemic_circulation" title="Circulatory system">Systemic</a></th><td class="navbox-list navbox-odd" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"><a class="mw-selflink selflink">Left heart</a> → <a href="/index.php/Aorta" title="Aorta">Aorta</a> → <a href="/index.php/Artery" title="Artery">Arteries</a> → <a href="/index.php/Arteriole" title="Arteriole">Arterioles</a> → <a href="/index.php/Capillary" title="Capillary">Capillaries</a> → <a href="/index.php/Venule" title="Venule">Venules</a> → <a href="/index.php/Vein" title="Vein">Veins</a> → <a href="/index.php/Venae_cavae" class="mw-redirect" title="Venae cavae">Vena cava</a> → (<a class="mw-selflink selflink">Right heart</a>)</div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/index.php/Pulmonary_circulation" title="Pulmonary circulation">Pulmonary</a></th><td class="navbox-list navbox-even" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em"><a class="mw-selflink selflink">Right heart</a> → <a href="/index.php/Pulmonary_artery" title="Pulmonary artery">Pulmonary arteries</a> → <a href="/index.php/Lung" title="Lung">Lungs</a> → <a href="/index.php/Pulmonary_vein" class="mw-redirect" title="Pulmonary vein">Pulmonary vein</a> → <a class="mw-selflink selflink">Left heart</a></div></td></tr></tbody></table><div></div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/index.php/Histology" title="Histology">Microanatomy</a></th><td class="navbox-list navbox-odd hlist" style="text-align:left;border-left-width:2px;border-left-style:solid;width:100%;padding:0px"><div style="padding:0em 0.25em">
<ul><li><a href="/index.php/Microvessel" title="Microvessel">Microvessel</a></li>
<li><a href="/index.php/Microcirculation" title="Microcirculation">Microcirculation</a></li>
<li><a href="/index.php/Tunica_intima" title="Tunica intima">Tunica intima</a>
<ul><li><a href="/index.php/Endothelium" title="Endothelium">Endothelium</a></li>
<li><a href="/index.php/Internal_elastic_lamina" title="Internal elastic lamina">Internal elastic lamina</a></li></ul></li>
<li><a href="/index.php/Tunica_media" title="Tunica media">Tunica media</a></li>
<li><a href="/index.php/Tunica_externa" class="mw-redirect" title="Tunica externa">Tunica externa</a></li>
<li><a href="/index.php/Vasa_vasorum" title="Vasa vasorum">Vasa vasorum</a></li>
<li><a href="/index.php/Vasa_nervorum" title="Vasa nervorum">Vasa nervorum</a></li>
<li><a href="/index.php/Vascular_nerves" title="Vascular nerves">Vascular nerves</a></li>
<li><a href="/index.php/Rete_mirabile" title="Rete mirabile">Rete mirabile</a></li>
<li><a href="/index.php/Circulatory_anastomosis" title="Circulatory anastomosis">Circulatory anastomosis</a></li></ul>
</div></td></tr></tbody></table></div>
<p><br>
<a href="/index.php?title=Template:Link_FA&amp;action=edit&amp;redlink=1" class="new" title="Template:Link FA (page does not exist)">Template:Link FA</a>
<a href="/index.php?title=Template:Link_FA&amp;action=edit&amp;redlink=1" class="new" title="Template:Link FA (page does not exist)">Template:Link FA</a>
</p><p><a href="/index.php?title=Template:WikiDoc_Sources&amp;action=edit&amp;redlink=1" class="new" title="Template:WikiDoc Sources (page does not exist)">Template:WikiDoc Sources</a>
</p>
<!-- 
NewPP limit report
Cached time: 20240821172504
Cache expiry: 1209600
Reduced expiry: false
Complications: [no‐toc‐conversion]
CPU time usage: 0.112 seconds
Real time usage: 0.241 seconds
Preprocessor visited node count: 764/1000000
Post‐expand include size: 64032/2097152 bytes
Template argument size: 86/2097152 bytes
Highest expansion depth: 8/100
Expensive parser function count: 0/100
Unstrip recursion depth: 1/20
Unstrip post‐expand size: 7927/5000000 bytes
Lua time usage: 0.090/7 seconds
Lua virtual size: 8048640/52428800 bytes
Lua estimated memory usage: 0 bytes
-->
<!--
Transclusion expansion time report (%,ms,calls,template)
100.00%  202.750      1 -total
 60.89%  123.464      8 Template:Navbox
 34.22%   69.383      1 Template:Reflist
 27.72%   56.199      2 Template:Cite_web
 24.05%   48.763      1 Template:Heart
 16.38%   33.219      1 Template:Main
 14.74%   29.892      1 Template:Cardiovascular_system
  2.34%    4.738      1 Template:SI
  1.93%    3.920      1 Template:CMG
  1.23%    2.500      1 Template:Main_other
-->

<!-- Saved in parser cache with key wikidb-wd_:pcache:idhash:17028-0!canonical and timestamp 20240821172504 and revision id 904517. Rendering was triggered because: page-view
 -->
</div>"""
```

```python
practiceHtml = """<div class="mw-parser-output"><p><b>Editor-In-Chief:</b> <a href="/index.php/User:C_Michael_Gibson" title="User:C Michael Gibson">C. Michael Gibson, M.S., M.D.</a> <a rel="nofollow" class="external autonumber" href="mailto:charlesmichaelgibson@gmail.com">[1]</a>
</p><p><br>
</p>
<div id="toc" class="toc" role="navigation" aria-labelledby="mw-toc-heading"><input type="checkbox" role="button" id="toctogglecheckbox" class="toctogglecheckbox" style="display:none"><div class="toctitle" lang="en" dir="ltr"><h2 id="mw-toc-heading">Contents</h2><span class="toctogglespan"><label class="toctogglelabel" for="toctogglecheckbox"></label></span></div>
<ul>
<li class="toclevel-1 tocsection-1"><a href="#Overview"><span class="tocnumber">1</span> <span class="toctext">Overview</span></a></li>
<li class="toclevel-1 tocsection-2"><a href="#Physical_and_chemical_properties"><span class="tocnumber">2</span> <span class="toctext">Physical and chemical properties</span></a>
<ul>
<li class="toclevel-2 tocsection-3"><a href="#Major_and_minor_grooves"><span class="tocnumber">2.1</span> <span class="toctext">Major and minor grooves</span></a></li>
<li class="toclevel-2 tocsection-4"><a href="#Base_pairing"><span class="tocnumber">2.2</span> <span class="toctext">Base pairing</span></a></li>
<li class="toclevel-2 tocsection-5"><a href="#Sense_and_antisense"><span class="tocnumber">2.3</span> <span class="toctext">Sense and antisense</span></a></li>
<li class="toclevel-2 tocsection-6"><a href="#Supercoiling"><span class="tocnumber">2.4</span> <span class="toctext">Supercoiling</span></a></li>
<li class="toclevel-2 tocsection-7"><a href="#Alternative_double-helical_structures"><span class="tocnumber">2.5</span> <span class="toctext">Alternative double-helical structures</span></a></li>
<li class="toclevel-2 tocsection-8"><a href="#Quadruplex_structures"><span class="tocnumber">2.6</span> <span class="toctext">Quadruplex structures</span></a></li>
</ul>
</li>
<li class="toclevel-1 tocsection-9"><a href="#Chemical_modifications"><span class="tocnumber">3</span> <span class="toctext">Chemical modifications</span></a>
<ul>
<li class="toclevel-2 tocsection-10"><a href="#Base_modifications"><span class="tocnumber">3.1</span> <span class="toctext">Base modifications</span></a></li>
<li class="toclevel-2 tocsection-11"><a href="#DNA_damage"><span class="tocnumber">3.2</span> <span class="toctext">DNA damage</span></a></li>
</ul>
</li>
<li class="toclevel-1 tocsection-12"><a href="#Overview_of_biological_functions"><span class="tocnumber">4</span> <span class="toctext">Overview of biological functions</span></a>
<ul>
<li class="toclevel-2 tocsection-13"><a href="#Genes_and_genomes"><span class="tocnumber">4.1</span> <span class="toctext">Genes and genomes</span></a></li>
<li class="toclevel-2 tocsection-14"><a href="#Transcription_and_translation"><span class="tocnumber">4.2</span> <span class="toctext">Transcription and translation</span></a></li>
<li class="toclevel-2 tocsection-15"><a href="#Replication"><span class="tocnumber">4.3</span> <span class="toctext">Replication</span></a></li>
</ul>
</li>
<li class="toclevel-1 tocsection-16"><a href="#Interactions_with_proteins"><span class="tocnumber">5</span> <span class="toctext">Interactions with proteins</span></a>
<ul>
<li class="toclevel-2 tocsection-17"><a href="#DNA-binding_proteins"><span class="tocnumber">5.1</span> <span class="toctext">DNA-binding proteins</span></a></li>
<li class="toclevel-2 tocsection-18"><a href="#DNA-modifying_enzymes"><span class="tocnumber">5.2</span> <span class="toctext">DNA-modifying enzymes</span></a>
<ul>
<li class="toclevel-3 tocsection-19"><a href="#Nucleases_and_ligases"><span class="tocnumber">5.2.1</span> <span class="toctext">Nucleases and ligases</span></a></li>
<li class="toclevel-3 tocsection-20"><a href="#Topoisomerases_and_helicases"><span class="tocnumber">5.2.2</span> <span class="toctext">Topoisomerases and helicases</span></a></li>
<li class="toclevel-3 tocsection-21"><a href="#Polymerases"><span class="tocnumber">5.2.3</span> <span class="toctext">Polymerases</span></a></li>
</ul>
</li>
</ul>
</li>
<li class="toclevel-1 tocsection-22"><a href="#Genetic_recombination"><span class="tocnumber">6</span> <span class="toctext">Genetic recombination</span></a></li>
<li class="toclevel-1 tocsection-23"><a href="#Evolution_of_DNA_metabolism"><span class="tocnumber">7</span> <span class="toctext">Evolution of DNA metabolism</span></a></li>
<li class="toclevel-1 tocsection-24"><a href="#Uses_in_technology"><span class="tocnumber">8</span> <span class="toctext">Uses in technology</span></a>
<ul>
<li class="toclevel-2 tocsection-25"><a href="#Genetic_engineering"><span class="tocnumber">8.1</span> <span class="toctext">Genetic engineering</span></a></li>
<li class="toclevel-2 tocsection-26"><a href="#Forensics"><span class="tocnumber">8.2</span> <span class="toctext">Forensics</span></a></li>
<li class="toclevel-2 tocsection-27"><a href="#Bioinformatics"><span class="tocnumber">8.3</span> <span class="toctext">Bioinformatics</span></a></li>
<li class="toclevel-2 tocsection-28"><a href="#DNA_nanotechnology"><span class="tocnumber">8.4</span> <span class="toctext">DNA nanotechnology</span></a></li>
<li class="toclevel-2 tocsection-29"><a href="#History_and_anthropology"><span class="tocnumber">8.5</span> <span class="toctext">History and anthropology</span></a></li>
</ul>
</li>
<li class="toclevel-1 tocsection-30"><a href="#History_of_DNA_research"><span class="tocnumber">9</span> <span class="toctext">History of DNA research</span></a></li>
<li class="toclevel-1 tocsection-31"><a href="#See_also"><span class="tocnumber">10</span> <span class="toctext">See also</span></a></li>
<li class="toclevel-1 tocsection-32"><a href="#References"><span class="tocnumber">11</span> <span class="toctext">References</span></a></li>
<li class="toclevel-1 tocsection-33"><a href="#Further_reading"><span class="tocnumber">12</span> <span class="toctext">Further reading</span></a></li>
<li class="toclevel-1 tocsection-34"><a href="#External_links"><span class="tocnumber">13</span> <span class="toctext">External links</span></a></li>
</ul>
</div>

<h2><span class="mw-headline" id="Overview">Overview</span></h2>
<figure typeof="mw:File/Thumb"><a href="/index.php/File:DNA_Overview.png" class="mw-file-description"><img src="/images/f/f0/DNA_Overview.png" decoding="async" width="220" height="550"></a><figcaption>The structure of part of a DNA double helix</figcaption></figure>
<p><b>Deoxyribonucleic acid</b> (<b>DNA</b>) is a <a href="/index.php/Nucleic_acid" title="Nucleic acid">nucleic acid</a> that contains the <a href="/index.php/Genetics" title="Genetics">genetic</a> instructions used in the <a href="/index.php/Developmental_biology" title="Developmental biology">development</a> and functioning of all known <a href="/index.php/Life" title="Life">living organisms</a> and some <a href="/index.php/Viruses" class="mw-redirect" title="Viruses">viruses</a>. The main role of DNA <a href="/index.php/Molecule" title="Molecule">molecules</a> is the long-term storage of information. DNA is often compared to a set of blueprints or a recipe, since it contains the instructions needed to construct other components of <a href="/index.php/Cell_(biology)" title="Cell (biology)">cells</a>, such as <a href="/index.php/Protein" title="Protein">proteins</a> and <a href="/index.php/RNA" title="RNA">RNA</a> molecules. The DNA segments that carry this genetic information are called <a href="/index.php/Gene" title="Gene">genes</a>, but other DNA sequences have structural purposes, or are involved in regulating the use of this genetic information.
</p><p>Chemically, DNA is a long <a href="/index.php/Polymer" title="Polymer">polymer</a> of simple units called <a href="/index.php/Nucleotide" title="Nucleotide">nucleotides</a>, with a backbone made of sugars and phosphate groups joined by <a href="/index.php/Ester" title="Ester">ester</a> bonds. Attached to each sugar is one of four types of molecules called <a href="/index.php/Nucleobase" title="Nucleobase">bases</a>. It is the sequence of these four bases along the backbone that encodes information. This information is read using the <a href="/index.php/Genetic_code" title="Genetic code">genetic code</a>, which specifies the sequence of the <a href="/index.php/Amino_acid" title="Amino acid">amino acids</a> within proteins. The code is read by copying stretches of DNA into the related nucleic acid RNA, in a process called <a href="/index.php/Transcription_(genetics)" title="Transcription (genetics)">transcription</a>. 
</p><p>Within cells, DNA is organized into structures called <a href="/index.php/Chromosome" title="Chromosome">chromosomes</a>. These chromosomes are duplicated before cells <a href="/index.php/Cell_division" title="Cell division">divide</a>, in a process called <a href="/index.php/DNA_replication" title="DNA replication">DNA replication</a>. <a href="/index.php/Eukaryote" title="Eukaryote">Eukaryotic organisms</a> (<a href="/index.php?title=Animal&amp;action=edit&amp;redlink=1" class="new" title="Animal (page does not exist)">animals</a>, <a href="/index.php/Plant" title="Plant">plants</a>, and <a href="/index.php/Fungus" title="Fungus">fungi</a>) store their DNA inside the <a href="/index.php/Cell_nucleus" title="Cell nucleus">cell nucleus</a>, while in <a href="/index.php/Prokaryote" title="Prokaryote">prokaryotes</a> (<a href="/index.php/Bacteria" title="Bacteria">bacteria</a> and <a href="/index.php/Archae" class="mw-redirect" title="Archae">archae</a>) it is found in the cell's <a href="/index.php/Cytoplasm" title="Cytoplasm">cytoplasm</a>. Within the chromosomes, <a href="/index.php/Chromatin" title="Chromatin">chromatin</a> proteins such as <a href="/index.php/Histone" title="Histone">histones</a> compact and organize DNA. These compact structures guide the interactions between DNA and other proteins, helping control which parts of the DNA are transcribed. 
</p>
<h2><span class="mw-headline" id="Physical_and_chemical_properties">Physical and chemical properties</span></h2>
<figure class="mw-halign-right" typeof="mw:File/Thumb"><a href="/index.php/File:DNA_chemical_structure_with_repositioned_labels.png" class="mw-file-description"><img src="/images/7/7c/DNA_chemical_structure_with_repositioned_labels.png" decoding="async" width="350" height="408"></a><figcaption>The chemical structure of DNA. <a href="/index.php/Hydrogen_bond" title="Hydrogen bond">Hydrogen bonds</a> are shown as dotted lines.</figcaption></figure>
<p>DNA is a long <a href="/index.php/Polymer" title="Polymer">polymer</a> made from repeating units called <a href="/index.php/Nucleotide" title="Nucleotide">nucleotides</a>.<sup id="cite_ref-Alberts_1-0" class="reference"><a href="#cite_note-Alberts-1">[1]</a></sup><sup id="cite_ref-Butler_2-0" class="reference"><a href="#cite_note-Butler-2">[2]</a></sup> The DNA chain is 22 to 26&nbsp;<a href="/index.php/%C3%85ngstr%C3%B6m" title="Ångström">Ångströms</a> wide (2.2 to 2.6&nbsp;<a href="/index.php/Nanometre" title="Nanometre">nanometres</a>), and one nucleotide unit is 3.3&nbsp;Å (0.33&nbsp;nm) long.<sup id="cite_ref-3" class="reference"><a href="#cite_note-3">[3]</a></sup> Although each individual repeating unit is very small, DNA polymers can be enormous molecules containing millions of nucleotides. For instance, the largest human chromosome, chromosome number 1, is approximately 220 million <a href="/index.php/Base_pair" title="Base pair">base pairs</a> long.<sup id="cite_ref-4" class="reference"><a href="#cite_note-4">[4]</a></sup>
</p><p>In living organisms, DNA does not usually exist as a single molecule, but instead as a tightly-associated pair of molecules.<sup id="cite_ref-Watson_5-0" class="reference"><a href="#cite_note-Watson-5">[5]</a></sup><sup id="cite_ref-berg_6-0" class="reference"><a href="#cite_note-berg-6">[6]</a></sup> These two long strands entwine like vines, in the shape of a <a href="/index.php/Helix" title="Helix">double helix</a>. The nucleotide repeats contain both the segment of the backbone of the molecule, which holds the chain together, and a base, which interacts with the other DNA strand in the helix. In general, a base linked to a sugar is called a <a href="/index.php/Nucleoside" title="Nucleoside">nucleoside</a> and a base linked to a sugar and one or more phosphate groups is called a <a href="/index.php/Nucleotide" title="Nucleotide">nucleotide</a>. If multiple nucleotides are linked together, as in DNA, this polymer is called a <a href="/index.php/Polynucleotide" title="Polynucleotide">polynucleotide</a>.<sup id="cite_ref-IUPAC_7-0" class="reference"><a href="#cite_note-IUPAC-7">[7]</a></sup>
</p><p>The backbone of the DNA strand is made from alternating <a href="/index.php/Phosphate" title="Phosphate">phosphate</a> and <a href="/index.php/Carbohydrate" title="Carbohydrate">sugar</a> residues.<sup id="cite_ref-Ghosh_8-0" class="reference"><a href="#cite_note-Ghosh-8">[8]</a></sup> The sugar in DNA is 2-deoxyribose, which is a <a href="/index.php/Pentose" title="Pentose">pentose</a> (five-<a href="/index.php/Carbon" title="Carbon">carbon</a>) sugar. The sugars are joined together by phosphate groups that form <a href="/index.php/Phosphodiester_bond" title="Phosphodiester bond">phosphodiester bonds</a> between the third and fifth carbon <a href="/index.php/Atom" title="Atom">atoms</a> of adjacent sugar rings. These asymmetric <a href="/index.php/Covalent_bond" title="Covalent bond">bonds</a> mean a strand of DNA has a direction. In a double helix the direction of the nucleotides in one strand is opposite to their direction in the other strand. This arrangement of DNA strands is called antiparallel. The asymmetric ends of DNA strands are referred to as the <a href="/index.php/Directionality_(molecular_biology)" title="Directionality (molecular biology)">5′</a> (<i>five prime</i>) and <a href="/index.php/Directionality_(molecular_biology)" title="Directionality (molecular biology)">3′</a> (<i>three prime</i>) ends, with the 5' end being that with a terminal phosphate group and the 3' end that with a terminal hydroxyl group.  One of the major differences between DNA and RNA is the sugar, with 2-deoxyribose being replaced by the alternative pentose sugar <a href="/index.php/Ribose" title="Ribose">ribose</a> in RNA.<sup id="cite_ref-berg_6-1" class="reference"><a href="#cite_note-berg-6">[6]</a></sup>
</p><p>The DNA double helix is stabilized by <a href="/index.php/Hydrogen_bond" title="Hydrogen bond">hydrogen bonds</a> between the bases attached to the two strands. The four bases found in DNA are <a href="/index.php/Adenine" title="Adenine">adenine</a> (abbreviated A), <a href="/index.php/Cytosine" title="Cytosine">cytosine</a> (C), <a href="/index.php/Guanine" title="Guanine">guanine</a> (G) and <a href="/index.php/Thymine" title="Thymine">thymine</a> (T). These four bases are attached to the sugar/phosphate to form the complete nucleotide, as shown for adenosine monophosphate.
</p><p>These bases are classified into two types; adenine and guanine are fused five- and six-membered <a href="/index.php/Heterocyclic_compound" title="Heterocyclic compound">heterocyclic compounds</a> called <a href="/index.php/Purine" title="Purine">purines</a>, while cytosine and thymine are six-membered rings called <a href="/index.php/Pyrimidine" title="Pyrimidine">pyrimidines</a>.<sup id="cite_ref-berg_6-2" class="reference"><a href="#cite_note-berg-6">[6]</a></sup> A fifth pyrimidine base, called <a href="/index.php/Uracil" title="Uracil">uracil</a> (U), usually takes the place of thymine in RNA and differs from thymine by lacking a <a href="/index.php/Methyl_group" title="Methyl group">methyl group</a> on its ring. Uracil is not usually found in DNA, occurring only as a breakdown product of cytosine. 
</p>
<h3><span class="mw-headline" id="Major_and_minor_grooves">Major and minor grooves</span></h3>
<figure class="mw-halign-right" typeof="mw:File/Frame"><a href="/index.php/File:DNA_orbit_animated_small.gif" class="mw-file-description"><img src="/images/8/87/DNA_orbit_animated_small.gif" decoding="async" width="182" height="300"></a><figcaption>Animation of the structure of a section of DNA. The bases lie horizontally between the two spiraling strands.<sup id="cite_ref-9" class="reference"><a href="#cite_note-9">[9]</a></sup></figcaption></figure>
<p>The double helix is a right-handed spiral. As the DNA strands wind around each other, they leave gaps between each set of phosphate backbones, revealing the sides of the bases inside (see animation). There are two of these grooves twisting around the surface of the double helix: one groove, the major groove, is 22&nbsp;Å wide and the other, the minor groove, is 12&nbsp;Å wide.<sup id="cite_ref-10" class="reference"><a href="#cite_note-10">[10]</a></sup> The narrowness of the minor groove means that the edges of the bases are more accessible in the major groove. As a result, proteins like <a href="/index.php/Transcription_factor" title="Transcription factor">transcription factors</a> that can bind to specific sequences in double-stranded DNA usually make contacts to the sides of the bases exposed in the major groove.<sup id="cite_ref-11" class="reference"><a href="#cite_note-11">[11]</a></sup>
</p>
<h3><span class="mw-headline" id="Base_pairing">Base pairing</span></h3>
<div role="note" class="hatnote navigation-not-searchable">Further information: [[:<a href="/index.php/Base_pair" title="Base pair">Base pair</a>]]</div>
<p>Each type of base on one strand forms a bond with just one type of base on the other strand. This is called complementary <a href="/index.php/Base_pair" title="Base pair">base pairing</a>. Here, purines form <a href="/index.php/Hydrogen_bond" title="Hydrogen bond">hydrogen bonds</a> to pyrimidines, with A bonding only to T, and C bonding only to G. This arrangement of two nucleotides binding together across the double helix is called a base pair. The double helix is also stabilized by the <a href="/index.php/Hydrophobic_effect" title="Hydrophobic effect">hydrophobic effect</a> and <a href="/index.php/Stacking_(chemistry)" title="Stacking (chemistry)">pi stacking</a>, which are not influenced by the sequence of the DNA.<sup id="cite_ref-12" class="reference"><a href="#cite_note-12">[12]</a></sup> As hydrogen bonds are not <a href="/index.php/Covalent_bond" title="Covalent bond">covalent</a>, they can be broken and rejoined relatively easily. The two strands of DNA in a double helix can therefore be pulled apart like a zipper, either by a mechanical force or high <a href="/index.php/Temperature" class="mw-redirect" title="Temperature">temperature</a>.<sup id="cite_ref-13" class="reference"><a href="#cite_note-13">[13]</a></sup> As a result of this complementarity, all the information in the double-stranded sequence of a DNA helix is duplicated on each strand, which is vital in DNA replication. Indeed, this reversible and specific interaction between complementary base pairs is critical for all the functions of DNA in living organisms.<sup id="cite_ref-Alberts_1-1" class="reference"><a href="#cite_note-Alberts-1">[1]</a></sup>
</p>
<div class="thumb tleft" style="background-color: #f9f9f9; border: 1px solid #CCCCCC; margin:0.5em;">
<table border="0" width="230" cellpadding="2" cellspacing="0" style="font-size: 85%; border: 1px solid #CCCCCC; margin: 0.3em;">
<tbody><tr>
<td><span typeof="mw:File"><a href="/index.php/File:GC_DNA_base_pair.svg" class="mw-file-description"><img src="/images/d/d7/GC_DNA_base_pair.svg" decoding="async" width="281" height="177"></a></span>
</td></tr></tbody></table>
<table border="0" width="230" cellpadding="2" cellspacing="0" style="font-size: 85%; border: 1px solid #CCCCCC; margin: 0.3em;">
<tbody><tr>
<td><span typeof="mw:File"><a href="/index.php/File:AT_DNA_base_pair.svg" class="mw-file-description"><img src="/images/9/91/AT_DNA_base_pair.svg" decoding="async" width="281" height="151"></a></span>
</td></tr></tbody></table>
<div style="border: none; width:281px;"><div class="thumbcaption">Top, a <b>GC</b> base pair with three <a href="/index.php/Hydrogen_bond" title="Hydrogen bond">hydrogen bonds</a>. Bottom, an <b>AT</b> base pair with two hydrogen bonds. Hydrogen bonds are shown as dashed lines.</div></div></div>
<p>The two types of base pairs form different numbers of hydrogen bonds, AT forming two hydrogen bonds, and GC forming three hydrogen bonds (see figures, left). The GC base pair is therefore stronger than the AT base pair. As a result, it is both the percentage of GC base pairs and the overall length of a DNA double helix that determine the strength of the association between the two strands of DNA. Long DNA helices with a high GC content have stronger-interacting strands, while short helices with high AT content have weaker-interacting strands.<sup id="cite_ref-14" class="reference"><a href="#cite_note-14">[14]</a></sup> In biology, parts of the DNA double helix that need to separate easily, such as the TATAAT <a href="/index.php/Pribnow_box" title="Pribnow box">Pribnow box</a> in some <a href="/index.php/Promoter" title="Promoter">promoters</a>, tend to have a high AT content, making the strands easier to pull apart.<sup id="cite_ref-15" class="reference"><a href="#cite_note-15">[15]</a></sup> In the laboratory, the strength of this interaction can be measured by finding the temperature required to break the hydrogen bonds, their <a href="/index.php/DNA_melting" title="DNA melting">melting temperature</a> (also called <i>T<sub>m</sub></i> value). When all the base pairs in a DNA double helix melt, the strands separate and exist in solution as two entirely independent molecules. These single-stranded DNA molecules have no single common shape, but some conformations are more stable than others.<sup id="cite_ref-16" class="reference"><a href="#cite_note-16">[16]</a></sup>
</p>
<h3><span class="mw-headline" id="Sense_and_antisense">Sense and antisense</span></h3>
<div role="note" class="hatnote navigation-not-searchable">Further information: [[:<a href="/index.php/Sense_(molecular_biology)" title="Sense (molecular biology)">Sense (molecular biology)</a>]]</div>
<p>A DNA sequence is called "sense" if its sequence is the same as that of a <a href="/index.php/Messenger_RNA" title="Messenger RNA">messenger RNA</a> copy that is translated into protein.<sup id="cite_ref-17" class="reference"><a href="#cite_note-17">[17]</a></sup> The sequence on the opposite strand is called the "antisense" sequence. Both sense and antisense sequences can exist on different parts of the same strand of DNA (i.e. both strands contain both sense and antisense sequences). In both prokaryotes and eukaryotes, antisense RNA sequences are produced, but the functions of these RNAs are not entirely clear.<sup id="cite_ref-18" class="reference"><a href="#cite_note-18">[18]</a></sup> One proposal is that antisense RNAs are involved in regulating <a href="/index.php/Gene_expression" title="Gene expression">gene expression</a> through RNA-RNA base pairing.<sup id="cite_ref-19" class="reference"><a href="#cite_note-19">[19]</a></sup>
</p><p>A few DNA sequences in prokaryotes and eukaryotes, and more in <a href="/index.php/Plasmid" title="Plasmid">plasmids</a> and <a href="/index.php/Virus" title="Virus">viruses</a>, blur the distinction between sense and antisense strands by having overlapping genes.<sup id="cite_ref-20" class="reference"><a href="#cite_note-20">[20]</a></sup> In these cases, some DNA sequences do double duty, encoding one protein when read along one strand, and a second protein when read in the opposite direction along the other strand. In <a href="/index.php/Bacteria" title="Bacteria">bacteria</a>, this overlap may be involved in the regulation of gene transcription,<sup id="cite_ref-21" class="reference"><a href="#cite_note-21">[21]</a></sup> while in viruses, overlapping genes increase the amount of information that can be encoded within the small viral genome.<sup id="cite_ref-22" class="reference"><a href="#cite_note-22">[22]</a></sup>
</p>
<h3><span class="mw-headline" id="Supercoiling">Supercoiling</span></h3>
<div role="note" class="hatnote navigation-not-searchable">Further information: [[:<a href="/index.php/DNA_supercoil" title="DNA supercoil">DNA supercoil</a>]]</div>
<p>DNA can be twisted like a rope in a process called <a href="/index.php/DNA_supercoil" title="DNA supercoil">DNA supercoiling</a>. With DNA in its "relaxed" state, a strand usually circles the axis of the double helix once every 10.4 base pairs, but if the DNA is twisted the strands become more tightly or more loosely wound.<sup id="cite_ref-23" class="reference"><a href="#cite_note-23">[23]</a></sup> If the DNA is twisted in the direction of the helix, this is positive supercoiling, and the bases are held more tightly together. If they are twisted in the opposite direction, this is negative supercoiling, and the bases come apart more easily. In nature, most DNA has slight negative supercoiling that is introduced by <a href="/index.php/Enzyme" title="Enzyme">enzymes</a> called <a href="/index.php/Topoisomerase" title="Topoisomerase">topoisomerases</a>.<sup id="cite_ref-Champoux_24-0" class="reference"><a href="#cite_note-Champoux-24">[24]</a></sup> These enzymes are also needed to relieve the twisting stresses introduced into DNA strands during processes such as <a href="/index.php/Transcription_(genetics)" title="Transcription (genetics)">transcription</a> and <a href="/index.php/DNA_replication" title="DNA replication">DNA replication</a>.<sup id="cite_ref-Wang_25-0" class="reference"><a href="#cite_note-Wang-25">[25]</a></sup>
</p>
<figure class="mw-halign-right" typeof="mw:File/Thumb"><a href="/index.php/File:A-DNA,_B-DNA_and_Z-DNA.png" class="mw-file-description"><img src="/images/b/b1/A-DNA%2C_B-DNA_and_Z-DNA.png" decoding="async" width="290" height="189"></a><figcaption>From left to right, the structures of A, B and Z DNA</figcaption></figure>
<h3><span class="mw-headline" id="Alternative_double-helical_structures">Alternative double-helical structures</span></h3>
<div role="note" class="hatnote navigation-not-searchable">Further information: [[:<a href="/index.php/Mechanical_properties_of_DNA" title="Mechanical properties of DNA">Mechanical properties of DNA</a>]]</div>
<p>DNA exists in many possible <a href="/index.php/Conformational_isomerism" title="Conformational isomerism">conformations</a>.<sup id="cite_ref-Ghosh_8-1" class="reference"><a href="#cite_note-Ghosh-8">[8]</a></sup> However, only <a href="/index.php/A-DNA" title="A-DNA">A-DNA</a>, B-DNA, and <a href="/index.php/Z-DNA" title="Z-DNA">Z-DNA</a> have been observed in organisms. Which conformation DNA adopts depends on the sequence of the DNA, the amount and direction of supercoiling, chemical modifications of the bases and also solution conditions, such as the concentration of <a href="/index.php/Metal" title="Metal">metal</a> <a href="/index.php/Ion" title="Ion">ions</a> and <a href="/index.php/Polyamine" title="Polyamine">polyamines</a>.<sup id="cite_ref-26" class="reference"><a href="#cite_note-26">[26]</a></sup> Of these three conformations, the "B" form described above is most common under the conditions found in cells.<sup id="cite_ref-27" class="reference"><a href="#cite_note-27">[27]</a></sup> The two alternative double-helical forms of DNA differ in their geometry and dimensions.
</p><p>The A form is a wider right-handed spiral, with a shallow, wide minor groove and a narrower, deeper major groove. The A form occurs under non-physiological conditions in dehydrated samples of DNA, while in the cell it may be produced in hybrid pairings of DNA and RNA strands, as well as in enzyme-DNA complexes.<sup id="cite_ref-28" class="reference"><a href="#cite_note-28">[28]</a></sup><sup id="cite_ref-29" class="reference"><a href="#cite_note-29">[29]</a></sup> Segments of DNA where the bases have been chemically-modified by <a href="/index.php/Methylation" title="Methylation">methylation</a> may undergo a larger change in conformation and adopt the <a href="/index.php/Z-DNA" title="Z-DNA">Z form</a>. Here, the strands turn about the helical axis in a left-handed spiral, the opposite of the more common B form.<sup id="cite_ref-30" class="reference"><a href="#cite_note-30">[30]</a></sup> These unusual structures can be recognized by specific Z-DNA binding proteins and may be involved in the regulation of transcription.<sup id="cite_ref-31" class="reference"><a href="#cite_note-31">[31]</a></sup>
</p>
<figure class="mw-halign-left" typeof="mw:File/Thumb"><a href="/index.php/File:Parallel_telomere_quadruple.png" class="mw-file-description"><img src="/images/9/9e/Parallel_telomere_quadruple.png" decoding="async" width="300" height="274"></a><figcaption>Structure of a DNA quadruplex formed by <a href="/index.php/Telomere" title="Telomere">telomere</a> repeats. The conformation of the DNA backbone diverges significantly from the typical helical structure<sup id="cite_ref-32" class="reference"><a href="#cite_note-32">[32]</a></sup></figcaption></figure>
<h3><span class="mw-headline" id="Quadruplex_structures">Quadruplex structures</span></h3>
<div role="note" class="hatnote navigation-not-searchable">Further information: [[:<a href="/index.php/G-quadruplex" title="G-quadruplex">G-quadruplex</a>]]</div>
<p>At the ends of the linear chromosomes are specialized regions of DNA called <a href="/index.php/Telomere" title="Telomere">telomeres</a>. The main function of these regions is to allow the cell to replicate chromosome ends using the enzyme <a href="/index.php/Telomerase" title="Telomerase">telomerase</a>, as the enzymes that normally replicate DNA cannot copy the extreme 3′ ends of chromosomes.<sup id="cite_ref-Greider_33-0" class="reference"><a href="#cite_note-Greider-33">[33]</a></sup> These specialized chromosome caps also help protect the DNA ends, and stop the <a href="/index.php/DNA_repair" title="DNA repair">DNA repair</a> systems in the cell from treating them as damage to be corrected.<sup id="cite_ref-Nugent_34-0" class="reference"><a href="#cite_note-Nugent-34">[34]</a></sup> In human cells, telomeres are usually lengths of single-stranded DNA containing several thousand repeats of a simple TTAGGG sequence.<sup id="cite_ref-35" class="reference"><a href="#cite_note-35">[35]</a></sup>
</p><p>These guanine-rich sequences may stabilize chromosome ends by forming structures of stacked sets of four-base units, rather than the usual base pairs found in other DNA molecules. Here, four guanine bases form a flat plate and these flat four-base units then stack on top of each other, to form a stable <i><a href="/index.php/G-quadruplex" title="G-quadruplex">G-quadruplex</a></i> structure.<sup id="cite_ref-Burge_36-0" class="reference"><a href="#cite_note-Burge-36">[36]</a></sup> These structures are stabilized by hydrogen bonding between the edges of the bases and <a href="/index.php/Chelation" title="Chelation">chelation</a> of a metal ion in the centre of each four-base unit.<sup id="cite_ref-37" class="reference"><a href="#cite_note-37">[37]</a></sup> Other structures can also be formed, with the central set of four bases coming from either a single strand folded around the bases, or several different parallel strands, each contributing one base to the central structure.
</p><p>In addition to these stacked structures, telomeres also form large loop structures called telomere loops, or T-loops. Here, the single-stranded DNA curls around in a long circle stabilized by telomere-binding proteins.<sup id="cite_ref-38" class="reference"><a href="#cite_note-38">[38]</a></sup> At the very end of the T-loop, the single-stranded telomere DNA is held onto a region of double-stranded DNA by the telomere strand disrupting the double-helical DNA and base pairing to one of the two strands. This <a href="/index.php/Triple-stranded_DNA" title="Triple-stranded DNA">triple-stranded</a> structure is called a displacement loop or <a href="/index.php/D-loop" title="D-loop">D-loop</a>.<sup id="cite_ref-Burge_36-1" class="reference"><a href="#cite_note-Burge-36">[36]</a></sup>
</p>
<h2><span class="mw-headline" id="Chemical_modifications">Chemical modifications</span></h2>
<div class="thumb tright" style="background-color: #f9f9f9; border: 1px solid #CCCCCC; margin:0.5em;">
<table border="0" width="300" cellpadding="2" cellspacing="0" style="font-size: 85%; border: 1px solid #CCCCCC; margin: 0.3em;">
<tbody><tr>
<td><span typeof="mw:File"><a href="/index.php/File:Cytosine_chemical_structure.png" class="mw-file-description"><img src="/images/1/10/Cytosine_chemical_structure.png" decoding="async" width="75" height="100"></a></span>
</td>
<td><span typeof="mw:File"><a href="/index.php/File:5-methylcytosine.png" class="mw-file-description"><img src="/images/9/9b/5-methylcytosine.png" decoding="async" width="95" height="98"></a></span>
</td>
<td><span typeof="mw:File"><a href="/index.php/File:Thymine_chemical_structure.png" class="mw-file-description"><img src="/images/e/ef/Thymine_chemical_structure.png" decoding="async" width="97" height="104"></a></span>
</td></tr>
<tr>
<td align="center"><a href="/index.php/Cytosine" title="Cytosine">cytosine</a>
</td>
<td align="center"><a href="/index.php/5-Methylcytosine" title="5-Methylcytosine">5-methylcytosine</a>
</td>
<td align="center"><a href="/index.php/Thymine" title="Thymine">thymine</a>
</td></tr></tbody></table>
<div style="border: none; width:300px;font-size: 85%;"><div class="thumbcaption">Structure of cytosine with and without the 5-methyl group. After deamination the 5-methylcytosine has the same structure as thymine</div></div></div>
<h3><span class="mw-headline" id="Base_modifications">Base modifications</span></h3>
<div role="note" class="hatnote navigation-not-searchable">Further information: [[:<a href="/index.php/DNA_methylation" title="DNA methylation">DNA methylation</a>]]</div>
<p>The expression of genes is influenced by the <a href="/index.php/Chromatin" title="Chromatin">chromatin</a> structure of a chromosome and regions of that have low or no gene expression usually contain high levels of <a href="/index.php/Methylation" title="Methylation">methylation</a> of <a href="/index.php/Cytosine" title="Cytosine">cytosine</a> bases. For example, cytosine methylation, producing <a href="/index.php/5-Methylcytosine" title="5-Methylcytosine">5-methylcytosine</a>, is important for <a href="/index.php/X-inactivation" title="X-inactivation">X-chromosome inactivation</a>.<sup id="cite_ref-39" class="reference"><a href="#cite_note-39">[39]</a></sup> The average level of methylation varies between organisms, with <i><a href="/index.php/Caenorhabditis_elegans" title="Caenorhabditis elegans">Caenorhabditis elegans</a></i> lacking cytosine methylation, while <a href="/index.php/Vertebrate" title="Vertebrate">vertebrates</a> show higher levels, with up to 1% of their DNA containing 5-methylcytosine.<sup id="cite_ref-40" class="reference"><a href="#cite_note-40">[40]</a></sup> Despite the biological role of 5-methylcytosine it can <a href="/index.php/Deamination" title="Deamination">deaminate</a> to leave a thymine base, methylated cytosines are therefore particularly prone to <a href="/index.php/Mutation" title="Mutation">mutations</a>.<sup id="cite_ref-41" class="reference"><a href="#cite_note-41">[41]</a></sup> Other base modifications include adenine methylation in bacteria and the <a href="/index.php/Glycosylation" title="Glycosylation">glycosylation</a> of uracil to produce the "J-base" in <a href="/index.php/Kinetoplastid" title="Kinetoplastid">kinetoplastids</a>.<sup id="cite_ref-42" class="reference"><a href="#cite_note-42">[42]</a></sup><sup id="cite_ref-43" class="reference"><a href="#cite_note-43">[43]</a></sup>
</p>
<h3><span class="mw-headline" id="DNA_damage">DNA damage</span></h3>
<div role="note" class="hatnote navigation-not-searchable">Further information: [[:<a href="/index.php/Mutation" title="Mutation">Mutation</a>]]</div>
<figure class="mw-halign-right" typeof="mw:File/Thumb"><a href="/index.php/File:Benzopyrene_DNA_adduct_1JDG.png" class="mw-file-description"><img src="/images/d/d8/Benzopyrene_DNA_adduct_1JDG.png" decoding="async" width="250" height="346"></a><figcaption><a href="/index.php/Benzopyrene" title="Benzopyrene">Benzopyrene</a>, the major mutagen in <a href="/index.php/Tobacco_smoking" title="Tobacco smoking">tobacco smoke</a>, in an adduct to DNA<sup id="cite_ref-44" class="reference"><a href="#cite_note-44">[44]</a></sup></figcaption></figure>
<p>DNA can be damaged by many different sorts of <a href="/index.php/Mutagen" title="Mutagen">mutagens</a>, which are agents that change the DNA sequence. These agents include <a href="/index.php/Oxidizing_agent" title="Oxidizing agent">oxidizing agents</a>, <a href="/index.php/Alkylation" title="Alkylation">alkylating agents</a> and also high-energy <a href="/index.php/Electromagnetic_radiation" title="Electromagnetic radiation">electromagnetic radiation</a> such as <a href="/index.php/Ultraviolet" title="Ultraviolet">ultraviolet</a> light and <a href="/index.php/X-ray" class="mw-redirect" title="X-ray">X-rays</a>. The type of DNA damage produced depends on the type of mutagen. For example, UV light mostly damages DNA by producing <a href="/index.php/Thymine_dimer" title="Thymine dimer">thymine dimers</a>, which are cross-links between adjacent pyrimidine bases in a DNA strand.<sup id="cite_ref-45" class="reference"><a href="#cite_note-45">[45]</a></sup> On the other hand, oxidants such as <a href="/index.php/Radical_(chemistry)" title="Radical (chemistry)">free radicals</a> or <a href="/index.php/Hydrogen_peroxide" title="Hydrogen peroxide">hydrogen peroxide</a> produce multiple forms of damage, including base modifications, particularly of guanosine, as well as double-strand breaks.<sup id="cite_ref-46" class="reference"><a href="#cite_note-46">[46]</a></sup> It has been estimated that in each human cell, about 500 bases suffer oxidative damage per day.<sup id="cite_ref-47" class="reference"><a href="#cite_note-47">[47]</a></sup><sup id="cite_ref-48" class="reference"><a href="#cite_note-48">[48]</a></sup> Of these oxidative lesions, the most dangerous are double-strand breaks, as these are difficult to repair and can produce <a href="/index.php/Point_mutation" title="Point mutation">point mutations</a>, <a href="/index.php/Genetic_insertion" title="Genetic insertion">insertions</a> and <a href="/index.php/Deletion_(genetics)" title="Deletion (genetics)">deletions</a> from the DNA sequence, as well as <a href="/index.php/Chromosomal_translocation" title="Chromosomal translocation">chromosomal translocations</a>.<sup id="cite_ref-49" class="reference"><a href="#cite_note-49">[49]</a></sup>
</p><p>Many mutagens <a href="/index.php/Intercalation_(chemistry)" title="Intercalation (chemistry)">intercalate</a> into the space between two adjacent base pairs. Intercalators are mostly <a href="/index.php/Aromaticity" title="Aromaticity">aromatic</a> and planar molecules, and include <a href="/index.php/Ethidium_bromide" title="Ethidium bromide">ethidium</a>, <a href="/index.php/Daunorubicin" title="Daunorubicin">daunomycin</a>, <a href="/index.php/Doxorubicin" class="mw-redirect" title="Doxorubicin">doxorubicin</a> and <a href="/index.php/Thalidomide" title="Thalidomide">thalidomide</a>. In order for an intercalator to fit between base pairs, the bases must separate, distorting the DNA strands by unwinding of the double helix. These structural changes inhibit both transcription and DNA replication, causing toxicity and mutations. As a result, DNA intercalators are often <a href="/index.php/Carcinogen" title="Carcinogen">carcinogens</a>, with <a href="/index.php/Benzopyrene" title="Benzopyrene">benzopyrene diol epoxide</a>, <a href="/index.php/Acridine" title="Acridine">acridines</a>, <a href="/index.php/Aflatoxin" title="Aflatoxin">aflatoxin</a> and <a href="/index.php/Ethidium_bromide" title="Ethidium bromide">ethidium bromide</a> being well-known examples.<sup id="cite_ref-50" class="reference"><a href="#cite_note-50">[50]</a></sup><sup id="cite_ref-51" class="reference"><a href="#cite_note-51">[51]</a></sup><sup id="cite_ref-52" class="reference"><a href="#cite_note-52">[52]</a></sup> Nevertheless, due to their properties of inhibiting DNA transcription and replication, they are also used in <a href="/index.php/Chemotherapy" title="Chemotherapy">chemotherapy</a> to inhibit rapidly-growing <a href="/index.php/Cancer" title="Cancer">cancer</a> cells.<sup id="cite_ref-53" class="reference"><a href="#cite_note-53">[53]</a></sup>
</p>
<h2><span class="mw-headline" id="Overview_of_biological_functions">Overview of biological functions</span></h2>
<p>DNA usually occurs as linear <a href="/index.php/Chromosome" title="Chromosome">chromosomes</a> in eukaryotes, and circular chromosomes in prokaryotes. The set of chromosomes in a cell makes up its <a href="/index.php/Genome" title="Genome">genome</a>; the <a href="/index.php/Human_genome" title="Human genome">human genome</a> has approximately 3 billion base pairs of DNA arranged into 46 chromosomes.<sup id="cite_ref-54" class="reference"><a href="#cite_note-54">[54]</a></sup> The information carried by DNA is held in the <a href="/index.php/DNA_sequence" title="DNA sequence">sequence</a> of pieces of DNA called <a href="/index.php/Gene" title="Gene">genes</a>. <a href="/index.php/Transmission_(genetics)" title="Transmission (genetics)">Transmission</a> of genetic information in genes is achieved via complementary base pairing. For example, in transcription, when a cell uses the information in a gene, the DNA sequence is copied into a complementary RNA sequence through the attraction between the DNA and the correct RNA nucleotides. Usually, this RNA copy is then used to make a matching protein sequence in a process called <a href="/index.php/Translation_(biology)" title="Translation (biology)">translation</a> which depends on the same interaction between RNA nucleotides. Alternatively, a cell may simply copy its genetic information in a process called DNA replication. The details of these functions are covered in other articles; here we focus on the interactions between DNA and other molecules that mediate the function of the genome.
</p>
<h3><span class="mw-headline" id="Genes_and_genomes">Genes and genomes</span></h3>
<div role="note" class="hatnote navigation-not-searchable">Further information: [[:<a href="/index.php/Cell_nucleus" title="Cell nucleus">Cell nucleus</a>, <a href="/index.php/Chromatin" title="Chromatin">Chromatin</a>, <a href="/index.php/Chromosome" title="Chromosome">Chromosome</a>, <a href="/index.php/Gene" title="Gene">Gene</a>, <a href="/index.php/Noncoding_DNA" title="Noncoding DNA">Noncoding DNA</a>]]</div>
<p>Genomic DNA is located in the <a href="/index.php/Cell_nucleus" title="Cell nucleus">cell nucleus</a> of eukaryotes, as well as small amounts in <a href="/index.php/Mitochondrion" title="Mitochondrion">mitochondria</a> and <a href="/index.php/Chloroplast" title="Chloroplast">chloroplasts</a>. In prokaryotes, the DNA is held within an irregularly shaped body in the cytoplasm called the <a href="/index.php/Nucleoid" title="Nucleoid">nucleoid</a>.<sup id="cite_ref-55" class="reference"><a href="#cite_note-55">[55]</a></sup> The genetic information in a genome is held within genes, and the complete set of this information in an organism is called its <a href="/index.php/Genotype" title="Genotype">genotype</a>. A gene is a unit of <a href="/index.php/Heredity" title="Heredity">heredity</a> and is a region of DNA that influences a particular characteristic in an organism. Genes contain an <a href="/index.php/Open_reading_frame" title="Open reading frame">open reading frame</a> that can be transcribed, as well as <a href="/index.php/Regulatory_sequence" title="Regulatory sequence">regulatory sequences</a> such as <a href="/index.php/Promoter" title="Promoter">promoters</a> and <a href="/index.php/Enhancer_(genetics)" title="Enhancer (genetics)">enhancers</a>, which control the transcription of the open reading frame. 
</p><p>In many <a href="/index.php/Species" title="Species">species</a>, only a small fraction of the total sequence of the <a href="/index.php/Genome" title="Genome">genome</a> encodes protein. For example, only about 1.5% of the human genome consists of protein-coding <a href="/index.php/Exon" title="Exon">exons</a>, with over 50% of human DNA consisting of non-coding <a href="/index.php/Repeated_sequence_(DNA)" title="Repeated sequence (DNA)">repetitive sequences</a>.<sup id="cite_ref-56" class="reference"><a href="#cite_note-56">[56]</a></sup> The reasons for the presence of so much <a href="/index.php/Noncoding_DNA" title="Noncoding DNA">non-coding DNA</a> in eukaryotic genomes and the extraordinary differences in <a href="/index.php/Genome_size" title="Genome size">genome size</a>, or <i><a href="/index.php/C-value" title="C-value">C-value</a></i>, among species represent a long-standing puzzle known as the "<a href="/index.php/C-value_enigma" title="C-value enigma">C-value enigma</a>."<sup id="cite_ref-57" class="reference"><a href="#cite_note-57">[57]</a></sup> However, DNA sequences that do not code protein may still encode functional <a href="/index.php/Non-coding_RNA" title="Non-coding RNA">non-coding RNA</a> molecules, which are involved in the regulation of gene expression.<sup id="cite_ref-58" class="reference"><a href="#cite_note-58">[58]</a></sup>
</p>
<figure class="mw-halign-left" typeof="mw:File/Thumb"><a href="/index.php/File:T7_RNA_polymerase_at_work.png" class="mw-file-description"><img src="/images/f/f3/T7_RNA_polymerase_at_work.png" decoding="async" width="300" height="208"></a><figcaption><a href="/index.php/T7_RNA_polymerase" title="T7 RNA polymerase">T7 RNA polymerase</a> (blue) producing a mRNA (green) from a DNA template (orange).<sup id="cite_ref-59" class="reference"><a href="#cite_note-59">[59]</a></sup></figcaption></figure> 
<p>Some non-coding DNA sequences play structural roles in chromosomes. <a href="/index.php/Telomere" title="Telomere">Telomeres</a> and <a href="/index.php/Centromere" title="Centromere">centromeres</a> typically contain few genes, but are important for the function and stability of chromosomes.<sup id="cite_ref-Nugent_34-1" class="reference"><a href="#cite_note-Nugent-34">[34]</a></sup><sup id="cite_ref-60" class="reference"><a href="#cite_note-60">[60]</a></sup> An abundant form of non-coding DNA in humans are <a href="/index.php/Pseudogene" title="Pseudogene">pseudogenes</a>, which are copies of genes that have been disabled by mutation.<sup id="cite_ref-61" class="reference"><a href="#cite_note-61">[61]</a></sup> These sequences are usually just molecular <a href="/index.php/Fossil" title="Fossil">fossils</a>, although they can occasionally serve as raw genetic material for the creation of new genes through the process of <a href="/index.php/Gene_duplication" title="Gene duplication">gene duplication</a> and <a href="/index.php/Divergent_evolution" title="Divergent evolution">divergence</a>.<sup id="cite_ref-62" class="reference"><a href="#cite_note-62">[62]</a></sup>
</p>
<h3><span class="mw-headline" id="Transcription_and_translation">Transcription and translation</span></h3>
<div role="note" class="hatnote navigation-not-searchable">Further information: [[:<a href="/index.php/Genetic_code" title="Genetic code">Genetic code</a>, <a href="/index.php/Transcription_(genetics)" title="Transcription (genetics)">Transcription (genetics)</a>, <a href="/index.php/Protein_biosynthesis" title="Protein biosynthesis">Protein biosynthesis</a>]]</div>
<p>A gene is a sequence of DNA that contains genetic information and can influence the <a href="/index.php/Phenotype" title="Phenotype">phenotype</a> of an organism. Within a gene, the sequence of bases along a DNA strand defines a <a href="/index.php/Messenger_RNA" title="Messenger RNA">messenger RNA</a> sequence, which then defines one or more protein sequences. The relationship between the nucleotide sequences of genes and the <a href="/index.php/Amino_acid" title="Amino acid">amino-acid</a> sequences of proteins is determined by the rules of <a href="/index.php/Translation_(biology)" title="Translation (biology)">translation</a>, known collectively as the <a href="/index.php/Genetic_code" title="Genetic code">genetic code</a>. The genetic code consists of three-letter 'words' called <i>codons</i> formed from a sequence of three nucleotides (e.g. ACT, CAG, TTT).
</p><p>In transcription, the codons of a gene are copied into messenger RNA by <a href="/index.php/RNA_polymerase" title="RNA polymerase">RNA polymerase</a>. This RNA copy is then decoded by a <a href="/index.php/Ribosome" title="Ribosome">ribosome</a> that reads the RNA sequence by base-pairing the messenger RNA to <a href="/index.php/Transfer_RNA" title="Transfer RNA">transfer RNA</a>, which carries amino acids. Since there are 4 bases in 3-letter combinations, there are 64 possible codons (&lt;math&gt;4^3&lt;/math&gt; combinations). These encode the twenty <a href="/index.php/List_of_standard_amino_acids" title="List of standard amino acids">standard amino acids</a>, giving most amino acids more than one possible codon. There are also three 'stop' or 'nonsense' codons signifying the end of the coding region; these are the TAA, TGA and TAG codons.
</p>
<figure class="mw-halign-right" typeof="mw:File/Thumb"><a href="/index.php/File:DNA_replication.svg" class="mw-file-description"><img src="/images/9/9f/DNA_replication.svg" decoding="async" width="450" height="219"></a><figcaption>DNA replication. The double helix is unwound by a <a href="/index.php/Helicase" title="Helicase">helicase</a> and <a href="/index.php/Topoisomerase" title="Topoisomerase">topoisomerase</a>. Next, one <a href="/index.php/DNA_polymerase" title="DNA polymerase">DNA polymerase</a> produces the <a href="/index.php/Replication_fork" title="Replication fork">leading strand</a> copy. Another DNA polymerase binds to the <a href="/index.php/Replication_fork" title="Replication fork">lagging strand</a>. This enzyme makes discontinuous segments (called <a href="/index.php/Okazaki_fragment" title="Okazaki fragment">Okazaki fragments</a>) before <a href="/index.php/DNA_ligase" title="DNA ligase">DNA ligase</a> joins them together.</figcaption></figure>
<h3><span class="mw-headline" id="Replication">Replication</span></h3>
<div role="note" class="hatnote navigation-not-searchable">Further information: [[:<a href="/index.php/DNA_replication" title="DNA replication">DNA replication</a>]]</div>
<p><a href="/index.php/Cell_division" title="Cell division">Cell division</a> is essential for an organism to grow, but when a cell divides it must replicate the DNA in its genome so that the two daughter cells have the same genetic information as their parent. The double-stranded structure of DNA provides a simple mechanism for <a href="/index.php/DNA_replication" title="DNA replication">DNA replication</a>. Here, the two strands are separated and then each strand's complementary DNA sequence is recreated by an <a href="/index.php/Enzyme" title="Enzyme">enzyme</a> called <a href="/index.php/DNA_polymerase" title="DNA polymerase">DNA polymerase</a>. This enzyme makes the complementary strand by finding the correct base through complementary base pairing, and bonding it onto the original strand. As DNA polymerases can only extend a DNA strand in a 5′ to 3′ direction, different mechanisms are used to copy the antiparallel strands of the double helix.<sup id="cite_ref-63" class="reference"><a href="#cite_note-63">[63]</a></sup> In this way, the base on the old strand dictates which base appears on the new strand, and the cell ends up with a perfect copy of its DNA.
</p>
<h2><span class="mw-headline" id="Interactions_with_proteins">Interactions with proteins</span></h2>
<p>All the functions of DNA depend on interactions with proteins. These protein interactions can be non-specific, or the protein can bind specifically to a single DNA sequence. Enzymes can also bind to DNA and of these, the polymerases that copy the DNA base sequence in transcription and DNA replication are particularly important.
</p>
<h3><span class="mw-headline" id="DNA-binding_proteins">DNA-binding proteins</span></h3>
<div class="thumb tleft" style="background-color: #f9f9f9; border: 1px solid #CCCCCC; margin:0.5em;">
<table border="0" width="260" cellpadding="0" cellspacing="0" style="font-size: 85%; border: 1px solid #CCCCCC; margin: 0.3em;">
<tbody><tr>
<td><span typeof="mw:File"><a href="/index.php/File:Nucleosome_2.jpg" class="mw-file-description"><img src="/images/6/60/Nucleosome_2.jpg" decoding="async" width="260" height="185"></a></span>
</td></tr>
<tr>
<td><span typeof="mw:File"><a href="/index.php/File:Nucleosome_(opposites_attracts).JPG" class="mw-file-description"><img src="/images/3/3e/Nucleosome_%28opposites_attracts%29.JPG" decoding="async" width="260" height="143"></a></span>
</td></tr></tbody></table>
<div style="border: none; width:260px;"><div class="thumbcaption">Interaction of DNA with <a href="/index.php/Histone" title="Histone">histones</a> (shown in white, top). These proteins' basic amino acids (below left, blue) bind to the acidic phosphate groups on DNA (below right, red).</div></div></div>
<p>Structural proteins that bind DNA are well-understood examples of non-specific DNA-protein interactions. Within chromosomes, DNA is held in complexes with structural proteins. These proteins organize the DNA into a compact structure called <a href="/index.php/Chromatin" title="Chromatin">chromatin</a>. In eukaryotes this structure involves DNA binding to a complex of small basic proteins called <a href="/index.php/Histone" title="Histone">histones</a>, while in prokaryotes multiple types of proteins are involved.<sup id="cite_ref-64" class="reference"><a href="#cite_note-64">[64]</a></sup><sup id="cite_ref-65" class="reference"><a href="#cite_note-65">[65]</a></sup> The histones form a disk-shaped complex called a <a href="/index.php/Nucleosome" title="Nucleosome">nucleosome</a>, which contains two complete turns of double-stranded DNA wrapped around its surface. These non-specific interactions are formed through basic residues in the histones making <a href="/index.php/Ionic_bond" title="Ionic bond">ionic bonds</a> to the acidic sugar-phosphate backbone of the DNA, and are therefore largely independent of the base sequence.<sup id="cite_ref-66" class="reference"><a href="#cite_note-66">[66]</a></sup> Chemical modifications of these basic amino acid residues include <a href="/index.php/Methylation" title="Methylation">methylation</a>, <a href="/index.php/Phosphorylation" title="Phosphorylation">phosphorylation</a> and <a href="/index.php/Acetylation" title="Acetylation">acetylation</a>.<sup id="cite_ref-67" class="reference"><a href="#cite_note-67">[67]</a></sup> These chemical changes alter the strength of the interaction between the DNA and the histones, making the DNA more or less accessible to <a href="/index.php/Transcription_factor" title="Transcription factor">transcription factors</a> and changing the rate of transcription.<sup id="cite_ref-68" class="reference"><a href="#cite_note-68">[68]</a></sup> Other non-specific DNA-binding proteins found in chromatin include the high-mobility group proteins, which bind preferentially to bent or distorted DNA.<sup id="cite_ref-69" class="reference"><a href="#cite_note-69">[69]</a></sup> These proteins are important in bending arrays of nucleosomes and arranging them into more complex chromatin structures.<sup id="cite_ref-70" class="reference"><a href="#cite_note-70">[70]</a></sup>
</p><p>A distinct group of DNA-binding proteins are the single-stranded-DNA-binding proteins that specifically bind single-stranded DNA. In humans, replication protein A is the best-characterised member of this family and is essential for most processes where the double helix is separated, including DNA replication, recombination and DNA repair.<sup id="cite_ref-71" class="reference"><a href="#cite_note-71">[71]</a></sup> These binding proteins seem to stabilize single-stranded DNA and protect it from forming <a href="/index.php/Stem-loop" title="Stem-loop">stem-loops</a> or being degraded by <a href="/index.php/Nuclease" title="Nuclease">nucleases</a>.
</p>
<figure class="mw-halign-right" typeof="mw:File/Thumb"><a href="/index.php/File:Lambda_repressor_1LMB.png" class="mw-file-description"><img src="/images/8/8f/Lambda_repressor_1LMB.png" decoding="async" width="185" height="272"></a><figcaption>The lambda repressor <a href="/index.php/Helix-turn-helix" title="Helix-turn-helix">helix-turn-helix</a> transcription factor bound to its DNA target<sup id="cite_ref-72" class="reference"><a href="#cite_note-72">[72]</a></sup></figcaption></figure>
<p>In contrast, other proteins have evolved to specifically bind particular DNA sequences. The most intensively studied of these are the various classes of <a href="/index.php/Transcription_factor" title="Transcription factor">transcription factors</a>, which are proteins that regulate transcription. Each one of these proteins bind to one particular set of DNA sequences and thereby activates or inhibits the transcription of genes with these sequences close to their promoters. The transcription factors do this in two ways. Firstly, they can bind the RNA polymerase responsible for transcription, either directly or through other mediator proteins; this locates the polymerase at the promoter and allows it to begin transcription.<sup id="cite_ref-73" class="reference"><a href="#cite_note-73">[73]</a></sup> Alternatively, transcription factors can bind <a href="/index.php/Enzyme" title="Enzyme">enzymes</a> that modify the histones at the promoter; this will change the accessibility of the DNA template to the polymerase.<sup id="cite_ref-74" class="reference"><a href="#cite_note-74">[74]</a></sup>
</p><p>As these DNA targets can occur throughout an organism's genome, changes in the activity of one type of transcription factor can affect thousands of genes.<sup id="cite_ref-75" class="reference"><a href="#cite_note-75">[75]</a></sup> Consequently, these proteins are often the targets of the <a href="/index.php/Signal_transduction" title="Signal transduction">signal transduction</a> processes that mediate responses to environmental changes or cellular differentiation and development. The specificity of these transcription factors' interactions with DNA come from the proteins making multiple contacts to the edges of the DNA bases, allowing them to "read" the DNA sequence. Most of these base-interactions are made in the major groove, where the bases are most accessible.<sup id="cite_ref-76" class="reference"><a href="#cite_note-76">[76]</a></sup>
</p>
<figure class="mw-halign-left" typeof="mw:File/Thumb"><a href="/index.php/File:EcoRV_1RVA.png" class="mw-file-description"><img src="/images/d/dc/EcoRV_1RVA.png" decoding="async" width="255" height="179"></a><figcaption>The <a href="/index.php/Restriction_enzyme" title="Restriction enzyme">restriction enzyme</a> <a href="/index.php/EcoRV" title="EcoRV">EcoRV</a> (green) in a complex with its substrate DNA<sup id="cite_ref-77" class="reference"><a href="#cite_note-77">[77]</a></sup></figcaption></figure>
<h3><span class="mw-headline" id="DNA-modifying_enzymes">DNA-modifying enzymes</span></h3>
<h4><span class="mw-headline" id="Nucleases_and_ligases">Nucleases and ligases</span></h4>
<p><a href="/index.php/Nuclease" title="Nuclease">Nucleases</a> are <a href="/index.php/Enzyme" title="Enzyme">enzymes</a> that cut DNA strands by catalyzing the <a href="/index.php/Hydrolysis" title="Hydrolysis">hydrolysis</a> of the <a href="/index.php/Phosphodiester_bond" title="Phosphodiester bond">phosphodiester bonds</a>. Nucleases that hydrolyse nucleotides from the ends of DNA strands are called <a href="/index.php/Exonuclease" title="Exonuclease">exonucleases</a>, while <a href="/index.php/Endonuclease" title="Endonuclease">endonucleases</a> cut within strands. The most frequently-used nucleases in <a href="/index.php/Molecular_biology" title="Molecular biology">molecular biology</a> are the <a href="/index.php/Restriction_enzyme" title="Restriction enzyme">restriction endonucleases</a>, which cut DNA at specific sequences. For instance, the EcoRV enzyme shown to the left recognizes the 6-base sequence 5′-GAT|ATC-3′ and makes a cut at the vertical line. In nature, these enzymes protect <a href="/index.php/Bacteria" title="Bacteria">bacteria</a> against <a href="/index.php/Bacteriophage" title="Bacteriophage">phage</a> infection by digesting the phage DNA when it enters the bacterial cell, acting as part of the <a href="/index.php/Restriction_modification_system" title="Restriction modification system">restriction modification system</a>.<sup id="cite_ref-78" class="reference"><a href="#cite_note-78">[78]</a></sup> In technology, these sequence-specific nucleases are used in <a href="/index.php/Molecular_cloning" title="Molecular cloning">molecular cloning</a> and <a href="/index.php/Genetic_fingerprinting" title="Genetic fingerprinting">DNA fingerprinting</a>.
</p><p>Enzymes called <a href="/index.php/DNA_ligase" title="DNA ligase">DNA ligases</a> can rejoin cut or broken DNA strands.<sup id="cite_ref-Doherty_79-0" class="reference"><a href="#cite_note-Doherty-79">[79]</a></sup> Ligases are particularly important in <a href="/index.php/Replication_fork" title="Replication fork">lagging strand</a> DNA replication, as they join together the short segments of DNA produced at the <a href="/index.php/Replication_fork" title="Replication fork">replication fork</a> into a complete copy of the DNA template. They are also used in <a href="/index.php/DNA_repair" title="DNA repair">DNA repair</a> and <a href="/index.php/Genetic_recombination" title="Genetic recombination">genetic recombination</a>.<sup id="cite_ref-Doherty_79-1" class="reference"><a href="#cite_note-Doherty-79">[79]</a></sup>
</p>
<h4><span class="mw-headline" id="Topoisomerases_and_helicases">Topoisomerases and helicases</span></h4>
<p><a href="/index.php/Topoisomerase" title="Topoisomerase">Topoisomerases</a> are enzymes with both nuclease and ligase activity. These proteins change the amount of <a href="/index.php/DNA_supercoil" title="DNA supercoil">supercoiling</a> in DNA. Some of these enzyme work by cutting the DNA helix and allowing one section to rotate, thereby reducing its level of supercoiling; the enzyme then seals the DNA break.<sup id="cite_ref-Champoux_24-1" class="reference"><a href="#cite_note-Champoux-24">[24]</a></sup> Other types of these enzymes are capable of cutting one DNA helix and then passing a second strand of DNA through this break, before rejoining the helix.<sup id="cite_ref-80" class="reference"><a href="#cite_note-80">[80]</a></sup> Topoisomerases are required for many processes involving DNA, such as DNA replication and transcription.<sup id="cite_ref-Wang_25-1" class="reference"><a href="#cite_note-Wang-25">[25]</a></sup>
</p><p><a href="/index.php/Helicase" title="Helicase">Helicases</a> are proteins that are a type of <a href="/index.php/Molecular_motor" title="Molecular motor">molecular motor</a>. They use the chemical energy in <a href="/index.php/Nucleoside_triphosphate" title="Nucleoside triphosphate">nucleoside triphosphates</a>, predominantly <a href="/index.php/Adenosine_triphosphate" title="Adenosine triphosphate">ATP</a>, to break hydrogen bonds between bases and unwind the DNA double helix into single strands.<sup id="cite_ref-81" class="reference"><a href="#cite_note-81">[81]</a></sup> These enzymes are essential for most processes where enzymes need to access the DNA bases.
</p>
<h4><span class="mw-headline" id="Polymerases">Polymerases</span></h4>
<p><a href="/index.php/Polymerase" title="Polymerase">Polymerases</a> are <a href="/index.php/Enzyme" title="Enzyme">enzymes</a> that synthesize polynucleotide chains from <a href="/index.php/Nucleoside_triphosphate" title="Nucleoside triphosphate">nucleoside triphosphates</a>. The sequence of their products are copies of existing polynucleotide chains - which are called <i>templates</i>. These enzymes function by adding nucleotides onto the 3′ <a href="/index.php/Hydroxyl" title="Hydroxyl">hydroxyl group</a> of the previous nucleotide in a DNA strand. Consequently, all polymerases work in a 5′ to 3′ direction.<sup id="cite_ref-Joyce_82-0" class="reference"><a href="#cite_note-Joyce-82">[82]</a></sup> In the <a href="/index.php/Active_site" title="Active site">active site</a> of these enzymes, the incoming nucleoside triphosphate base-pairs to the template: this allows polymerases to accurately synthesize the complementary strand of their template. Polymerases are classified according to the type of template that they use.
</p><p>In DNA replication, a DNA-dependent <a href="/index.php/DNA_polymerase" title="DNA polymerase">DNA polymerase</a> makes a DNA copy of a DNA sequence. Accuracy is vital in this process, so many of these polymerases have a <a href="/index.php?title=Proofreading&amp;action=edit&amp;redlink=1" class="new" title="Proofreading (page does not exist)">proofreading</a> activity. Here, the polymerase recognizes the occasional mistakes in the synthesis reaction by the lack of base pairing between the mismatched nucleotides. If a mismatch is detected, a 3′ to 5′ <a href="/index.php/Exonuclease" title="Exonuclease">exonuclease</a> activity is activated and the incorrect base removed.<sup id="cite_ref-83" class="reference"><a href="#cite_note-83">[83]</a></sup> In most organisms DNA polymerases function in a large complex called the <a href="/index.php/Replisome" title="Replisome">replisome</a> that contains multiple accessory subunits, such as the <a href="/index.php/DNA_clamp" title="DNA clamp">DNA clamp</a> or <a href="/index.php/Helicase" title="Helicase">helicases</a>.<sup id="cite_ref-84" class="reference"><a href="#cite_note-84">[84]</a></sup>
</p><p>RNA-dependent DNA polymerases are a specialized class of polymerases that copy the sequence of an RNA strand into DNA. They include <a href="/index.php/Reverse_transcriptase" title="Reverse transcriptase">reverse transcriptase</a>, which is a <a href="/index.php/Virus" title="Virus">viral</a> enzyme involved in the infection of cells by <a href="/index.php/Retrovirus" title="Retrovirus">retroviruses</a>, and <a href="/index.php/Telomerase" title="Telomerase">telomerase</a>, which is required for the replication of telomeres.<sup id="cite_ref-85" class="reference"><a href="#cite_note-85">[85]</a></sup><sup id="cite_ref-Greider_33-1" class="reference"><a href="#cite_note-Greider-33">[33]</a></sup> Telomerase is an unusual polymerase because it contains its own RNA template as part of its structure.<sup id="cite_ref-Nugent_34-2" class="reference"><a href="#cite_note-Nugent-34">[34]</a></sup>
</p><p>Transcription is carried out by a DNA-dependent <a href="/index.php/RNA_polymerase" title="RNA polymerase">RNA polymerase</a> that copies the sequence of a DNA strand into RNA. To begin transcribing a gene, the RNA polymerase binds to a sequence of DNA called a promoter and separates the DNA strands. It then copies the gene sequence into a <a href="/index.php/Messenger_RNA" title="Messenger RNA">messenger RNA</a> transcript until it reaches a region of DNA called the <a href="/index.php/Terminator_(genetics)" title="Terminator (genetics)">terminator</a>, where it halts and detaches from the DNA. As with human DNA-dependent DNA polymerases, RNA polymerase II, the enzyme that transcribes most of the genes in the human genome, operates as part of a large protein complex with multiple regulatory and accessory subunits.<sup id="cite_ref-86" class="reference"><a href="#cite_note-86">[86]</a></sup>
</p>
<h2><span class="mw-headline" id="Genetic_recombination">Genetic recombination</span></h2>
<div class="thumb tright" style="background-color: #f9f9f9; border: 1px solid #CCCCCC; margin:0.5em;">
<table border="0" width="250" cellpadding="0" cellspacing="0" style="font-size: 85%; border: 1px solid #CCCCCC; margin: 0.3em;">
<tbody><tr>
<td><span typeof="mw:File"><a href="/index.php/File:Holliday_Junction_cropped.png" class="mw-file-description"><img src="/images/9/9b/Holliday_Junction_cropped.png" decoding="async" width="250" height="200"></a></span>
</td></tr>
<tr>
<td><span typeof="mw:File"><a href="/index.php/File:Holliday_junction_coloured.png" class="mw-file-description"><img src="/images/9/92/Holliday_junction_coloured.png" decoding="async" width="250" height="250"></a></span>
</td></tr></tbody></table>
<div style="border: none; width:250px;"><div class="thumbcaption">Structure of the <a href="/index.php/Holliday_junction" title="Holliday junction">Holliday junction</a> intermediate in <a href="/index.php/Genetic_recombination" title="Genetic recombination">genetic recombination</a>. The four separate DNA strands are coloured red, blue, green and yellow.<sup id="cite_ref-87" class="reference"><a href="#cite_note-87">[87]</a></sup></div></div></div>
<div role="note" class="hatnote navigation-not-searchable">Further information: [[:<a href="/index.php/Genetic_recombination" title="Genetic recombination">Genetic recombination</a>]]</div>
<figure class="mw-halign-left" typeof="mw:File/Thumb"><a href="/index.php/File:Chromosomal_Recombination.svg" class="mw-file-description"><img src="/images/b/b2/Chromosomal_Recombination.svg" decoding="async" width="250" height="162"></a><figcaption>Recombination involves the breakage and rejoining of two chromosomes (M and F) to produce two re-arranged chromosomes (C1 and C2).</figcaption></figure>
<p>A DNA helix usually does not  interact with other segments of DNA, and in human cells the different chromosomes even occupy separate areas in the nucleus called "chromosome territories".<sup id="cite_ref-88" class="reference"><a href="#cite_note-88">[88]</a></sup> This physical separation of different chromosomes is important for the ability of DNA to function as a stable repository for information, as one of the few times chromosomes interact is during <a href="/index.php/Chromosomal_crossover" title="Chromosomal crossover">chromosomal crossover</a> when they <a href="/index.php/Genetic_recombination" title="Genetic recombination">recombine</a>. Chromosomal crossover is when two DNA helices break, swap a section and then rejoin.
</p><p>Recombination allows chromosomes to exchange genetic information and produces new combinations of genes, which increases the efficiency of <a href="/index.php?title=Natural_selection&amp;action=edit&amp;redlink=1" class="new" title="Natural selection (page does not exist)">natural selection</a> and can be important in the rapid evolution of new proteins.<sup id="cite_ref-89" class="reference"><a href="#cite_note-89">[89]</a></sup> Genetic recombination can also be involved in DNA repair, particularly in the cell's response to double-strand breaks.<sup id="cite_ref-90" class="reference"><a href="#cite_note-90">[90]</a></sup>
</p><p>The most common form of chromosomal crossover is <a href="/index.php/Homologous_recombination" title="Homologous recombination">homologous recombination</a>, where the two chromosomes involved share very similar sequences. Non-homologous recombination can be damaging to cells, as it can produce <a href="/index.php/Chromosomal_translocation" title="Chromosomal translocation">chromosomal translocations</a> and genetic abnormalities. The recombination reaction is catalyzed by enzymes known as <i>recombinases</i>, such as <a href="/index.php/RAD51" title="RAD51">RAD51</a>.<sup id="cite_ref-91" class="reference"><a href="#cite_note-91">[91]</a></sup>  The first step in recombination is a double-stranded break either caused by an <a href="/index.php/Endonuclease" title="Endonuclease">endonuclease</a> or damage to the DNA.<sup id="cite_ref-92" class="reference"><a href="#cite_note-92">[92]</a></sup>  A series of steps catalyzed in part by the recombinase then leads to joining of the two helices by at least one <a href="/index.php/Holliday_junction" title="Holliday junction">Holliday junction</a>, in which a segment of a single strand in each helix is annealed to the complementary strand in the other helix. The Holliday junction is a tetrahedral junction structure that can be moved along the pair of chromosomes, swapping one strand for another. The recombination reaction is then halted by cleavage of the junction and re-ligation of the released DNA.<sup id="cite_ref-93" class="reference"><a href="#cite_note-93">[93]</a></sup>
</p>
<h2><span class="mw-headline" id="Evolution_of_DNA_metabolism">Evolution of DNA metabolism</span></h2>
<div role="note" class="hatnote navigation-not-searchable">Further information: [[:<a href="/index.php/RNA_world_hypothesis" title="RNA world hypothesis">RNA world hypothesis</a>]]</div>
<p>DNA contains the genetic information that allows all modern living things to function, grow and reproduce. However, it is unclear how long in the 4-billion-year <a href="/index.php/Timeline_of_evolution" title="Timeline of evolution">history of life</a> DNA has performed this function, as it has been proposed that the earliest forms of life may have used RNA as their genetic material.<sup id="cite_ref-Joyce_82-1" class="reference"><a href="#cite_note-Joyce-82">[82]</a></sup><sup id="cite_ref-94" class="reference"><a href="#cite_note-94">[94]</a></sup> RNA may have acted as the central part of early cell metabolism as it can both transmit genetic information and carry out <a href="/index.php/Catalysis" title="Catalysis">catalysis</a> as part of <a href="/index.php/Ribozyme" title="Ribozyme">ribozymes</a>.<sup id="cite_ref-95" class="reference"><a href="#cite_note-95">[95]</a></sup> This ancient <a href="/index.php/RNA_world_hypothesis" title="RNA world hypothesis">RNA world</a> where nucleic acid would have been used for both catalysis and genetics may have influenced the <a href="/index.php?title=Evolution&amp;action=edit&amp;redlink=1" class="new" title="Evolution (page does not exist)">evolution</a> of the current genetic code based on four nucleotide bases. This would occur since the number of unique bases in such an organism is a trade-off between a small number of bases increasing replication accuracy and a large number of bases increasing the catalytic efficiency of ribozymes.<sup id="cite_ref-96" class="reference"><a href="#cite_note-96">[96]</a></sup> 
</p><p>Unfortunately, there is no direct evidence of ancient genetic systems, as recovery of DNA from most fossils is impossible. This is because DNA will survive in the environment for less than one million years and slowly degrades into short fragments in solution.<sup id="cite_ref-97" class="reference"><a href="#cite_note-97">[97]</a></sup> Claims for older DNA have been made, most notably a report of the isolation of a viable bacterium from a salt crystal 250-million years old,<sup id="cite_ref-98" class="reference"><a href="#cite_note-98">[98]</a></sup> but these claims are controversial.<sup id="cite_ref-99" class="reference"><a href="#cite_note-99">[99]</a></sup><sup id="cite_ref-100" class="reference"><a href="#cite_note-100">[100]</a></sup>
</p>
<h2><span class="mw-headline" id="Uses_in_technology">Uses in technology</span></h2>
<h3><span class="mw-headline" id="Genetic_engineering">Genetic engineering</span></h3>
<div role="note" class="hatnote navigation-not-searchable">Further information: [[:<a href="/index.php/Molecular_biology" title="Molecular biology">Molecular biology</a> and <a href="/index.php/Genetic_engineering" title="Genetic engineering">genetic engineering</a>]]</div>
<p>Modern <a href="/index.php/Biology" title="Biology">biology</a> and <a href="/index.php/Biochemistry" title="Biochemistry">biochemistry</a> make intensive use of recombinant DNA technology. <a href="/index.php/Recombinant_DNA" title="Recombinant DNA">Recombinant DNA</a> is a man-made DNA sequence that has been assembled from other DNA sequences. They can be <a href="/index.php/Transformation_(genetics)" title="Transformation (genetics)">transformed</a> into organisms in the form of <a href="/index.php/Plasmid" title="Plasmid">plasmids</a> or in the appropriate format, by using a <a href="/index.php/Viral_vector" title="Viral vector">viral vector</a>.<sup id="cite_ref-101" class="reference"><a href="#cite_note-101">[101]</a></sup> The <a href="/index.php/Genetic_engineering" title="Genetic engineering">genetically modified</a> organisms produced can be used to produce products such as recombinant <a href="/index.php/Protein" title="Protein">proteins</a>, used in medical research,<sup id="cite_ref-102" class="reference"><a href="#cite_note-102">[102]</a></sup> or be grown in agriculture.<sup id="cite_ref-103" class="reference"><a href="#cite_note-103">[103]</a></sup><sup id="cite_ref-104" class="reference"><a href="#cite_note-104">[104]</a></sup>
</p>
<h3><span class="mw-headline" id="Forensics">Forensics</span></h3>
<div role="note" class="hatnote navigation-not-searchable">Further information: [[:<a href="/index.php/Genetic_fingerprinting" title="Genetic fingerprinting">Genetic fingerprinting</a>]]</div>
<p><a href="/index.php/Forensic_science" title="Forensic science">Forensic scientists</a> can use DNA in <a href="/index.php/Blood" title="Blood">blood</a>, <a href="/index.php/Semen" title="Semen">semen</a>, <a href="/index.php/Skin" title="Skin">skin</a>, <a href="/index.php/Saliva" title="Saliva">saliva</a> or <a href="/index.php/Hair" title="Hair">hair</a> at a crime scene to identify a perpetrator. This process is called <a href="/index.php/Genetic_fingerprinting" title="Genetic fingerprinting">genetic fingerprinting</a>, or more accurately, DNA profiling. In DNA profiling, the lengths of variable sections of repetitive DNA, such as <a href="/index.php/Short_tandem_repeat" title="Short tandem repeat">short tandem repeats</a> and <a href="/index.php/Minisatellite" title="Minisatellite">minisatellites</a>, are compared between people. This method is usually an extremely reliable technique for identifying a criminal.<sup id="cite_ref-105" class="reference"><a href="#cite_note-105">[105]</a></sup> However, identification can be complicated if the scene is contaminated with DNA from several people.<sup id="cite_ref-106" class="reference"><a href="#cite_note-106">[106]</a></sup> DNA profiling was developed in 1984 by British geneticist Sir <a href="/index.php?title=Alec_Jeffreys&amp;action=edit&amp;redlink=1" class="new" title="Alec Jeffreys (page does not exist)">Alec Jeffreys</a>,<sup id="cite_ref-107" class="reference"><a href="#cite_note-107">[107]</a></sup> and first used in forensic science to convict Colin Pitchfork in the 1988 Enderby murders case.<sup id="cite_ref-108" class="reference"><a href="#cite_note-108">[108]</a></sup> People convicted of certain types of crimes may be required to provide a sample of DNA for a database. This has helped investigators solve old cases where only a DNA sample was obtained from the scene. DNA profiling can also be used to identify victims of mass casualty incidents.<sup id="cite_ref-109" class="reference"><a href="#cite_note-109">[109]</a></sup>
</p>
<h3><span class="mw-headline" id="Bioinformatics">Bioinformatics</span></h3>
<div role="note" class="hatnote navigation-not-searchable">Further information: [[:<a href="/index.php/Bioinformatics" title="Bioinformatics">Bioinformatics</a>]]</div>
<p><a href="/index.php/Bioinformatics" title="Bioinformatics">Bioinformatics</a> involves the manipulation, searching, and <a href="/index.php/Data_mining" title="Data mining">data mining</a> of DNA sequence data. The development of techniques to store and search DNA sequences have led to widely-applied advances in <a href="/index.php?title=Computer_science&amp;action=edit&amp;redlink=1" class="new" title="Computer science (page does not exist)">computer science</a>, especially <a href="/index.php/String_searching_algorithm" title="String searching algorithm">string searching algorithms</a>, <a href="/index.php/Machine_learning" title="Machine learning">machine learning</a> and <a href="/index.php/Database_theory" title="Database theory">database theory</a>.<sup id="cite_ref-110" class="reference"><a href="#cite_note-110">[110]</a></sup> String searching or matching algorithms, which find an occurrence of a sequence of letters inside a larger sequence of letters, were developed to search for specific sequences of nucleotides.<sup id="cite_ref-111" class="reference"><a href="#cite_note-111">[111]</a></sup> In other applications such as text editors, even simple algorithms for this problem usually suffice, but DNA sequences cause these algorithms to exhibit near-worst-case behaviour due to their small number of distinct characters. The related problem of <a href="/index.php/Sequence_alignment" title="Sequence alignment">sequence alignment</a> aims to identify <a href="/index.php/Homology_(biology)" title="Homology (biology)">homologous</a> sequences and locate the specific <a href="/index.php/Mutation" title="Mutation">mutations</a> that make them distinct. These techniques, especially <a href="/index.php/Multiple_sequence_alignment" title="Multiple sequence alignment">multiple sequence alignment</a>, are used in studying <a href="/index.php/Phylogenetics" title="Phylogenetics">phylogenetic</a> relationships and protein function.<sup id="cite_ref-112" class="reference"><a href="#cite_note-112">[112]</a></sup> Data sets representing entire genomes' worth of DNA sequences, such as those produced by the <a href="/index.php/Human_Genome_Project" class="mw-redirect" title="Human Genome Project">Human Genome Project</a>, are difficult to use without annotations, which label the locations of genes and regulatory elements on each chromosome. Regions of DNA sequence that have the characteristic patterns associated with protein- or RNA-coding genes can be identified by <a href="/index.php/Gene_prediction" title="Gene prediction">gene finding</a> algorithms, which allow researchers to predict the presence of particular <a href="/index.php/Gene_product" title="Gene product">gene products</a> in an organism even before they have been isolated experimentally.<sup id="cite_ref-Mount_113-0" class="reference"><a href="#cite_note-Mount-113">[113]</a></sup>
</p>
<h3><span class="mw-headline" id="DNA_nanotechnology">DNA nanotechnology</span></h3>
<figure typeof="mw:File/Thumb"><a href="/index.php/File:DNA_nanostructures.png" class="mw-file-description"><img src="/images/5/55/DNA_nanostructures.png" decoding="async" width="400" height="220"></a><figcaption>The DNA structure at left (schematic shown) will self-assemble into the structure visualized by <a href="/index.php/Atomic_force_microscope" title="Atomic force microscope">atomic force microscopy</a> at right.  <a href="/index.php/DNA_nanotechnology" title="DNA nanotechnology">DNA nanotechnology</a> is the field which seeks to design nanoscale structures using the <a href="/index.php/Molecular_recognition" title="Molecular recognition">molecular recognition</a> properties of DNA molecules.  Image from Strong, 2004.<a href="/index.php?title=Template:Doi-inline&amp;action=edit&amp;redlink=1" class="new" title="Template:Doi-inline (page does not exist)">Template:Doi-inline</a></figcaption></figure>
<div role="note" class="hatnote navigation-not-searchable">Further information: [[:<a href="/index.php/DNA_nanotechnology" title="DNA nanotechnology">DNA nanotechnology</a>]]</div>
<p>DNA nanotechnology uses the unique <a href="/index.php/Molecular_recognition" title="Molecular recognition">molecular recognition</a> properties of DNA and other nucleic acids to create self-assembling branched DNA complexes with useful properties.  DNA is thus used as a structural material rather than as a carrier of biological information.  This has led to the creation of two-dimensional periodic lattices (both tile-based as well as using the "<a href="/index.php/DNA_origami" title="DNA origami">DNA origami</a>" method) as well as three-dimensional structures in the shapes of <a href="/index.php/Polyhedron" title="Polyhedron">polyhedra</a>.  <a href="/index.php/DNA_machine" title="DNA machine">Nanomechanical devices</a> and <a href="/index.php/DNA_computing" title="DNA computing">algorithmic self-assembly</a> have also been demonstrated, and these DNA structures have been used to template the arrangement of other molecules such as <a href="/index.php/Colloidal_gold" title="Colloidal gold">gold nanoparticles</a> and <a href="/index.php/Streptavidin" title="Streptavidin">streptavidin</a> proteins.
</p>
<h3><span class="mw-headline" id="History_and_anthropology">History and anthropology</span></h3>
<div role="note" class="hatnote navigation-not-searchable">Further information: [[:<a href="/index.php/Phylogenetics" title="Phylogenetics">Phylogenetics</a> and <a href="/index.php/Genetic_genealogy" title="Genetic genealogy">Genetic genealogy</a>]]</div>
<p>Because DNA collects mutations over time, which are then inherited, it contains historical information and by comparing DNA sequences, geneticists can infer the evolutionary history of organisms, their <a href="/index.php/Phylogenetics" title="Phylogenetics">phylogeny</a>.<sup id="cite_ref-114" class="reference"><a href="#cite_note-114">[114]</a></sup> This field of phylogenetics is a powerful tool in <a href="/index.php?title=Evolutionary_biology&amp;action=edit&amp;redlink=1" class="new" title="Evolutionary biology (page does not exist)">evolutionary biology</a>. If DNA sequences within a species are compared, <a href="/index.php/Population_genetics" title="Population genetics">population geneticists</a> can learn the history of particular populations. This can be used in studies ranging from <a href="/index.php/Ecological_genetics" title="Ecological genetics">ecological genetics</a> to <a href="/index.php?title=Anthropology&amp;action=edit&amp;redlink=1" class="new" title="Anthropology (page does not exist)">anthropology</a>; for example, DNA evidence is being used to try to identify the Ten Lost Tribes of Israel.<sup id="cite_ref-115" class="reference"><a href="#cite_note-115">[115]</a></sup><sup id="cite_ref-116" class="reference"><a href="#cite_note-116">[116]</a></sup>
</p><p>DNA has also been used to look at modern family relationships, such as establishing family relationships between the descendants of Sally Hemings and Thomas Jefferson. This usage is closely related to the use of DNA in criminal investigations detailed above. Indeed, some criminal investigations have been solved when DNA from crime scenes has matched relatives of the guilty individual.<sup id="cite_ref-117" class="reference"><a href="#cite_note-117">[117]</a></sup>
</p>
<h2><span class="mw-headline" id="History_of_DNA_research">History of DNA research</span></h2>
<figure class="mw-halign-right" typeof="mw:File/Thumb"><a href="/index.php/File:Rosalind_Franklin.jpg" class="mw-file-description"><img src="/images/9/97/Rosalind_Franklin.jpg" decoding="async" width="125" height="182"></a><figcaption><a href="/index.php/Rosalind_Franklin" title="Rosalind Franklin">Rosalind Franklin</a></figcaption></figure>
<figure class="mw-halign-right" typeof="mw:File/Thumb"><a href="/index.php/File:Francis_Crick.png" class="mw-file-description"><img src="/images/d/d2/Francis_Crick.png" decoding="async" width="125" height="93"></a><figcaption><a href="/index.php/Francis_Crick" title="Francis Crick">Francis Crick</a></figcaption></figure>
<figure class="mw-halign-right" typeof="mw:File/Thumb"><a href="/index.php/File:JamesDWatson.jpg" class="mw-file-description"><img src="/images/5/52/JamesDWatson.jpg" decoding="async" width="125" height="156"></a><figcaption><a href="/index.php?title=James_D._Watson&amp;action=edit&amp;redlink=1" class="new" title="James D. Watson (page does not exist)">James Watson</a></figcaption></figure>
<div role="note" class="hatnote navigation-not-searchable">Further information: [[:<a href="/index.php/History_of_molecular_biology" title="History of molecular biology">History of molecular biology</a>]]</div>
<p>DNA was first isolated by the Swiss physician <a href="/index.php?title=Friedrich_Miescher&amp;action=edit&amp;redlink=1" class="new" title="Friedrich Miescher (page does not exist)">Friedrich Miescher</a> who, in 1869, discovered a microscopic substance in the <a href="/index.php/Pus" title="Pus">pus</a> of discarded surgical bandages. As it resided in the nuclei of cells, he called it "nuclein".<sup id="cite_ref-118" class="reference"><a href="#cite_note-118">[118]</a></sup> In 1919 this discovery was followed by <a href="/index.php?title=Phoebus_Levene&amp;action=edit&amp;redlink=1" class="new" title="Phoebus Levene (page does not exist)">Phoebus Levene</a>'s identification of the base, sugar and phosphate nucleotide unit.<sup id="cite_ref-119" class="reference"><a href="#cite_note-119">[119]</a></sup> Levene suggested that DNA consisted of a string of nucleotide units linked together through the phosphate groups. However, Levene thought the chain was short and the bases repeated in a fixed order. In 1937 <a href="/index.php?title=William_Astbury&amp;action=edit&amp;redlink=1" class="new" title="William Astbury (page does not exist)">William Astbury</a> produced the first <a href="/index.php/X-ray_scattering_techniques" title="X-ray scattering techniques">X-ray diffraction</a> patterns that showed that DNA had a regular structure.<sup id="cite_ref-120" class="reference"><a href="#cite_note-120">[120]</a></sup>
</p><p>In 1928, <a href="/index.php?title=Frederick_Griffith&amp;action=edit&amp;redlink=1" class="new" title="Frederick Griffith (page does not exist)">Frederick Griffith</a> discovered that <a href="/index.php/Trait_(biology)" title="Trait (biology)">traits</a> of the "smooth" form of the <i>Pneumococcus</i> could be transferred to the "rough" form of the same bacteria by mixing killed "smooth" bacteria with the live "rough" form.<sup id="cite_ref-121" class="reference"><a href="#cite_note-121">[121]</a></sup> This system provided the first clear suggestion that DNA carried genetic information, when <a href="/index.php?title=Oswald_Avery&amp;action=edit&amp;redlink=1" class="new" title="Oswald Avery (page does not exist)">Oswald Avery</a>, along with coworkers <a href="/index.php?title=Colin_Munro_MacLeod&amp;action=edit&amp;redlink=1" class="new" title="Colin Munro MacLeod (page does not exist)">Colin MacLeod</a> and <a href="/index.php?title=Maclyn_McCarty&amp;action=edit&amp;redlink=1" class="new" title="Maclyn McCarty (page does not exist)">Maclyn McCarty</a>, identified DNA as the <a href="/index.php?title=Griffith%27s_experiment&amp;action=edit&amp;redlink=1" class="new" title="Griffith's experiment (page does not exist)">transforming principle</a> in 1943.<sup id="cite_ref-122" class="reference"><a href="#cite_note-122">[122]</a></sup> DNA's role in <a href="/index.php/Heredity" title="Heredity">heredity</a> was confirmed in 1952, when <a href="/index.php?title=Alfred_Hershey&amp;action=edit&amp;redlink=1" class="new" title="Alfred Hershey (page does not exist)">Alfred Hershey</a> and <a href="/index.php?title=Martha_Chase&amp;action=edit&amp;redlink=1" class="new" title="Martha Chase (page does not exist)">Martha Chase</a> in the <a href="/index.php/Hershey-Chase_experiment" title="Hershey-Chase experiment">Hershey-Chase experiment</a> showed that DNA is the <a href="/index.php/Genetic_material" title="Genetic material">genetic material</a> of the <a href="/index.php/Enterobacteria_phage_T2" title="Enterobacteria phage T2">T2 phage</a>.<sup id="cite_ref-123" class="reference"><a href="#cite_note-123">[123]</a></sup>
</p><p>In 1953, based on <a href="/index.php/Photo_51" title="Photo 51">X-ray diffraction images</a><sup id="cite_ref-FWPUB_124-0" class="reference"><a href="#cite_note-FWPUB-124">[124]</a></sup> taken by <a href="/index.php/Rosalind_Franklin" title="Rosalind Franklin">Rosalind Franklin</a> and the information that the bases were paired, <a href="/index.php?title=James_D._Watson&amp;action=edit&amp;redlink=1" class="new" title="James D. Watson (page does not exist)">James D. Watson</a> and <a href="/index.php/Francis_Crick" title="Francis Crick">Francis Crick</a> suggested<sup id="cite_ref-FWPUB_124-1" class="reference"><a href="#cite_note-FWPUB-124">[124]</a></sup> what is now accepted as the first accurate model of <a href="/index.php?title=Molecular_structure_of_Nucleic_Acids&amp;action=edit&amp;redlink=1" class="new" title="Molecular structure of Nucleic Acids (page does not exist)">DNA structure</a> in the journal <a href="/index.php/Nature_(journal)" title="Nature (journal)"><i>Nature</i></a>.<sup id="cite_ref-Watson_5-1" class="reference"><a href="#cite_note-Watson-5">[5]</a></sup> Experimental evidence for Watson and Crick's model were published in a series of five articles in the same issue of <i>Nature</i>.<sup id="cite_ref-NatureDNA50_125-0" class="reference"><a href="#cite_note-NatureDNA50-125">[125]</a></sup> Of these, <a href="/index.php/Rosalind_Franklin" title="Rosalind Franklin">Franklin</a> and <a href="/index.php/Raymond_Gosling" title="Raymond Gosling">Raymond Gosling</a>'s paper was the first publication of X-ray diffraction data that supported the Watson and Crick model,<sup id="cite_ref-NatFranGos_126-0" class="reference"><a href="#cite_note-NatFranGos-126">[126]</a></sup><sup id="cite_ref-127" class="reference"><a href="#cite_note-127">[127]</a></sup> this issue also contained an article on DNA structure by <a href="/index.php?title=Maurice_Wilkins&amp;action=edit&amp;redlink=1" class="new" title="Maurice Wilkins (page does not exist)">Maurice Wilkins</a> and his colleagues.<sup id="cite_ref-NatWilk_128-0" class="reference"><a href="#cite_note-NatWilk-128">[128]</a></sup> In 1962, after Franklin's death, Watson, Crick, and Wilkins jointly received the <a href="/index.php/Nobel_Prize" title="Nobel Prize">Nobel Prize</a> in <a href="/index.php/Nobel_Prize_in_Physiology_or_Medicine" title="Nobel Prize in Physiology or Medicine">Physiology or Medicine</a>.<sup id="cite_ref-129" class="reference"><a href="#cite_note-129">[129]</a></sup> However, debate continues on who should receive credit for the discovery.<sup id="cite_ref-130" class="reference"><a href="#cite_note-130">[130]</a></sup> 
</p><p>In an influential presentation in 1957, Crick laid out the <a href="/index.php/Central_dogma_of_molecular_biology" title="Central dogma of molecular biology">"Central Dogma" of molecular biology</a>, which foretold the relationship between DNA, RNA, and proteins, and articulated the "adaptor hypothesis".<sup id="cite_ref-131" class="reference"><a href="#cite_note-131">[131]</a></sup> Final confirmation of the replication mechanism that was implied by the double-helical structure followed in 1958 through the <a href="/index.php/Meselson-Stahl_experiment" title="Meselson-Stahl experiment">Meselson-Stahl experiment</a>.<sup id="cite_ref-132" class="reference"><a href="#cite_note-132">[132]</a></sup> Further work by Crick and coworkers showed that the genetic code was based on non-overlapping triplets of bases, called codons, allowing <a href="/index.php?title=Har_Gobind_Khorana&amp;action=edit&amp;redlink=1" class="new" title="Har Gobind Khorana (page does not exist)">Har Gobind Khorana</a>, <a href="/index.php?title=Robert_W._Holley&amp;action=edit&amp;redlink=1" class="new" title="Robert W. Holley (page does not exist)">Robert W. Holley</a> and <a href="/index.php?title=Marshall_Warren_Nirenberg&amp;action=edit&amp;redlink=1" class="new" title="Marshall Warren Nirenberg (page does not exist)">Marshall Warren Nirenberg</a> to decipher the genetic code.<sup id="cite_ref-133" class="reference"><a href="#cite_note-133">[133]</a></sup> These findings represent the birth of <a href="/index.php/Molecular_biology" title="Molecular biology">molecular biology</a>.
</p>
<h2><span class="mw-headline" id="See_also">See also</span></h2>
<div style="-moz-column-count:2; column-count:2;">
<ul><li><a href="/index.php/Genetic_disorder" title="Genetic disorder">Genetic disorder</a></li>
<li><a href="/index.php/Plasmid" title="Plasmid">Plasmid</a></li>
<li><a href="/index.php/Junk_DNA" title="Junk DNA">Junk DNA</a></li>
<li><a href="/index.php/Triple-stranded_DNA" title="Triple-stranded DNA">Triple-stranded DNA</a></li>
<li><a href="/index.php/Nucleic_acid_analogues" title="Nucleic acid analogues">Nucleic acid analogues</a></li>
<li><a href="/index.php/Nucleic_acid_methods" title="Nucleic acid methods">Nucleic acid methods</a></li>
<li><a href="/index.php/DNA_sequencing" title="DNA sequencing">DNA sequencing</a></li>
<li><a href="/index.php/Southern_blot" title="Southern blot">Southern blot</a></li>
<li><a href="/index.php/DNA_microarray" title="DNA microarray">DNA microarray</a></li>
<li><a href="/index.php/Polymerase_chain_reaction" title="Polymerase chain reaction">Polymerase chain reaction</a></li>
<li><a href="/index.php/Phosphoramidite" title="Phosphoramidite">Phosphoramidite</a></li></ul>
</div>
<h2><span class="mw-headline" id="References">References</span></h2>
<div class="reflist columns references-column-width" style="-moz-column-width: 30em; -webkit-column-width: 30em; column-width: 30em; list-style-type: decimal;">
<ol class="references">
<li id="cite_note-Alberts-1"><span class="mw-cite-backlink">↑ <sup><a href="#cite_ref-Alberts_1-0"><span class="cite-accessibility-label">Jump up to: </span>1.0</a></sup> <sup><a href="#cite_ref-Alberts_1-1">1.1</a></sup></span> <span class="reference-text"><cite class="citation book">Alberts, Bruce (2002). <a rel="nofollow" class="external text" href="http://www.ncbi.nlm.nih.gov/books/bv.fcgi?call=bv.View..ShowTOC&amp;rid=mboc4.TOC&amp;depth=2"><i>Molecular Biology of the Cell; Fourth Edition</i></a>. New York and London: Garland Science. <a href="/index.php?title=International_Standard_Book_Number&amp;action=edit&amp;redlink=1" class="new" title="International Standard Book Number (page does not exist)">ISBN</a>&nbsp;<a href="/index.php/Special:BookSources/0-8153-3218-1" title="Special:BookSources/0-8153-3218-1">0-8153-3218-1</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=book&amp;rft.btitle=Molecular+Biology+of+the+Cell%3B+Fourth+Edition&amp;rft.place=New+York+and+London&amp;rft.pub=Garland+Science&amp;rft.date=2002&amp;rft.isbn=0-8153-3218-1&amp;rft.aulast=Alberts&amp;rft.aufirst=Bruce&amp;rft_id=http%3A%2F%2Fwww.ncbi.nlm.nih.gov%2Fbooks%2Fbv.fcgi%3Fcall%3Dbv.View..ShowTOC%26rid%3Dmboc4.TOC%26depth%3D2&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span> <span class="cs1-visible-error error citation-comment">Unknown parameter <code class="cs1-code">|coauthors=</code> ignored (<a href="/index.php?title=Help:CS1_errors&amp;action=edit&amp;redlink=1" class="new" title="Help:CS1 errors (page does not exist)">help</a>)</span><style data-mw-deduplicate="TemplateStyles:r1594688">.mw-parser-output cite.citation{font-style:inherit}.mw-parser-output q{quotes:"\"""\"""'""'"}.mw-parser-output code.cs1-code{color:inherit;background:inherit;border:inherit;padding:inherit}.mw-parser-output .cs1-lock-free a{background:url("https://upload.wikimedia.org/wikipedia/commons/thumb/6/65/Lock-green.svg/9px-Lock-green.svg.png")no-repeat;background-position:right .1em center}.mw-parser-output .cs1-lock-limited a,.mw-parser-output .cs1-lock-registration a{background:url("https://upload.wikimedia.org/wikipedia/commons/thumb/d/d6/Lock-gray-alt-2.svg/9px-Lock-gray-alt-2.svg.png")no-repeat;background-position:right .1em center}.mw-parser-output .cs1-lock-subscription a{background:url("https://upload.wikimedia.org/wikipedia/commons/thumb/a/aa/Lock-red-alt-2.svg/9px-Lock-red-alt-2.svg.png")no-repeat;background-position:right .1em center}.mw-parser-output .cs1-subscription,.mw-parser-output .cs1-registration{color:#555}.mw-parser-output .cs1-subscription span,.mw-parser-output .cs1-registration span{border-bottom:1px dotted;cursor:help}.mw-parser-output .cs1-hidden-error{display:none;font-size:100%}.mw-parser-output .cs1-visible-error{display:none;font-size:100%}.mw-parser-output .cs1-subscription,.mw-parser-output .cs1-registration,.mw-parser-output .cs1-format{font-size:95%}.mw-parser-output .cs1-kern-left,.mw-parser-output .cs1-kern-wl-left{padding-left:0.2em}.mw-parser-output .cs1-kern-right,.mw-parser-output .cs1-kern-wl-right{padding-right:0.2em}</style></span>
</li>
<li id="cite_note-Butler-2"><span class="mw-cite-backlink"><a href="#cite_ref-Butler_2-0" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation book">Butler, John M. (2001). <i>Forensic DNA Typing</i>. Elsevier. <a href="/index.php?title=International_Standard_Book_Number&amp;action=edit&amp;redlink=1" class="new" title="International Standard Book Number (page does not exist)">ISBN</a>&nbsp;<a href="/index.php/Special:BookSources/978-0-12-147951-0" title="Special:BookSources/978-0-12-147951-0">978-0-12-147951-0</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=book&amp;rft.btitle=Forensic+DNA+Typing&amp;rft.pub=Elsevier&amp;rft.date=2001&amp;rft.isbn=978-0-12-147951-0&amp;rft.au=Butler%2C+John+M.&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688">pp. 14–15.</span>
</li>
<li id="cite_note-3"><span class="mw-cite-backlink"><a href="#cite_ref-3" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Mandelkern M, Elias J, Eden D, Crothers D (1981). "The dimensions of DNA in solution". <i>J Mol Biol</i>. <b>152</b> (1): 153–61. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/7338906">7338906</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=J+Mol+Biol&amp;rft.atitle=The+dimensions+of+DNA+in+solution&amp;rft.volume=152&amp;rft.issue=1&amp;rft.pages=153-61&amp;rft.date=1981&amp;rft_id=info%3Apmid%2F7338906&amp;rft.au=Mandelkern+M%2C+Elias+J%2C+Eden+D%2C+Crothers+D&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-4"><span class="mw-cite-backlink"><a href="#cite_ref-4" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Gregory S;  et al. (2006). "The DNA sequence and biological annotation of human chromosome 1". <i>Nature</i>. <b>441</b> (7091): 315–21. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1038%2Fnature04727">10.1038/nature04727</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/16710414">16710414</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Nature&amp;rft.atitle=The+DNA+sequence+and+biological+annotation+of+human+chromosome+1&amp;rft.volume=441&amp;rft.issue=7091&amp;rft.pages=315-21&amp;rft.date=2006&amp;rft_id=info%3Adoi%2F10.1038%2Fnature04727&amp;rft_id=info%3Apmid%2F16710414&amp;rft.au=Gregory+S&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Explicit use of et al. (<a href="/index.php?title=Category:CS1_maint:_Explicit_use_of_et_al.&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Explicit use of et al. (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-Watson-5"><span class="mw-cite-backlink">↑ <sup><a href="#cite_ref-Watson_5-0"><span class="cite-accessibility-label">Jump up to: </span>5.0</a></sup> <sup><a href="#cite_ref-Watson_5-1">5.1</a></sup></span> <span class="reference-text"><cite class="citation journal">Watson J, Crick F (1953). <a rel="nofollow" class="external text" href="http://profiles.nlm.nih.gov/SC/B/B/Y/W/_/scbbyw.pdf">"Molecular structure of nucleic acids; a structure for deoxyribose nucleic acid"</a> <span class="cs1-format">(PDF)</span>. <i>Nature</i>. <b>171</b> (4356): 737–8. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/13054692">13054692</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Nature&amp;rft.atitle=Molecular+structure+of+nucleic+acids%3B+a+structure+for+deoxyribose+nucleic+acid&amp;rft.volume=171&amp;rft.issue=4356&amp;rft.pages=737-8&amp;rft.date=1953&amp;rft_id=info%3Apmid%2F13054692&amp;rft.au=Watson+J%2C+Crick+F&amp;rft_id=http%3A%2F%2Fprofiles.nlm.nih.gov%2FSC%2FB%2FB%2FY%2FW%2F_%2Fscbbyw.pdf&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-berg-6"><span class="mw-cite-backlink">↑ <sup><a href="#cite_ref-berg_6-0"><span class="cite-accessibility-label">Jump up to: </span>6.0</a></sup> <sup><a href="#cite_ref-berg_6-1">6.1</a></sup> <sup><a href="#cite_ref-berg_6-2">6.2</a></sup></span> <span class="reference-text">Berg J., Tymoczko J. and Stryer L. (2002) <i>Biochemistry.</i> W. H. Freeman and Company ISBN 0-7167-4955-6</span>
</li>
<li id="cite_note-IUPAC-7"><span class="mw-cite-backlink"><a href="#cite_ref-IUPAC_7-0" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><a rel="nofollow" class="external text" href="http://www.chem.qmul.ac.uk/iupac/misc/naabb.html">Abbreviations and Symbols for Nucleic Acids, Polynucleotides and their Constituents</a> IUPAC-IUB Commission on Biochemical Nomenclature (CBN), Accessed 03 Jan 2006</span>
</li>
<li id="cite_note-Ghosh-8"><span class="mw-cite-backlink">↑ <sup><a href="#cite_ref-Ghosh_8-0"><span class="cite-accessibility-label">Jump up to: </span>8.0</a></sup> <sup><a href="#cite_ref-Ghosh_8-1">8.1</a></sup></span> <span class="reference-text"><cite class="citation journal">Ghosh A, Bansal M (2003). "A glossary of DNA structures from A to Z". <i>Acta Crystallogr D Biol Crystallogr</i>. <b>59</b> (Pt 4): 620–6. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/12657780">12657780</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Acta+Crystallogr+D+Biol+Crystallogr&amp;rft.atitle=A+glossary+of+DNA+structures+from+A+to+Z&amp;rft.volume=59&amp;rft.issue=Pt+4&amp;rft.pages=620-6&amp;rft.date=2003&amp;rft_id=info%3Apmid%2F12657780&amp;rft.au=Ghosh+A%2C+Bansal+M&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-9"><span class="mw-cite-backlink"><a href="#cite_ref-9" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text">Created from <a rel="nofollow" class="external text" href="http://www.rcsb.org/pdb/cgi/explore.cgi?pdbId=1D65">PDB 1D65</a></span>
</li>
<li id="cite_note-10"><span class="mw-cite-backlink"><a href="#cite_ref-10" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Wing R, Drew H, Takano T, Broka C, Tanaka S, Itakura K, Dickerson R (1980). "Crystal structure analysis of a complete turn of B-DNA". <i>Nature</i>. <b>287</b> (5784): 755–8. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/7432492">7432492</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Nature&amp;rft.atitle=Crystal+structure+analysis+of+a+complete+turn+of+B-DNA&amp;rft.volume=287&amp;rft.issue=5784&amp;rft.pages=755-8&amp;rft.date=1980&amp;rft_id=info%3Apmid%2F7432492&amp;rft.au=Wing+R%2C+Drew+H%2C+Takano+T%2C+Broka+C%2C+Tanaka+S%2C+Itakura+K%2C+Dickerson+R&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-11"><span class="mw-cite-backlink"><a href="#cite_ref-11" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Pabo C, Sauer R (1984). "Protein-DNA recognition". <i>Annu Rev Biochem</i>. <b>53</b>: 293–321. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1146%2Fannurev.bi.53.070184.001453">10.1146/annurev.bi.53.070184.001453</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/6236744">6236744</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Annu+Rev+Biochem&amp;rft.atitle=Protein-DNA+recognition&amp;rft.volume=53&amp;rft.pages=293-321&amp;rft.date=1984&amp;rft_id=info%3Adoi%2F10.1146%2Fannurev.bi.53.070184.001453&amp;rft_id=info%3Apmid%2F6236744&amp;rft.au=Pabo+C%2C+Sauer+R&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-12"><span class="mw-cite-backlink"><a href="#cite_ref-12" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Ponnuswamy P, Gromiha M (1994). "On the conformational stability of oligonucleotide duplexes and tRNA molecules". <i>J Theor Biol</i>. <b>169</b> (4): 419–32. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/7526075">7526075</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=J+Theor+Biol&amp;rft.atitle=On+the+conformational+stability+of+oligonucleotide+duplexes+and+tRNA+molecules&amp;rft.volume=169&amp;rft.issue=4&amp;rft.pages=419-32&amp;rft.date=1994&amp;rft_id=info%3Apmid%2F7526075&amp;rft.au=Ponnuswamy+P%2C+Gromiha+M&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-13"><span class="mw-cite-backlink"><a href="#cite_ref-13" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Clausen-Schaumann H, Rief M, Tolksdorf C, Gaub H (2000). <a rel="nofollow" class="external text" href="http://www.pubmedcentral.nih.gov/picrender.fcgi?artid=1300792&amp;blobtype=pdf">"Mechanical stability of single DNA molecules"</a>. <i>Biophys J</i>. <b>78</b> (4): 1997–2007. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/10733978">10733978</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Biophys+J&amp;rft.atitle=Mechanical+stability+of+single+DNA+molecules&amp;rft.volume=78&amp;rft.issue=4&amp;rft.pages=1997-2007&amp;rft.date=2000&amp;rft_id=info%3Apmid%2F10733978&amp;rft.au=Clausen-Schaumann+H%2C+Rief+M%2C+Tolksdorf+C%2C+Gaub+H&amp;rft_id=http%3A%2F%2Fwww.pubmedcentral.nih.gov%2Fpicrender.fcgi%3Fartid%3D1300792%26blobtype%3Dpdf&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-14"><span class="mw-cite-backlink"><a href="#cite_ref-14" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Chalikian T, Völker J, Plum G, Breslauer K (1999). <a rel="nofollow" class="external text" href="http://www.pubmedcentral.nih.gov/picrender.fcgi?artid=22151&amp;blobtype=pdf">"A more unified picture for the thermodynamics of nucleic acid duplex melting: a characterization by calorimetric and volumetric techniques"</a>. <i>Proc Natl Acad Sci U S A</i>. <b>96</b> (14): 7853–8. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/10393911">10393911</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Proc+Natl+Acad+Sci+U+S+A&amp;rft.atitle=A+more+unified+picture+for+the+thermodynamics+of+nucleic+acid+duplex+melting%3A+a+characterization+by+calorimetric+and+volumetric+techniques&amp;rft.volume=96&amp;rft.issue=14&amp;rft.pages=7853-8&amp;rft.date=1999&amp;rft_id=info%3Apmid%2F10393911&amp;rft.au=Chalikian+T%2C+V%C3%B6lker+J%2C+Plum+G%2C+Breslauer+K&amp;rft_id=http%3A%2F%2Fwww.pubmedcentral.nih.gov%2Fpicrender.fcgi%3Fartid%3D22151%26blobtype%3Dpdf&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-15"><span class="mw-cite-backlink"><a href="#cite_ref-15" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">deHaseth P, Helmann J (1995). "Open complex formation by Escherichia coli RNA polymerase: the mechanism of polymerase-induced strand separation of double helical DNA". <i>Mol Microbiol</i>. <b>16</b> (5): 817–24. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/7476180">7476180</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Mol+Microbiol&amp;rft.atitle=Open+complex+formation+by+Escherichia+coli+RNA+polymerase%3A+the+mechanism+of+polymerase-induced+strand+separation+of+double+helical+DNA&amp;rft.volume=16&amp;rft.issue=5&amp;rft.pages=817-24&amp;rft.date=1995&amp;rft_id=info%3Apmid%2F7476180&amp;rft.au=deHaseth+P%2C+Helmann+J&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-16"><span class="mw-cite-backlink"><a href="#cite_ref-16" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Isaksson J, Acharya S, Barman J, Cheruku P, Chattopadhyaya J (2004). "Single-stranded adenine-rich DNA and RNA retain structural characteristics of their respective double-stranded conformations and show directional differences in stacking pattern". <i>Biochemistry</i>. <b>43</b> (51): 15996–6010. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1021%2Fbi048221v">10.1021/bi048221v</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/15609994">15609994</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Biochemistry&amp;rft.atitle=Single-stranded+adenine-rich+DNA+and+RNA+retain+structural+characteristics+of+their+respective+double-stranded+conformations+and+show+directional+differences+in+stacking+pattern&amp;rft.volume=43&amp;rft.issue=51&amp;rft.pages=15996-6010&amp;rft.date=2004&amp;rft_id=info%3Adoi%2F10.1021%2Fbi048221v&amp;rft_id=info%3Apmid%2F15609994&amp;rft.au=Isaksson+J%2C+Acharya+S%2C+Barman+J%2C+Cheruku+P%2C+Chattopadhyaya+J&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-17"><span class="mw-cite-backlink"><a href="#cite_ref-17" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><a rel="nofollow" class="external text" href="http://www.chem.qmul.ac.uk/iubmb/newsletter/misc/DNA.html">Designation of the two strands of DNA</a> JCBN/NC-IUB Newsletter 1989, Accessed 07 May 2008</span>
</li>
<li id="cite_note-18"><span class="mw-cite-backlink"><a href="#cite_ref-18" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Hüttenhofer A, Schattner P, Polacek N (2005). "Non-coding RNAs: hope or hype?". <i>Trends Genet</i>. <b>21</b> (5): 289–97. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/15851066">15851066</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Trends+Genet&amp;rft.atitle=Non-coding+RNAs%3A+hope+or+hype%3F&amp;rft.volume=21&amp;rft.issue=5&amp;rft.pages=289-97&amp;rft.date=2005&amp;rft_id=info%3Apmid%2F15851066&amp;rft.au=H%C3%BCttenhofer+A%2C+Schattner+P%2C+Polacek+N&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-19"><span class="mw-cite-backlink"><a href="#cite_ref-19" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Munroe S (2004). "Diversity of antisense regulation in eukaryotes: multiple mechanisms, emerging patterns". <i>J Cell Biochem</i>. <b>93</b> (4): 664–71. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1002%2Fjcb.20252">10.1002/jcb.20252</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/15389973">15389973</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=J+Cell+Biochem&amp;rft.atitle=Diversity+of+antisense+regulation+in+eukaryotes%3A+multiple+mechanisms%2C+emerging+patterns&amp;rft.volume=93&amp;rft.issue=4&amp;rft.pages=664-71&amp;rft.date=2004&amp;rft_id=info%3Adoi%2F10.1002%2Fjcb.20252&amp;rft_id=info%3Apmid%2F15389973&amp;rft.au=Munroe+S&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-20"><span class="mw-cite-backlink"><a href="#cite_ref-20" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Makalowska I, Lin C, Makalowski W (2005). "Overlapping genes in vertebrate genomes". <i>Comput Biol Chem</i>. <b>29</b> (1): 1&amp;ndash, 12. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/15680581">15680581</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Comput+Biol+Chem&amp;rft.atitle=Overlapping+genes+in+vertebrate+genomes&amp;rft.volume=29&amp;rft.issue=1&amp;rft.pages=1%26ndash%2C+12&amp;rft.date=2005&amp;rft_id=info%3Apmid%2F15680581&amp;rft.au=Makalowska+I%2C+Lin+C%2C+Makalowski+W&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-21"><span class="mw-cite-backlink"><a href="#cite_ref-21" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Johnson Z, Chisholm S (2004). "Properties of overlapping genes are conserved across microbial genomes". <i>Genome Res</i>. <b>14</b> (11): 2268–72. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1101%2Fgr.2433104">10.1101/gr.2433104</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/15520290">15520290</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Genome+Res&amp;rft.atitle=Properties+of+overlapping+genes+are+conserved+across+microbial+genomes&amp;rft.volume=14&amp;rft.issue=11&amp;rft.pages=2268-72&amp;rft.date=2004&amp;rft_id=info%3Adoi%2F10.1101%2Fgr.2433104&amp;rft_id=info%3Apmid%2F15520290&amp;rft.au=Johnson+Z%2C+Chisholm+S&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-22"><span class="mw-cite-backlink"><a href="#cite_ref-22" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Lamb R, Horvath C (1991). "Diversity of coding strategies in influenza viruses". <i>Trends Genet</i>. <b>7</b> (8): 261–6. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/1771674">1771674</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Trends+Genet&amp;rft.atitle=Diversity+of+coding+strategies+in+influenza+viruses&amp;rft.volume=7&amp;rft.issue=8&amp;rft.pages=261-6&amp;rft.date=1991&amp;rft_id=info%3Apmid%2F1771674&amp;rft.au=Lamb+R%2C+Horvath+C&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-23"><span class="mw-cite-backlink"><a href="#cite_ref-23" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Benham C, Mielke S (2005). "DNA mechanics". <i>Annu Rev Biomed Eng</i>. <b>7</b>: 21–53. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1146%2Fannurev.bioeng.6.062403.132016">10.1146/annurev.bioeng.6.062403.132016</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/16004565">16004565</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Annu+Rev+Biomed+Eng&amp;rft.atitle=DNA+mechanics&amp;rft.volume=7&amp;rft.pages=21-53&amp;rft.date=2005&amp;rft_id=info%3Adoi%2F10.1146%2Fannurev.bioeng.6.062403.132016&amp;rft_id=info%3Apmid%2F16004565&amp;rft.au=Benham+C%2C+Mielke+S&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-Champoux-24"><span class="mw-cite-backlink">↑ <sup><a href="#cite_ref-Champoux_24-0"><span class="cite-accessibility-label">Jump up to: </span>24.0</a></sup> <sup><a href="#cite_ref-Champoux_24-1">24.1</a></sup></span> <span class="reference-text"><cite class="citation journal">Champoux J (2001). "DNA topoisomerases: structure, function, and mechanism". <i>Annu Rev Biochem</i>. <b>70</b>: 369–413. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1146%2Fannurev.biochem.70.1.369">10.1146/annurev.biochem.70.1.369</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/11395412">11395412</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Annu+Rev+Biochem&amp;rft.atitle=DNA+topoisomerases%3A+structure%2C+function%2C+and+mechanism&amp;rft.volume=70&amp;rft.pages=369-413&amp;rft.date=2001&amp;rft_id=info%3Adoi%2F10.1146%2Fannurev.biochem.70.1.369&amp;rft_id=info%3Apmid%2F11395412&amp;rft.au=Champoux+J&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-Wang-25"><span class="mw-cite-backlink">↑ <sup><a href="#cite_ref-Wang_25-0"><span class="cite-accessibility-label">Jump up to: </span>25.0</a></sup> <sup><a href="#cite_ref-Wang_25-1">25.1</a></sup></span> <span class="reference-text"><cite class="citation journal">Wang J (2002). "Cellular roles of DNA topoisomerases: a molecular perspective". <i>Nat Rev Mol Cell Biol</i>. <b>3</b> (6): 430–40. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1038%2Fnrm831">10.1038/nrm831</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/12042765">12042765</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Nat+Rev+Mol+Cell+Biol&amp;rft.atitle=Cellular+roles+of+DNA+topoisomerases%3A+a+molecular+perspective&amp;rft.volume=3&amp;rft.issue=6&amp;rft.pages=430-40&amp;rft.date=2002&amp;rft_id=info%3Adoi%2F10.1038%2Fnrm831&amp;rft_id=info%3Apmid%2F12042765&amp;rft.au=Wang+J&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-26"><span class="mw-cite-backlink"><a href="#cite_ref-26" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Basu H, Feuerstein B, Zarling D, Shafer R, Marton L (1988). "Recognition of Z-RNA and Z-DNA determinants by polyamines in solution: experimental and theoretical studies". <i>J Biomol Struct Dyn</i>. <b>6</b> (2): 299&amp;ndash, 309. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/2482766">2482766</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=J+Biomol+Struct+Dyn&amp;rft.atitle=Recognition+of+Z-RNA+and+Z-DNA+determinants+by+polyamines+in+solution%3A+experimental+and+theoretical+studies&amp;rft.volume=6&amp;rft.issue=2&amp;rft.pages=299%26ndash%2C+309&amp;rft.date=1988&amp;rft_id=info%3Apmid%2F2482766&amp;rft.au=Basu+H%2C+Feuerstein+B%2C+Zarling+D%2C+Shafer+R%2C+Marton+L&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-27"><span class="mw-cite-backlink"><a href="#cite_ref-27" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Leslie AG, Arnott S, Chandrasekaran R, Ratliff RL (1980). "Polymorphism of DNA double helices". <i>J. Mol. Biol</i>. <b>143</b> (1): 49–72. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/7441761">7441761</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=J.+Mol.+Biol.&amp;rft.atitle=Polymorphism+of+DNA+double+helices&amp;rft.volume=143&amp;rft.issue=1&amp;rft.pages=49-72&amp;rft.date=1980&amp;rft_id=info%3Apmid%2F7441761&amp;rft.au=Leslie+AG%2C+Arnott+S%2C+Chandrasekaran+R%2C+Ratliff+RL&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-28"><span class="mw-cite-backlink"><a href="#cite_ref-28" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Wahl M, Sundaralingam M (1997). "Crystal structures of A-DNA duplexes". <i>Biopolymers</i>. <b>44</b> (1): 45&amp;ndash, 63. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1002%2F%28SICI%291097-0282%281997%2944%3A1">10.1002/(SICI)1097-0282(1997)44:1</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/9097733">9097733</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Biopolymers&amp;rft.atitle=Crystal+structures+of+A-DNA+duplexes&amp;rft.volume=44&amp;rft.issue=1&amp;rft.pages=45%26ndash%2C+63&amp;rft.date=1997&amp;rft_id=info%3Adoi%2F10.1002%2F%28SICI%291097-0282%281997%2944%3A1&amp;rft_id=info%3Apmid%2F9097733&amp;rft.au=Wahl+M%2C+Sundaralingam+M&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-29"><span class="mw-cite-backlink"><a href="#cite_ref-29" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Lu XJ, Shakked Z, Olson WK (2000). "A-form conformational motifs in ligand-bound DNA structures". <i>J. Mol. Biol</i>. <b>300</b> (4): 819–40. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/10891271">10891271</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=J.+Mol.+Biol.&amp;rft.atitle=A-form+conformational+motifs+in+ligand-bound+DNA+structures&amp;rft.volume=300&amp;rft.issue=4&amp;rft.pages=819-40&amp;rft.date=2000&amp;rft_id=info%3Apmid%2F10891271&amp;rft.au=Lu+XJ%2C+Shakked+Z%2C+Olson+WK&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-30"><span class="mw-cite-backlink"><a href="#cite_ref-30" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Rothenburg S, Koch-Nolte F, Haag F. "DNA methylation and Z-DNA formation as mediators of quantitative differences in the expression of alleles". <i>Immunol Rev</i>. <b>184</b>: 286&amp;ndash, 98. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/12086319">12086319</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Immunol+Rev&amp;rft.atitle=DNA+methylation+and+Z-DNA+formation+as+mediators+of+quantitative+differences+in+the+expression+of+alleles&amp;rft.volume=184&amp;rft.pages=286%26ndash%2C+98&amp;rft_id=info%3Apmid%2F12086319&amp;rft.au=Rothenburg+S%2C+Koch-Nolte+F%2C+Haag+F&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-31"><span class="mw-cite-backlink"><a href="#cite_ref-31" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Oh D, Kim Y, Rich A (2002). <a rel="nofollow" class="external text" href="http://www.pubmedcentral.nih.gov/articlerender.fcgi?tool=pubmed&amp;pubmedid=12486233">"Z-DNA-binding proteins can act as potent effectors of gene expression in vivo"</a>. <i>Proc. Natl. Acad. Sci. U.S.A</i>. <b>99</b> (26): 16666–71. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1073%2Fpnas.262672699">10.1073/pnas.262672699</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/12486233">12486233</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Proc.+Natl.+Acad.+Sci.+U.S.A.&amp;rft.atitle=Z-DNA-binding+proteins+can+act+as+potent+effectors+of+gene+expression+in+vivo&amp;rft.volume=99&amp;rft.issue=26&amp;rft.pages=16666-71&amp;rft.date=2002&amp;rft_id=info%3Adoi%2F10.1073%2Fpnas.262672699&amp;rft_id=info%3Apmid%2F12486233&amp;rft.au=Oh+D%2C+Kim+Y%2C+Rich+A&amp;rft_id=http%3A%2F%2Fwww.pubmedcentral.nih.gov%2Farticlerender.fcgi%3Ftool%3Dpubmed%26pubmedid%3D12486233&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-32"><span class="mw-cite-backlink"><a href="#cite_ref-32" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text">Created from <a rel="nofollow" class="external text" href="http://ndbserver.rutgers.edu/atlas/xray/structures/U/ud0017/ud0017.html">NDB UD0017</a></span>
</li>
<li id="cite_note-Greider-33"><span class="mw-cite-backlink">↑ <sup><a href="#cite_ref-Greider_33-0"><span class="cite-accessibility-label">Jump up to: </span>33.0</a></sup> <sup><a href="#cite_ref-Greider_33-1">33.1</a></sup></span> <span class="reference-text"><cite class="citation journal">Greider C, Blackburn E (1985). "Identification of a specific telomere terminal transferase activity in Tetrahymena extracts". <i>Cell</i>. <b>43</b> (2 Pt 1): 405&amp;ndash, 13. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/3907856">3907856</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Cell&amp;rft.atitle=Identification+of+a+specific+telomere+terminal+transferase+activity+in+Tetrahymena+extracts&amp;rft.volume=43&amp;rft.issue=2+Pt+1&amp;rft.pages=405%26ndash%2C+13&amp;rft.date=1985&amp;rft_id=info%3Apmid%2F3907856&amp;rft.au=Greider+C%2C+Blackburn+E&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-Nugent-34"><span class="mw-cite-backlink">↑ <sup><a href="#cite_ref-Nugent_34-0"><span class="cite-accessibility-label">Jump up to: </span>34.0</a></sup> <sup><a href="#cite_ref-Nugent_34-1">34.1</a></sup> <sup><a href="#cite_ref-Nugent_34-2">34.2</a></sup></span> <span class="reference-text"><cite class="citation journal">Nugent C, Lundblad V (1998). <a rel="nofollow" class="external text" href="http://www.genesdev.org/cgi/content/full/12/8/1073">"The telomerase reverse transcriptase: components and regulation"</a>. <i>Genes Dev</i>. <b>12</b> (8): 1073&amp;ndash, 85. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/9553037">9553037</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Genes+Dev&amp;rft.atitle=The+telomerase+reverse+transcriptase%3A+components+and+regulation&amp;rft.volume=12&amp;rft.issue=8&amp;rft.pages=1073%26ndash%2C+85&amp;rft.date=1998&amp;rft_id=info%3Apmid%2F9553037&amp;rft.au=Nugent+C%2C+Lundblad+V&amp;rft_id=http%3A%2F%2Fwww.genesdev.org%2Fcgi%2Fcontent%2Ffull%2F12%2F8%2F1073&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-35"><span class="mw-cite-backlink"><a href="#cite_ref-35" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Wright W, Tesmer V, Huffman K, Levene S, Shay J (1997). <a rel="nofollow" class="external text" href="http://www.genesdev.org/cgi/content/full/11/21/2801">"Normal human chromosomes have long G-rich telomeric overhangs at one end"</a>. <i>Genes Dev</i>. <b>11</b> (21): 2801&amp;ndash, 9. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/9353250">9353250</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Genes+Dev&amp;rft.atitle=Normal+human+chromosomes+have+long+G-rich+telomeric+overhangs+at+one+end&amp;rft.volume=11&amp;rft.issue=21&amp;rft.pages=2801%26ndash%2C+9&amp;rft.date=1997&amp;rft_id=info%3Apmid%2F9353250&amp;rft.au=Wright+W%2C+Tesmer+V%2C+Huffman+K%2C+Levene+S%2C+Shay+J&amp;rft_id=http%3A%2F%2Fwww.genesdev.org%2Fcgi%2Fcontent%2Ffull%2F11%2F21%2F2801&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-Burge-36"><span class="mw-cite-backlink">↑ <sup><a href="#cite_ref-Burge_36-0"><span class="cite-accessibility-label">Jump up to: </span>36.0</a></sup> <sup><a href="#cite_ref-Burge_36-1">36.1</a></sup></span> <span class="reference-text"><cite class="citation journal">Burge S, Parkinson G, Hazel P, Todd A, Neidle S (2006). <a rel="nofollow" class="external text" href="http://www.pubmedcentral.nih.gov/articlerender.fcgi?tool=pubmed&amp;pubmedid=17012276">"Quadruplex DNA: sequence, topology and structure"</a>. <i>Nucleic Acids Res</i>. <b>34</b> (19): 5402&amp;ndash, 15. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1093%2Fnar%2Fgkl655">10.1093/nar/gkl655</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/17012276">17012276</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Nucleic+Acids+Res&amp;rft.atitle=Quadruplex+DNA%3A+sequence%2C+topology+and+structure&amp;rft.volume=34&amp;rft.issue=19&amp;rft.pages=5402%26ndash%2C+15&amp;rft.date=2006&amp;rft_id=info%3Adoi%2F10.1093%2Fnar%2Fgkl655&amp;rft_id=info%3Apmid%2F17012276&amp;rft.au=Burge+S%2C+Parkinson+G%2C+Hazel+P%2C+Todd+A%2C+Neidle+S&amp;rft_id=http%3A%2F%2Fwww.pubmedcentral.nih.gov%2Farticlerender.fcgi%3Ftool%3Dpubmed%26pubmedid%3D17012276&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-37"><span class="mw-cite-backlink"><a href="#cite_ref-37" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Parkinson G, Lee M, Neidle S (2002). "Crystal structure of parallel quadruplexes from human telomeric DNA". <i>Nature</i>. <b>417</b> (6891): 876&amp;ndash, 80. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1038%2Fnature755">10.1038/nature755</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/12050675">12050675</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Nature&amp;rft.atitle=Crystal+structure+of+parallel+quadruplexes+from+human+telomeric+DNA&amp;rft.volume=417&amp;rft.issue=6891&amp;rft.pages=876%26ndash%2C+80&amp;rft.date=2002&amp;rft_id=info%3Adoi%2F10.1038%2Fnature755&amp;rft_id=info%3Apmid%2F12050675&amp;rft.au=Parkinson+G%2C+Lee+M%2C+Neidle+S&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-38"><span class="mw-cite-backlink"><a href="#cite_ref-38" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Griffith J, Comeau L, Rosenfield S, Stansel R, Bianchi A, Moss H, de Lange T (1999). "Mammalian telomeres end in a large duplex loop". <i>Cell</i>. <b>97</b> (4): 503&amp;ndash, 14. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/10338214">10338214</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Cell&amp;rft.atitle=Mammalian+telomeres+end+in+a+large+duplex+loop&amp;rft.volume=97&amp;rft.issue=4&amp;rft.pages=503%26ndash%2C+14&amp;rft.date=1999&amp;rft_id=info%3Apmid%2F10338214&amp;rft.au=Griffith+J%2C+Comeau+L%2C+Rosenfield+S%2C+Stansel+R%2C+Bianchi+A%2C+Moss+H%2C+de+Lange+T&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-39"><span class="mw-cite-backlink"><a href="#cite_ref-39" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Klose R, Bird A (2006). "Genomic DNA methylation: the mark and its mediators". <i>Trends Biochem Sci</i>. <b>31</b> (2): 89&amp;ndash, 97. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/16403636">16403636</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Trends+Biochem+Sci&amp;rft.atitle=Genomic+DNA+methylation%3A+the+mark+and+its+mediators&amp;rft.volume=31&amp;rft.issue=2&amp;rft.pages=89%26ndash%2C+97&amp;rft.date=2006&amp;rft_id=info%3Apmid%2F16403636&amp;rft.au=Klose+R%2C+Bird+A&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-40"><span class="mw-cite-backlink"><a href="#cite_ref-40" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Bird A (2002). "DNA methylation patterns and epigenetic memory". <i>Genes Dev</i>. <b>16</b> (1): 6&amp;ndash, 21. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/11782440">11782440</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Genes+Dev&amp;rft.atitle=DNA+methylation+patterns+and+epigenetic+memory&amp;rft.volume=16&amp;rft.issue=1&amp;rft.pages=6%26ndash%2C+21&amp;rft.date=2002&amp;rft_id=info%3Apmid%2F11782440&amp;rft.au=Bird+A&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-41"><span class="mw-cite-backlink"><a href="#cite_ref-41" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Walsh C, Xu G. "Cytosine methylation and DNA repair". <i>Curr Top Microbiol Immunol</i>. <b>301</b>: 283&amp;ndash, 315. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/16570853">16570853</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Curr+Top+Microbiol+Immunol&amp;rft.atitle=Cytosine+methylation+and+DNA+repair&amp;rft.volume=301&amp;rft.pages=283%26ndash%2C+315&amp;rft_id=info%3Apmid%2F16570853&amp;rft.au=Walsh+C%2C+Xu+G&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-42"><span class="mw-cite-backlink"><a href="#cite_ref-42" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Ratel D, Ravanat J, Berger F, Wion D (2006). "N6-methyladenine: the other methylated base of DNA". <i>Bioessays</i>. <b>28</b> (3): 309&amp;ndash, 15. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1002%2Fbies.20342">10.1002/bies.20342</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/16479578">16479578</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Bioessays&amp;rft.atitle=N6-methyladenine%3A+the+other+methylated+base+of+DNA&amp;rft.volume=28&amp;rft.issue=3&amp;rft.pages=309%26ndash%2C+15&amp;rft.date=2006&amp;rft_id=info%3Adoi%2F10.1002%2Fbies.20342&amp;rft_id=info%3Apmid%2F16479578&amp;rft.au=Ratel+D%2C+Ravanat+J%2C+Berger+F%2C+Wion+D&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-43"><span class="mw-cite-backlink"><a href="#cite_ref-43" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Gommers-Ampt J, Van Leeuwen F, de Beer A, Vliegenthart J, Dizdaroglu M, Kowalak J, Crain P, Borst P (1993). "beta-D-glucosyl-hydroxymethyluracil: a novel modified base present in the DNA of the parasitic protozoan T. brucei". <i>Cell</i>. <b>75</b> (6): 1129&amp;ndash, 36. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/8261512">8261512</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Cell&amp;rft.atitle=beta-D-glucosyl-hydroxymethyluracil%3A+a+novel+modified+base+present+in+the+DNA+of+the+parasitic+protozoan+T.+brucei&amp;rft.volume=75&amp;rft.issue=6&amp;rft.pages=1129%26ndash%2C+36&amp;rft.date=1993&amp;rft_id=info%3Apmid%2F8261512&amp;rft.au=Gommers-Ampt+J%2C+Van+Leeuwen+F%2C+de+Beer+A%2C+Vliegenthart+J%2C+Dizdaroglu+M%2C+Kowalak+J%2C+Crain+P%2C+Borst+P&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-44"><span class="mw-cite-backlink"><a href="#cite_ref-44" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text">Created from <a rel="nofollow" class="external text" href="http://www.rcsb.org/pdb/cgi/explore.cgi?pdbId=1JDG">PDB 1JDG</a></span>
</li>
<li id="cite_note-45"><span class="mw-cite-backlink"><a href="#cite_ref-45" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Douki T, Reynaud-Angelin A, Cadet J, Sage E (2003). "Bipyrimidine photoproducts rather than oxidative lesions are the main type of DNA damage involved in the genotoxic effect of solar UVA radiation". <i>Biochemistry</i>. <b>42</b> (30): 9221&amp;ndash, 6. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1021%2Fbi034593c">10.1021/bi034593c</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/12885257">12885257</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Biochemistry&amp;rft.atitle=Bipyrimidine+photoproducts+rather+than+oxidative+lesions+are+the+main+type+of+DNA+damage+involved+in+the+genotoxic+effect+of+solar+UVA+radiation&amp;rft.volume=42&amp;rft.issue=30&amp;rft.pages=9221%26ndash%2C+6&amp;rft.date=2003&amp;rft_id=info%3Adoi%2F10.1021%2Fbi034593c&amp;rft_id=info%3Apmid%2F12885257&amp;rft.au=Douki+T%2C+Reynaud-Angelin+A%2C+Cadet+J%2C+Sage+E&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688">,</span>
</li>
<li id="cite_note-46"><span class="mw-cite-backlink"><a href="#cite_ref-46" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Cadet J, Delatour T, Douki T, Gasparutto D, Pouget J, Ravanat J, Sauvaigo S (1999). "Hydroxyl radicals and DNA base damage". <i>Mutat Res</i>. <b>424</b> (1&amp;ndash, 2): 9&amp;ndash, 21. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/10064846">10064846</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Mutat+Res&amp;rft.atitle=Hydroxyl+radicals+and+DNA+base+damage&amp;rft.volume=424&amp;rft.issue=1%26ndash%2C+2&amp;rft.pages=9%26ndash%2C+21&amp;rft.date=1999&amp;rft_id=info%3Apmid%2F10064846&amp;rft.au=Cadet+J%2C+Delatour+T%2C+Douki+T%2C+Gasparutto+D%2C+Pouget+J%2C+Ravanat+J%2C+Sauvaigo+S&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-47"><span class="mw-cite-backlink"><a href="#cite_ref-47" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Shigenaga M, Gimeno C, Ames B (1989). <a rel="nofollow" class="external text" href="http://www.pnas.org/cgi/reprint/86/24/9697">"Urinary 8-hydroxy-2′-deoxyguanosine as a biological marker of <i>in vivo</i> oxidative DNA damage"</a>. <i>Proc Natl Acad Sci U S A</i>. <b>86</b> (24): 9697&amp;ndash, 701. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1073%2Fpnas.86.24.9697">10.1073/pnas.86.24.9697</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/2602371">2602371</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Proc+Natl+Acad+Sci+U+S+A&amp;rft.atitle=Urinary+8-hydroxy-2%E2%80%B2-deoxyguanosine+as+a+biological+marker+of+in+vivo+oxidative+DNA+damage&amp;rft.volume=86&amp;rft.issue=24&amp;rft.pages=9697%26ndash%2C+701&amp;rft.date=1989&amp;rft_id=info%3Adoi%2F10.1073%2Fpnas.86.24.9697&amp;rft_id=info%3Apmid%2F2602371&amp;rft.au=Shigenaga+M%2C+Gimeno+C%2C+Ames+B&amp;rft_id=http%3A%2F%2Fwww.pnas.org%2Fcgi%2Freprint%2F86%2F24%2F9697&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-48"><span class="mw-cite-backlink"><a href="#cite_ref-48" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Cathcart R, Schwiers E, Saul R, Ames B (1984). <a rel="nofollow" class="external text" href="http://www.pnas.org/cgi/reprint/81/18/5633.pdf">"Thymine glycol and thymidine glycol in human and rat urine: a possible assay for oxidative DNA damage"</a> <span class="cs1-format">(PDF)</span>. <i>Proc Natl Acad Sci U S A</i>. <b>81</b> (18): 5633&amp;ndash, 7. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/6592579">6592579</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Proc+Natl+Acad+Sci+U+S+A&amp;rft.atitle=Thymine+glycol+and+thymidine+glycol+in+human+and+rat+urine%3A+a+possible+assay+for+oxidative+DNA+damage&amp;rft.volume=81&amp;rft.issue=18&amp;rft.pages=5633%26ndash%2C+7&amp;rft.date=1984&amp;rft_id=info%3Apmid%2F6592579&amp;rft.au=Cathcart+R%2C+Schwiers+E%2C+Saul+R%2C+Ames+B&amp;rft_id=http%3A%2F%2Fwww.pnas.org%2Fcgi%2Freprint%2F81%2F18%2F5633.pdf&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-49"><span class="mw-cite-backlink"><a href="#cite_ref-49" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Valerie K, Povirk L (2003). "Regulation and mechanisms of mammalian double-strand break repair". <i>Oncogene</i>. <b>22</b> (37): 5792&amp;ndash, 812. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1038%2Fsj.onc.1206679">10.1038/sj.onc.1206679</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/12947387">12947387</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Oncogene&amp;rft.atitle=Regulation+and+mechanisms+of+mammalian+double-strand+break+repair&amp;rft.volume=22&amp;rft.issue=37&amp;rft.pages=5792%26ndash%2C+812&amp;rft.date=2003&amp;rft_id=info%3Adoi%2F10.1038%2Fsj.onc.1206679&amp;rft_id=info%3Apmid%2F12947387&amp;rft.au=Valerie+K%2C+Povirk+L&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-50"><span class="mw-cite-backlink"><a href="#cite_ref-50" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Ferguson L, Denny W (1991). "The genetic toxicology of acridines". <i>Mutat Res</i>. <b>258</b> (2): 123&amp;ndash, 60. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/1881402">1881402</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Mutat+Res&amp;rft.atitle=The+genetic+toxicology+of+acridines&amp;rft.volume=258&amp;rft.issue=2&amp;rft.pages=123%26ndash%2C+60&amp;rft.date=1991&amp;rft_id=info%3Apmid%2F1881402&amp;rft.au=Ferguson+L%2C+Denny+W&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-51"><span class="mw-cite-backlink"><a href="#cite_ref-51" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Jeffrey A (1985). "DNA modification by chemical carcinogens". <i>Pharmacol Ther</i>. <b>28</b> (2): 237&amp;ndash, 72. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/3936066">3936066</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Pharmacol+Ther&amp;rft.atitle=DNA+modification+by+chemical+carcinogens&amp;rft.volume=28&amp;rft.issue=2&amp;rft.pages=237%26ndash%2C+72&amp;rft.date=1985&amp;rft_id=info%3Apmid%2F3936066&amp;rft.au=Jeffrey+A&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-52"><span class="mw-cite-backlink"><a href="#cite_ref-52" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Stephens T, Bunde C, Fillmore B (2000). "Mechanism of action in thalidomide teratogenesis". <i>Biochem Pharmacol</i>. <b>59</b> (12): 1489&amp;ndash, 99. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/10799645">10799645</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Biochem+Pharmacol&amp;rft.atitle=Mechanism+of+action+in+thalidomide+teratogenesis&amp;rft.volume=59&amp;rft.issue=12&amp;rft.pages=1489%26ndash%2C+99&amp;rft.date=2000&amp;rft_id=info%3Apmid%2F10799645&amp;rft.au=Stephens+T%2C+Bunde+C%2C+Fillmore+B&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-53"><span class="mw-cite-backlink"><a href="#cite_ref-53" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Braña M, Cacho M, Gradillas A, de Pascual-Teresa B, Ramos A (2001). "Intercalators as anticancer drugs". <i>Curr Pharm Des</i>. <b>7</b> (17): 1745&amp;ndash, 80. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/11562309">11562309</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Curr+Pharm+Des&amp;rft.atitle=Intercalators+as+anticancer+drugs&amp;rft.volume=7&amp;rft.issue=17&amp;rft.pages=1745%26ndash%2C+80&amp;rft.date=2001&amp;rft_id=info%3Apmid%2F11562309&amp;rft.au=Bra%C3%B1a+M%2C+Cacho+M%2C+Gradillas+A%2C+de+Pascual-Teresa+B%2C+Ramos+A&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-54"><span class="mw-cite-backlink"><a href="#cite_ref-54" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Venter J;  et al. (2001). "The sequence of the human genome". <i>Science</i>. <b>291</b> (5507): 1304&amp;ndash, 51. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/11181995">11181995</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Science&amp;rft.atitle=The+sequence+of+the+human+genome&amp;rft.volume=291&amp;rft.issue=5507&amp;rft.pages=1304%26ndash%2C+51&amp;rft.date=2001&amp;rft_id=info%3Apmid%2F11181995&amp;rft.au=Venter+J&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Explicit use of et al. (<a href="/index.php?title=Category:CS1_maint:_Explicit_use_of_et_al.&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Explicit use of et al. (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-55"><span class="mw-cite-backlink"><a href="#cite_ref-55" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Thanbichler M, Wang S, Shapiro L (2005). "The bacterial nucleoid: a highly organized and dynamic structure". <i>J Cell Biochem</i>. <b>96</b> (3): 506&amp;ndash, 21. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1002%2Fjcb.20519">10.1002/jcb.20519</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/15988757">15988757</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=J+Cell+Biochem&amp;rft.atitle=The+bacterial+nucleoid%3A+a+highly+organized+and+dynamic+structure&amp;rft.volume=96&amp;rft.issue=3&amp;rft.pages=506%26ndash%2C+21&amp;rft.date=2005&amp;rft_id=info%3Adoi%2F10.1002%2Fjcb.20519&amp;rft_id=info%3Apmid%2F15988757&amp;rft.au=Thanbichler+M%2C+Wang+S%2C+Shapiro+L&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-56"><span class="mw-cite-backlink"><a href="#cite_ref-56" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Wolfsberg T, McEntyre J, Schuler G (2001). "Guide to the draft human genome". <i>Nature</i>. <b>409</b> (6822): 824&amp;ndash, 6. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1038%2F35057000">10.1038/35057000</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/11236998">11236998</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Nature&amp;rft.atitle=Guide+to+the+draft+human+genome&amp;rft.volume=409&amp;rft.issue=6822&amp;rft.pages=824%26ndash%2C+6&amp;rft.date=2001&amp;rft_id=info%3Adoi%2F10.1038%2F35057000&amp;rft_id=info%3Apmid%2F11236998&amp;rft.au=Wolfsberg+T%2C+McEntyre+J%2C+Schuler+G&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-57"><span class="mw-cite-backlink"><a href="#cite_ref-57" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Gregory T (2005). <a rel="nofollow" class="external text" href="http://aob.oxfordjournals.org/cgi/content/full/95/1/133">"The C-value enigma in plants and animals: a review of parallels and an appeal for partnership"</a>. <i>Ann Bot (Lond)</i>. <b>95</b> (1): 133&amp;ndash, 46. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/15596463">15596463</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Ann+Bot+%28Lond%29&amp;rft.atitle=The+C-value+enigma+in+plants+and+animals%3A+a+review+of+parallels+and+an+appeal+for+partnership&amp;rft.volume=95&amp;rft.issue=1&amp;rft.pages=133%26ndash%2C+46&amp;rft.date=2005&amp;rft_id=info%3Apmid%2F15596463&amp;rft.au=Gregory+T&amp;rft_id=http%3A%2F%2Faob.oxfordjournals.org%2Fcgi%2Fcontent%2Ffull%2F95%2F1%2F133&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-58"><span class="mw-cite-backlink"><a href="#cite_ref-58" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">The ENCODE Project Consortium (2007). "Identification and analysis of functional elements in 1% of the human genome by the ENCODE pilot project". <i>Nature</i>. <b>447</b> (7146): 799–816. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/doi%3A10.1038%2Fnature05874">doi:10.1038/nature05874</a> <span class="cs1-visible-error error citation-comment">Check <code class="cs1-code">|doi=</code> value (<a href="/index.php?title=Help:CS1_errors&amp;action=edit&amp;redlink=1" class="new" title="Help:CS1 errors (page does not exist)">help</a>)</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Nature&amp;rft.atitle=Identification+and+analysis+of+functional+elements+in+1%25+of+the+human+genome+by+the+ENCODE+pilot+project&amp;rft.volume=447&amp;rft.issue=7146&amp;rft.pages=799-816&amp;rft.date=2007&amp;rft_id=info%3Adoi%2Fdoi%3A10.1038%2Fnature05874&amp;rft.au=The+ENCODE+Project+Consortium&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-59"><span class="mw-cite-backlink"><a href="#cite_ref-59" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text">Created from <a rel="nofollow" class="external text" href="http://www.rcsb.org/pdb/explore/explore.do?structureId=1MSW">PDB 1MSW</a></span>
</li>
<li id="cite_note-60"><span class="mw-cite-backlink"><a href="#cite_ref-60" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Pidoux A, Allshire R (2005). <a rel="nofollow" class="external text" href="http://www.pubmedcentral.nih.gov/articlerender.fcgi?artid=1569473">"The role of heterochromatin in centromere function"</a>. <i>Philos Trans R Soc Lond B Biol Sci</i>. <b>360</b> (1455): 569&amp;ndash, 79. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1098%2Frstb.2004.1611">10.1098/rstb.2004.1611</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/15905142">15905142</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Philos+Trans+R+Soc+Lond+B+Biol+Sci&amp;rft.atitle=The+role+of+heterochromatin+in+centromere+function&amp;rft.volume=360&amp;rft.issue=1455&amp;rft.pages=569%26ndash%2C+79&amp;rft.date=2005&amp;rft_id=info%3Adoi%2F10.1098%2Frstb.2004.1611&amp;rft_id=info%3Apmid%2F15905142&amp;rft.au=Pidoux+A%2C+Allshire+R&amp;rft_id=http%3A%2F%2Fwww.pubmedcentral.nih.gov%2Farticlerender.fcgi%3Fartid%3D1569473&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-61"><span class="mw-cite-backlink"><a href="#cite_ref-61" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Harrison P, Hegyi H, Balasubramanian S, Luscombe N, Bertone P, Echols N, Johnson T, Gerstein M (2002). <a rel="nofollow" class="external text" href="http://www.genome.org/cgi/content/full/12/2/272">"Molecular fossils in the human genome: identification and analysis of the pseudogenes in chromosomes 21 and 22"</a>. <i>Genome Res</i>. <b>12</b> (2): 272&amp;ndash, 80. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/11827946">11827946</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Genome+Res&amp;rft.atitle=Molecular+fossils+in+the+human+genome%3A+identification+and+analysis+of+the+pseudogenes+in+chromosomes+21+and+22&amp;rft.volume=12&amp;rft.issue=2&amp;rft.pages=272%26ndash%2C+80&amp;rft.date=2002&amp;rft_id=info%3Apmid%2F11827946&amp;rft.au=Harrison+P%2C+Hegyi+H%2C+Balasubramanian+S%2C+Luscombe+N%2C+Bertone+P%2C+Echols+N%2C+Johnson+T%2C+Gerstein+M&amp;rft_id=http%3A%2F%2Fwww.genome.org%2Fcgi%2Fcontent%2Ffull%2F12%2F2%2F272&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-62"><span class="mw-cite-backlink"><a href="#cite_ref-62" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Harrison P, Gerstein M (2002). "Studying genomes through the aeons: protein families, pseudogenes and proteome evolution". <i>J Mol Biol</i>. <b>318</b> (5): 1155&amp;ndash, 74. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/12083509">12083509</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=J+Mol+Biol&amp;rft.atitle=Studying+genomes+through+the+aeons%3A+protein+families%2C+pseudogenes+and+proteome+evolution&amp;rft.volume=318&amp;rft.issue=5&amp;rft.pages=1155%26ndash%2C+74&amp;rft.date=2002&amp;rft_id=info%3Apmid%2F12083509&amp;rft.au=Harrison+P%2C+Gerstein+M&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-63"><span class="mw-cite-backlink"><a href="#cite_ref-63" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Albà M (2001). <a rel="nofollow" class="external text" href="http://www.pubmedcentral.nih.gov/articlerender.fcgi?tool=pubmed&amp;pubmedid=11178285">"Replicative DNA polymerases"</a>. <i>Genome Biol</i>. <b>2</b> (1): REVIEWS3002. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/11178285">11178285</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Genome+Biol&amp;rft.atitle=Replicative+DNA+polymerases&amp;rft.volume=2&amp;rft.issue=1&amp;rft.pages=REVIEWS3002&amp;rft.date=2001&amp;rft_id=info%3Apmid%2F11178285&amp;rft.au=Alb%C3%A0+M&amp;rft_id=http%3A%2F%2Fwww.pubmedcentral.nih.gov%2Farticlerender.fcgi%3Ftool%3Dpubmed%26pubmedid%3D11178285&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-64"><span class="mw-cite-backlink"><a href="#cite_ref-64" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Sandman K, Pereira S, Reeve J (1998). "Diversity of prokaryotic chromosomal proteins and the origin of the nucleosome". <i>Cell Mol Life Sci</i>. <b>54</b> (12): 1350&amp;ndash, 64. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/9893710">9893710</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Cell+Mol+Life+Sci&amp;rft.atitle=Diversity+of+prokaryotic+chromosomal+proteins+and+the+origin+of+the+nucleosome&amp;rft.volume=54&amp;rft.issue=12&amp;rft.pages=1350%26ndash%2C+64&amp;rft.date=1998&amp;rft_id=info%3Apmid%2F9893710&amp;rft.au=Sandman+K%2C+Pereira+S%2C+Reeve+J&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-65"><span class="mw-cite-backlink"><a href="#cite_ref-65" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Dame RT (2005). "The role of nucleoid-associated proteins in the organization and compaction of bacterial chromatin". <i>Mol. Microbiol</i>. <b>56</b> (4): 858–70. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/15853876">15853876</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Mol.+Microbiol.&amp;rft.atitle=The+role+of+nucleoid-associated+proteins+in+the+organization+and+compaction+of+bacterial+chromatin&amp;rft.volume=56&amp;rft.issue=4&amp;rft.pages=858-70&amp;rft.date=2005&amp;rft_id=info%3Apmid%2F15853876&amp;rft.au=Dame+RT&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-66"><span class="mw-cite-backlink"><a href="#cite_ref-66" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Luger K, Mäder A, Richmond R, Sargent D, Richmond T (1997). "Crystal structure of the nucleosome core particle at 2.8 A resolution". <i>Nature</i>. <b>389</b> (6648): 251&amp;ndash, 60. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1038%2F38444">10.1038/38444</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/9305837">9305837</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Nature&amp;rft.atitle=Crystal+structure+of+the+nucleosome+core+particle+at+2.8+A+resolution&amp;rft.volume=389&amp;rft.issue=6648&amp;rft.pages=251%26ndash%2C+60&amp;rft.date=1997&amp;rft_id=info%3Adoi%2F10.1038%2F38444&amp;rft_id=info%3Apmid%2F9305837&amp;rft.au=Luger+K%2C+M%C3%A4der+A%2C+Richmond+R%2C+Sargent+D%2C+Richmond+T&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-67"><span class="mw-cite-backlink"><a href="#cite_ref-67" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Jenuwein T, Allis C (2001). "Translating the histone code". <i>Science</i>. <b>293</b> (5532): 1074&amp;ndash, 80. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/11498575">11498575</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Science&amp;rft.atitle=Translating+the+histone+code&amp;rft.volume=293&amp;rft.issue=5532&amp;rft.pages=1074%26ndash%2C+80&amp;rft.date=2001&amp;rft_id=info%3Apmid%2F11498575&amp;rft.au=Jenuwein+T%2C+Allis+C&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-68"><span class="mw-cite-backlink"><a href="#cite_ref-68" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Ito T. "Nucleosome assembly and remodelling". <i>Curr Top Microbiol Immunol</i>. <b>274</b>: 1&amp;ndash, 22. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/12596902">12596902</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Curr+Top+Microbiol+Immunol&amp;rft.atitle=Nucleosome+assembly+and+remodelling&amp;rft.volume=274&amp;rft.pages=1%26ndash%2C+22&amp;rft_id=info%3Apmid%2F12596902&amp;rft.au=Ito+T&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-69"><span class="mw-cite-backlink"><a href="#cite_ref-69" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Thomas J (2001). "HMG1 and 2: architectural DNA-binding proteins". <i>Biochem Soc Trans</i>. <b>29</b> (Pt 4): 395&amp;ndash, 401. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/11497996">11497996</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Biochem+Soc+Trans&amp;rft.atitle=HMG1+and+2%3A+architectural+DNA-binding+proteins&amp;rft.volume=29&amp;rft.issue=Pt+4&amp;rft.pages=395%26ndash%2C+401&amp;rft.date=2001&amp;rft_id=info%3Apmid%2F11497996&amp;rft.au=Thomas+J&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-70"><span class="mw-cite-backlink"><a href="#cite_ref-70" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Grosschedl R, Giese K, Pagel J (1994). "HMG domain proteins: architectural elements in the assembly of nucleoprotein structures". <i>Trends Genet</i>. <b>10</b> (3): 94–100. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/8178371">8178371</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Trends+Genet&amp;rft.atitle=HMG+domain+proteins%3A+architectural+elements+in+the+assembly+of+nucleoprotein+structures&amp;rft.volume=10&amp;rft.issue=3&amp;rft.pages=94-100&amp;rft.date=1994&amp;rft_id=info%3Apmid%2F8178371&amp;rft.au=Grosschedl+R%2C+Giese+K%2C+Pagel+J&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-71"><span class="mw-cite-backlink"><a href="#cite_ref-71" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Iftode C, Daniely Y, Borowiec J (1999). "Replication protein A (RPA): the eukaryotic SSB". <i>Crit Rev Biochem Mol Biol</i>. <b>34</b> (3): 141&amp;ndash, 80. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/10473346">10473346</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Crit+Rev+Biochem+Mol+Biol&amp;rft.atitle=Replication+protein+A+%28RPA%29%3A+the+eukaryotic+SSB&amp;rft.volume=34&amp;rft.issue=3&amp;rft.pages=141%26ndash%2C+80&amp;rft.date=1999&amp;rft_id=info%3Apmid%2F10473346&amp;rft.au=Iftode+C%2C+Daniely+Y%2C+Borowiec+J&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-72"><span class="mw-cite-backlink"><a href="#cite_ref-72" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text">Created from <a rel="nofollow" class="external text" href="http://www.rcsb.org/pdb/explore/explore.do?structureId=1LMB">PDB 1LMB</a></span>
</li>
<li id="cite_note-73"><span class="mw-cite-backlink"><a href="#cite_ref-73" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Myers L, Kornberg R (2000). "Mediator of transcriptional regulation". <i>Annu Rev Biochem</i>. <b>69</b>: 729&amp;ndash, 49. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1146%2Fannurev.biochem.69.1.729">10.1146/annurev.biochem.69.1.729</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/10966474">10966474</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Annu+Rev+Biochem&amp;rft.atitle=Mediator+of+transcriptional+regulation&amp;rft.volume=69&amp;rft.pages=729%26ndash%2C+49&amp;rft.date=2000&amp;rft_id=info%3Adoi%2F10.1146%2Fannurev.biochem.69.1.729&amp;rft_id=info%3Apmid%2F10966474&amp;rft.au=Myers+L%2C+Kornberg+R&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-74"><span class="mw-cite-backlink"><a href="#cite_ref-74" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Spiegelman B, Heinrich R (2004). "Biological control through regulated transcriptional coactivators". <i>Cell</i>. <b>119</b> (2): 157–67. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/15479634">15479634</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Cell&amp;rft.atitle=Biological+control+through+regulated+transcriptional+coactivators&amp;rft.volume=119&amp;rft.issue=2&amp;rft.pages=157-67&amp;rft.date=2004&amp;rft_id=info%3Apmid%2F15479634&amp;rft.au=Spiegelman+B%2C+Heinrich+R&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-75"><span class="mw-cite-backlink"><a href="#cite_ref-75" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Li Z, Van Calcar S, Qu C, Cavenee W, Zhang M, Ren B (2003). <a rel="nofollow" class="external text" href="http://www.pubmedcentral.nih.gov/articlerender.fcgi?tool=pubmed&amp;pubmedid=12808131">"A global transcriptional regulatory role for c-Myc in Burkitt's lymphoma cells"</a>. <i>Proc Natl Acad Sci U S A</i>. <b>100</b> (14): 8164&amp;ndash, 9. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1073%2Fpnas.1332764100">10.1073/pnas.1332764100</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/12808131">12808131</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Proc+Natl+Acad+Sci+U+S+A&amp;rft.atitle=A+global+transcriptional+regulatory+role+for+c-Myc+in+Burkitt%27s+lymphoma+cells&amp;rft.volume=100&amp;rft.issue=14&amp;rft.pages=8164%26ndash%2C+9&amp;rft.date=2003&amp;rft_id=info%3Adoi%2F10.1073%2Fpnas.1332764100&amp;rft_id=info%3Apmid%2F12808131&amp;rft.au=Li+Z%2C+Van+Calcar+S%2C+Qu+C%2C+Cavenee+W%2C+Zhang+M%2C+Ren+B&amp;rft_id=http%3A%2F%2Fwww.pubmedcentral.nih.gov%2Farticlerender.fcgi%3Ftool%3Dpubmed%26pubmedid%3D12808131&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-76"><span class="mw-cite-backlink"><a href="#cite_ref-76" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Pabo C, Sauer R (1984). "Protein-DNA recognition". <i>Annu Rev Biochem</i>. <b>53</b>: 293&amp;ndash, 321. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1146%2Fannurev.bi.53.070184.001453">10.1146/annurev.bi.53.070184.001453</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/6236744">6236744</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Annu+Rev+Biochem&amp;rft.atitle=Protein-DNA+recognition&amp;rft.volume=53&amp;rft.pages=293%26ndash%2C+321&amp;rft.date=1984&amp;rft_id=info%3Adoi%2F10.1146%2Fannurev.bi.53.070184.001453&amp;rft_id=info%3Apmid%2F6236744&amp;rft.au=Pabo+C%2C+Sauer+R&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-77"><span class="mw-cite-backlink"><a href="#cite_ref-77" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text">Created from <a rel="nofollow" class="external text" href="http://www.rcsb.org/pdb/explore/explore.do?structureId=1RVA">PDB 1RVA</a></span>
</li>
<li id="cite_note-78"><span class="mw-cite-backlink"><a href="#cite_ref-78" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Bickle T, Krüger D (1993). <a rel="nofollow" class="external text" href="http://www.pubmedcentral.nih.gov/picrender.fcgi?artid=372918&amp;blobtype=pdf">"Biology of DNA restriction"</a>. <i>Microbiol Rev</i>. <b>57</b> (2): 434&amp;ndash, 50. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/8336674">8336674</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Microbiol+Rev&amp;rft.atitle=Biology+of+DNA+restriction&amp;rft.volume=57&amp;rft.issue=2&amp;rft.pages=434%26ndash%2C+50&amp;rft.date=1993&amp;rft_id=info%3Apmid%2F8336674&amp;rft.au=Bickle+T%2C+Kr%C3%BCger+D&amp;rft_id=http%3A%2F%2Fwww.pubmedcentral.nih.gov%2Fpicrender.fcgi%3Fartid%3D372918%26blobtype%3Dpdf&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-Doherty-79"><span class="mw-cite-backlink">↑ <sup><a href="#cite_ref-Doherty_79-0"><span class="cite-accessibility-label">Jump up to: </span>79.0</a></sup> <sup><a href="#cite_ref-Doherty_79-1">79.1</a></sup></span> <span class="reference-text"><cite class="citation journal">Doherty A, Suh S (2000). <a rel="nofollow" class="external text" href="http://www.pubmedcentral.nih.gov/articlerender.fcgi?tool=pubmed&amp;pubmedid=11058099">"Structural and mechanistic conservation in DNA ligases"</a>. <i>Nucleic Acids Res</i>. <b>28</b> (21): 4051&amp;ndash, 8. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/11058099">11058099</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Nucleic+Acids+Res&amp;rft.atitle=Structural+and+mechanistic+conservation+in+DNA+ligases.&amp;rft.volume=28&amp;rft.issue=21&amp;rft.pages=4051%26ndash%2C+8&amp;rft.date=2000&amp;rft_id=info%3Apmid%2F11058099&amp;rft.au=Doherty+A%2C+Suh+S&amp;rft_id=http%3A%2F%2Fwww.pubmedcentral.nih.gov%2Farticlerender.fcgi%3Ftool%3Dpubmed%26pubmedid%3D11058099&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-80"><span class="mw-cite-backlink"><a href="#cite_ref-80" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Schoeffler A, Berger J (2005). "Recent advances in understanding structure-function relationships in the type II topoisomerase mechanism". <i>Biochem Soc Trans</i>. <b>33</b> (Pt 6): 1465&amp;ndash, 70. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/16246147">16246147</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Biochem+Soc+Trans&amp;rft.atitle=Recent+advances+in+understanding+structure-function+relationships+in+the+type+II+topoisomerase+mechanism&amp;rft.volume=33&amp;rft.issue=Pt+6&amp;rft.pages=1465%26ndash%2C+70&amp;rft.date=2005&amp;rft_id=info%3Apmid%2F16246147&amp;rft.au=Schoeffler+A%2C+Berger+J&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-81"><span class="mw-cite-backlink"><a href="#cite_ref-81" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Tuteja N, Tuteja R (2004). "Unraveling DNA helicases. Motif, structure, mechanism and function". <i>Eur J Biochem</i>. <b>271</b> (10): 1849–63. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1111%2Fj.1432-1033.2004.04094.x">10.1111/j.1432-1033.2004.04094.x</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/15128295">15128295</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Eur+J+Biochem&amp;rft.atitle=Unraveling+DNA+helicases.+Motif%2C+structure%2C+mechanism+and+function&amp;rft.volume=271&amp;rft.issue=10&amp;rft.pages=1849-63&amp;rft.date=2004&amp;rft_id=info%3Adoi%2F10.1111%2Fj.1432-1033.2004.04094.x&amp;rft_id=info%3Apmid%2F15128295&amp;rft.au=Tuteja+N%2C+Tuteja+R&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-Joyce-82"><span class="mw-cite-backlink">↑ <sup><a href="#cite_ref-Joyce_82-0"><span class="cite-accessibility-label">Jump up to: </span>82.0</a></sup> <sup><a href="#cite_ref-Joyce_82-1">82.1</a></sup></span> <span class="reference-text"><cite class="citation journal">Joyce C, Steitz T (1995). <a rel="nofollow" class="external text" href="http://www.pubmedcentral.nih.gov/picrender.fcgi?artid=177480&amp;blobtype=pdf">"Polymerase structures and function: variations on a theme?"</a>. <i>J Bacteriol</i>. <b>177</b> (22): 6321&amp;ndash, 9. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/7592405">7592405</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=J+Bacteriol&amp;rft.atitle=Polymerase+structures+and+function%3A+variations+on+a+theme%3F&amp;rft.volume=177&amp;rft.issue=22&amp;rft.pages=6321%26ndash%2C+9&amp;rft.date=1995&amp;rft_id=info%3Apmid%2F7592405&amp;rft.au=Joyce+C%2C+Steitz+T&amp;rft_id=http%3A%2F%2Fwww.pubmedcentral.nih.gov%2Fpicrender.fcgi%3Fartid%3D177480%26blobtype%3Dpdf&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"> <span class="error mw-ext-cite-error" lang="en" dir="ltr"></span></span>
</li>
<li id="cite_note-83"><span class="mw-cite-backlink"><a href="#cite_ref-83" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Hubscher U, Maga G, Spadari S (2002). "Eukaryotic DNA polymerases". <i>Annu Rev Biochem</i>. <b>71</b>: 133&amp;ndash, 63. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1146%2Fannurev.biochem.71.090501.150041">10.1146/annurev.biochem.71.090501.150041</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/12045093">12045093</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Annu+Rev+Biochem&amp;rft.atitle=Eukaryotic+DNA+polymerases&amp;rft.volume=71&amp;rft.pages=133%26ndash%2C+63&amp;rft.date=2002&amp;rft_id=info%3Adoi%2F10.1146%2Fannurev.biochem.71.090501.150041&amp;rft_id=info%3Apmid%2F12045093&amp;rft.au=Hubscher+U%2C+Maga+G%2C+Spadari+S&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-84"><span class="mw-cite-backlink"><a href="#cite_ref-84" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Johnson A, O'Donnell M (2005). "Cellular DNA replicases: components and dynamics at the replication fork". <i>Annu Rev Biochem</i>. <b>74</b>: 283&amp;ndash, 315. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1146%2Fannurev.biochem.73.011303.073859">10.1146/annurev.biochem.73.011303.073859</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/15952889">15952889</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Annu+Rev+Biochem&amp;rft.atitle=Cellular+DNA+replicases%3A+components+and+dynamics+at+the+replication+fork&amp;rft.volume=74&amp;rft.pages=283%26ndash%2C+315&amp;rft.date=2005&amp;rft_id=info%3Adoi%2F10.1146%2Fannurev.biochem.73.011303.073859&amp;rft_id=info%3Apmid%2F15952889&amp;rft.au=Johnson+A%2C+O%27Donnell+M&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-85"><span class="mw-cite-backlink"><a href="#cite_ref-85" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Tarrago-Litvak L, Andréola M, Nevinsky G, Sarih-Cottin L, Litvak S (1994). <a rel="nofollow" class="external text" href="http://www.fasebj.org/cgi/reprint/8/8/497">"The reverse transcriptase of HIV-1: from enzymology to therapeutic intervention"</a>. <i>FASEB J</i>. <b>8</b> (8): 497–503. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/7514143">7514143</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=FASEB+J&amp;rft.atitle=The+reverse+transcriptase+of+HIV-1%3A+from+enzymology+to+therapeutic+intervention&amp;rft.volume=8&amp;rft.issue=8&amp;rft.pages=497-503&amp;rft.date=1994&amp;rft_id=info%3Apmid%2F7514143&amp;rft.au=Tarrago-Litvak+L%2C+Andr%C3%A9ola+M%2C+Nevinsky+G%2C+Sarih-Cottin+L%2C+Litvak+S&amp;rft_id=http%3A%2F%2Fwww.fasebj.org%2Fcgi%2Freprint%2F8%2F8%2F497&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-86"><span class="mw-cite-backlink"><a href="#cite_ref-86" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Martinez E (2002). "Multi-protein complexes in eukaryotic gene transcription". <i>Plant Mol Biol</i>. <b>50</b> (6): 925&amp;ndash, 47. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/12516863">12516863</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Plant+Mol+Biol&amp;rft.atitle=Multi-protein+complexes+in+eukaryotic+gene+transcription&amp;rft.volume=50&amp;rft.issue=6&amp;rft.pages=925%26ndash%2C+47&amp;rft.date=2002&amp;rft_id=info%3Apmid%2F12516863&amp;rft.au=Martinez+E&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-87"><span class="mw-cite-backlink"><a href="#cite_ref-87" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text">Created from <a rel="nofollow" class="external text" href="http://www.rcsb.org/pdb/explore/explore.do?structureId=1M6G">PDB 1M6G</a></span>
</li>
<li id="cite_note-88"><span class="mw-cite-backlink"><a href="#cite_ref-88" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Cremer T, Cremer C (2001). "Chromosome territories, nuclear architecture and gene regulation in mammalian cells". <i>Nat Rev Genet</i>. <b>2</b> (4): 292–301. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1038%2F35066075">10.1038/35066075</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/11283701">11283701</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Nat+Rev+Genet&amp;rft.atitle=Chromosome+territories%2C+nuclear+architecture+and+gene+regulation+in+mammalian+cells&amp;rft.volume=2&amp;rft.issue=4&amp;rft.pages=292-301&amp;rft.date=2001&amp;rft_id=info%3Adoi%2F10.1038%2F35066075&amp;rft_id=info%3Apmid%2F11283701&amp;rft.au=Cremer+T%2C+Cremer+C&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-89"><span class="mw-cite-backlink"><a href="#cite_ref-89" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Pál C, Papp B, Lercher M (2006). "An integrated view of protein evolution". <i>Nat Rev Genet</i>. <b>7</b> (5): 337&amp;ndash, 48. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1038%2Fnrg1838">10.1038/nrg1838</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/16619049">16619049</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Nat+Rev+Genet&amp;rft.atitle=An+integrated+view+of+protein+evolution&amp;rft.volume=7&amp;rft.issue=5&amp;rft.pages=337%26ndash%2C+48&amp;rft.date=2006&amp;rft_id=info%3Adoi%2F10.1038%2Fnrg1838&amp;rft_id=info%3Apmid%2F16619049&amp;rft.au=P%C3%A1l+C%2C+Papp+B%2C+Lercher+M&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-90"><span class="mw-cite-backlink"><a href="#cite_ref-90" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">O'Driscoll M, Jeggo P (2006). "The role of double-strand break repair - insights from human genetics". <i>Nat Rev Genet</i>. <b>7</b> (1): 45&amp;ndash, 54. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1038%2Fnrg1746">10.1038/nrg1746</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/16369571">16369571</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Nat+Rev+Genet&amp;rft.atitle=The+role+of+double-strand+break+repair+-+insights+from+human+genetics&amp;rft.volume=7&amp;rft.issue=1&amp;rft.pages=45%26ndash%2C+54&amp;rft.date=2006&amp;rft_id=info%3Adoi%2F10.1038%2Fnrg1746&amp;rft_id=info%3Apmid%2F16369571&amp;rft.au=O%27Driscoll+M%2C+Jeggo+P&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-91"><span class="mw-cite-backlink"><a href="#cite_ref-91" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Vispé S, Defais M (1997). "Mammalian Rad51 protein: a RecA homologue with pleiotropic functions". <i>Biochimie</i>. <b>79</b> (9–10): 587–92. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/9466696">9466696</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Biochimie&amp;rft.atitle=Mammalian+Rad51+protein%3A+a+RecA+homologue+with+pleiotropic+functions&amp;rft.volume=79&amp;rft.issue=9%E2%80%9310&amp;rft.pages=587-92&amp;rft.date=1997&amp;rft_id=info%3Apmid%2F9466696&amp;rft.au=Visp%C3%A9+S%2C+Defais+M&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-92"><span class="mw-cite-backlink"><a href="#cite_ref-92" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Neale MJ, Keeney S (2006). "Clarifying the mechanics of DNA strand exchange in meiotic recombination". <i>Nature</i>. <b>442</b> (7099): 153–8. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1038%2Fnature04885">10.1038/nature04885</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/16838012">16838012</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Nature&amp;rft.atitle=Clarifying+the+mechanics+of+DNA+strand+exchange+in+meiotic+recombination&amp;rft.volume=442&amp;rft.issue=7099&amp;rft.pages=153-8&amp;rft.date=2006&amp;rft_id=info%3Adoi%2F10.1038%2Fnature04885&amp;rft_id=info%3Apmid%2F16838012&amp;rft.au=Neale+MJ%2C+Keeney+S&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-93"><span class="mw-cite-backlink"><a href="#cite_ref-93" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Dickman M, Ingleston S, Sedelnikova S, Rafferty J, Lloyd R, Grasby J, Hornby D (2002). "The RuvABC resolvasome". <i>Eur J Biochem</i>. <b>269</b> (22): 5492&amp;ndash, 501. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/12423347">12423347</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Eur+J+Biochem&amp;rft.atitle=The+RuvABC+resolvasome&amp;rft.volume=269&amp;rft.issue=22&amp;rft.pages=5492%26ndash%2C+501&amp;rft.date=2002&amp;rft_id=info%3Apmid%2F12423347&amp;rft.au=Dickman+M%2C+Ingleston+S%2C+Sedelnikova+S%2C+Rafferty+J%2C+Lloyd+R%2C+Grasby+J%2C+Hornby+D&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-94"><span class="mw-cite-backlink"><a href="#cite_ref-94" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Orgel L. <a rel="nofollow" class="external text" href="http://www.crbmb.com/cgi/reprint/39/2/99.pdf">"Prebiotic chemistry and the origin of the RNA world"</a> <span class="cs1-format">(PDF)</span>. <i>Crit Rev Biochem Mol Biol</i>. <b>39</b> (2): 99&amp;ndash, 123. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1080%2F10409230490460765">10.1080/10409230490460765</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/15217990">15217990</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Crit+Rev+Biochem+Mol+Biol&amp;rft.atitle=Prebiotic+chemistry+and+the+origin+of+the+RNA+world&amp;rft.volume=39&amp;rft.issue=2&amp;rft.pages=99%26ndash%2C+123&amp;rft_id=info%3Adoi%2F10.1080%2F10409230490460765&amp;rft_id=info%3Apmid%2F15217990&amp;rft.au=Orgel+L&amp;rft_id=http%3A%2F%2Fwww.crbmb.com%2Fcgi%2Freprint%2F39%2F2%2F99.pdf&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-95"><span class="mw-cite-backlink"><a href="#cite_ref-95" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Davenport R (2001). "Ribozymes. Making copies in the RNA world". <i>Science</i>. <b>292</b> (5520): 1278. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1126%2Fscience.292.5520.1278a">10.1126/science.292.5520.1278a</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/11360970">11360970</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Science&amp;rft.atitle=Ribozymes.+Making+copies+in+the+RNA+world&amp;rft.volume=292&amp;rft.issue=5520&amp;rft.pages=1278&amp;rft.date=2001&amp;rft_id=info%3Adoi%2F10.1126%2Fscience.292.5520.1278a&amp;rft_id=info%3Apmid%2F11360970&amp;rft.au=Davenport+R&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-96"><span class="mw-cite-backlink"><a href="#cite_ref-96" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Szathmáry E (1992). <a rel="nofollow" class="external text" href="http://www.pnas.org/cgi/reprint/89/7/2614.pdf">"What is the optimum size for the genetic alphabet?"</a> <span class="cs1-format">(PDF)</span>. <i>Proc Natl Acad Sci U S A</i>. <b>89</b> (7): 2614&amp;ndash, 8. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/1372984">1372984</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Proc+Natl+Acad+Sci+U+S+A&amp;rft.atitle=What+is+the+optimum+size+for+the+genetic+alphabet%3F&amp;rft.volume=89&amp;rft.issue=7&amp;rft.pages=2614%26ndash%2C+8&amp;rft.date=1992&amp;rft_id=info%3Apmid%2F1372984&amp;rft.au=Szathm%C3%A1ry+E&amp;rft_id=http%3A%2F%2Fwww.pnas.org%2Fcgi%2Freprint%2F89%2F7%2F2614.pdf&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-97"><span class="mw-cite-backlink"><a href="#cite_ref-97" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Lindahl T (1993). "Instability and decay of the primary structure of DNA". <i>Nature</i>. <b>362</b> (6422): 709&amp;ndash, 15. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1038%2F362709a0">10.1038/362709a0</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/8469282">8469282</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Nature&amp;rft.atitle=Instability+and+decay+of+the+primary+structure+of+DNA&amp;rft.volume=362&amp;rft.issue=6422&amp;rft.pages=709%26ndash%2C+15&amp;rft.date=1993&amp;rft_id=info%3Adoi%2F10.1038%2F362709a0&amp;rft_id=info%3Apmid%2F8469282&amp;rft.au=Lindahl+T&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-98"><span class="mw-cite-backlink"><a href="#cite_ref-98" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Vreeland R, Rosenzweig W, Powers D (2000). "Isolation of a 250 million-year-old halotolerant bacterium from a primary salt crystal". <i>Nature</i>. <b>407</b> (6806): 897&amp;ndash, 900. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1038%2F35038060">10.1038/35038060</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/11057666">11057666</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Nature&amp;rft.atitle=Isolation+of+a+250+million-year-old+halotolerant+bacterium+from+a+primary+salt+crystal&amp;rft.volume=407&amp;rft.issue=6806&amp;rft.pages=897%26ndash%2C+900&amp;rft.date=2000&amp;rft_id=info%3Adoi%2F10.1038%2F35038060&amp;rft_id=info%3Apmid%2F11057666&amp;rft.au=Vreeland+R%2C+Rosenzweig+W%2C+Powers+D&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-99"><span class="mw-cite-backlink"><a href="#cite_ref-99" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Hebsgaard M, Phillips M, Willerslev E (2005). "Geologically ancient DNA: fact or artefact?". <i>Trends Microbiol</i>. <b>13</b> (5): 212&amp;ndash, 20. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/15866038">15866038</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Trends+Microbiol&amp;rft.atitle=Geologically+ancient+DNA%3A+fact+or+artefact%3F&amp;rft.volume=13&amp;rft.issue=5&amp;rft.pages=212%26ndash%2C+20&amp;rft.date=2005&amp;rft_id=info%3Apmid%2F15866038&amp;rft.au=Hebsgaard+M%2C+Phillips+M%2C+Willerslev+E&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-100"><span class="mw-cite-backlink"><a href="#cite_ref-100" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Nickle D, Learn G, Rain M, Mullins J, Mittler J (2002). "Curiously modern DNA for a "250 million-year-old" bacterium". <i>J Mol Evol</i>. <b>54</b> (1): 134&amp;ndash, 7. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1007%2Fs00239-001-0025-x">10.1007/s00239-001-0025-x</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/11734907">11734907</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=J+Mol+Evol&amp;rft.atitle=Curiously+modern+DNA+for+a+%22250+million-year-old%22+bacterium&amp;rft.volume=54&amp;rft.issue=1&amp;rft.pages=134%26ndash%2C+7&amp;rft.date=2002&amp;rft_id=info%3Adoi%2F10.1007%2Fs00239-001-0025-x&amp;rft_id=info%3Apmid%2F11734907&amp;rft.au=Nickle+D%2C+Learn+G%2C+Rain+M%2C+Mullins+J%2C+Mittler+J&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-101"><span class="mw-cite-backlink"><a href="#cite_ref-101" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Goff SP, Berg P (1976). "Construction of hybrid viruses containing SV40 and lambda phage DNA segments and their propagation in cultured monkey cells". <i>Cell</i>. <b>9</b> (4 PT 2): 695–705. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/189942">189942</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Cell&amp;rft.atitle=Construction+of+hybrid+viruses+containing+SV40+and+lambda+phage+DNA+segments+and+their+propagation+in+cultured+monkey+cells&amp;rft.volume=9&amp;rft.issue=4+PT+2&amp;rft.pages=695-705&amp;rft.date=1976&amp;rft_id=info%3Apmid%2F189942&amp;rft.au=Goff+SP%2C+Berg+P&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-102"><span class="mw-cite-backlink"><a href="#cite_ref-102" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Houdebine L. "Transgenic animal models in biomedical research". <i>Methods Mol Biol</i>. <b>360</b>: 163&amp;ndash, 202. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/17172731">17172731</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Methods+Mol+Biol&amp;rft.atitle=Transgenic+animal+models+in+biomedical+research&amp;rft.volume=360&amp;rft.pages=163%26ndash%2C+202&amp;rft_id=info%3Apmid%2F17172731&amp;rft.au=Houdebine+L&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-103"><span class="mw-cite-backlink"><a href="#cite_ref-103" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Daniell H, Dhingra A (2002). "Multigene engineering: dawn of an exciting new era in biotechnology". <i>Curr Opin Biotechnol</i>. <b>13</b> (2): 136&amp;ndash, 41. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/11950565">11950565</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Curr+Opin+Biotechnol&amp;rft.atitle=Multigene+engineering%3A+dawn+of+an+exciting+new+era+in+biotechnology&amp;rft.volume=13&amp;rft.issue=2&amp;rft.pages=136%26ndash%2C+41&amp;rft.date=2002&amp;rft_id=info%3Apmid%2F11950565&amp;rft.au=Daniell+H%2C+Dhingra+A&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-104"><span class="mw-cite-backlink"><a href="#cite_ref-104" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Job D (2002). "Plant biotechnology in agriculture". <i>Biochimie</i>. <b>84</b> (11): 1105&amp;ndash, 10. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/12595138">12595138</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Biochimie&amp;rft.atitle=Plant+biotechnology+in+agriculture&amp;rft.volume=84&amp;rft.issue=11&amp;rft.pages=1105%26ndash%2C+10&amp;rft.date=2002&amp;rft_id=info%3Apmid%2F12595138&amp;rft.au=Job+D&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-105"><span class="mw-cite-backlink"><a href="#cite_ref-105" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Collins A, Morton N (1994). <a rel="nofollow" class="external text" href="http://www.pnas.org/cgi/reprint/91/13/6007.pdf">"Likelihood ratios for DNA identification"</a> <span class="cs1-format">(PDF)</span>. <i>Proc Natl Acad Sci U S A</i>. <b>91</b> (13): 6007&amp;ndash, 11. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/8016106">8016106</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Proc+Natl+Acad+Sci+U+S+A&amp;rft.atitle=Likelihood+ratios+for+DNA+identification&amp;rft.volume=91&amp;rft.issue=13&amp;rft.pages=6007%26ndash%2C+11&amp;rft.date=1994&amp;rft_id=info%3Apmid%2F8016106&amp;rft.au=Collins+A%2C+Morton+N&amp;rft_id=http%3A%2F%2Fwww.pnas.org%2Fcgi%2Freprint%2F91%2F13%2F6007.pdf&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-106"><span class="mw-cite-backlink"><a href="#cite_ref-106" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Weir B, Triggs C, Starling L, Stowell L, Walsh K, Buckleton J (1997). "Interpreting DNA mixtures". <i>J Forensic Sci</i>. <b>42</b> (2): 213&amp;ndash, 22. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/9068179">9068179</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=J+Forensic+Sci&amp;rft.atitle=Interpreting+DNA+mixtures&amp;rft.volume=42&amp;rft.issue=2&amp;rft.pages=213%26ndash%2C+22&amp;rft.date=1997&amp;rft_id=info%3Apmid%2F9068179&amp;rft.au=Weir+B%2C+Triggs+C%2C+Starling+L%2C+Stowell+L%2C+Walsh+K%2C+Buckleton+J&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-107"><span class="mw-cite-backlink"><a href="#cite_ref-107" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Jeffreys A, Wilson V, Thein S. "Individual-specific 'fingerprints' of human DNA". <i>Nature</i>. <b>316</b> (6023): 76&amp;ndash, 9. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/2989708">2989708</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Nature&amp;rft.atitle=Individual-specific+%27fingerprints%27+of+human+DNA.&amp;rft.volume=316&amp;rft.issue=6023&amp;rft.pages=76%26ndash%2C+9&amp;rft_id=info%3Apmid%2F2989708&amp;rft.au=Jeffreys+A%2C+Wilson+V%2C+Thein+S&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-108"><span class="mw-cite-backlink"><a href="#cite_ref-108" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><a rel="nofollow" class="external text" href="http://www.forensic.gov.uk/forensic_t/inside/news/list_casefiles.php?case=1">Colin Pitchfork — first murder conviction on DNA evidence also clears the prime suspect</a> Forensic Science Service Accessed 23 Dec 2006</span>
</li>
<li id="cite_note-109"><span class="mw-cite-backlink"><a href="#cite_ref-109" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="http://massfatality.dna.gov/Introduction/">"DNA Identification in Mass Fatality Incidents"</a>. National Institute of Justice. September 2006.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=DNA+Identification+in+Mass+Fatality+Incidents&amp;rft.pub=National+Institute+of+Justice&amp;rft.date=2006-09&amp;rft_id=http%3A%2F%2Fmassfatality.dna.gov%2FIntroduction%2F&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-110"><span class="mw-cite-backlink"><a href="#cite_ref-110" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation book">Baldi, Pierre. Brunak, Soren (2001). <i>Bioinformatics: The Machine Learning Approach</i>. MIT Press. <a href="/index.php?title=International_Standard_Book_Number&amp;action=edit&amp;redlink=1" class="new" title="International Standard Book Number (page does not exist)">ISBN</a>&nbsp;<a href="/index.php/Special:BookSources/978-0-262-02506-5" title="Special:BookSources/978-0-262-02506-5">978-0-262-02506-5</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=book&amp;rft.btitle=Bioinformatics%3A+The+Machine+Learning+Approach&amp;rft.pub=MIT+Press&amp;rft.date=2001&amp;rft.isbn=978-0-262-02506-5&amp;rft.au=Baldi%2C+Pierre.+Brunak%2C+Soren&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-111"><span class="mw-cite-backlink"><a href="#cite_ref-111" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text">Gusfield, Dan. <i>Algorithms on Strings, Trees, and Sequences: Computer Science and Computational Biology</i>. Cambridge University Press, 15 January 1997. ISBN 978-0-521-58519-4.</span>
</li>
<li id="cite_note-112"><span class="mw-cite-backlink"><a href="#cite_ref-112" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Sjölander K (2004). <a rel="nofollow" class="external text" href="http://bioinformatics.oxfordjournals.org/cgi/reprint/20/2/170">"Phylogenomic inference of protein molecular function: advances and challenges"</a>. <i>Bioinformatics</i>. <b>20</b> (2): 170–9. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1093%2Fbioinformatics%2Fbth021">10.1093/bioinformatics/bth021</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/14734307">14734307</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Bioinformatics&amp;rft.atitle=Phylogenomic+inference+of+protein+molecular+function%3A+advances+and+challenges&amp;rft.volume=20&amp;rft.issue=2&amp;rft.pages=170-9&amp;rft.date=2004&amp;rft_id=info%3Adoi%2F10.1093%2Fbioinformatics%2Fbth021&amp;rft_id=info%3Apmid%2F14734307&amp;rft.au=Sj%C3%B6lander+K&amp;rft_id=http%3A%2F%2Fbioinformatics.oxfordjournals.org%2Fcgi%2Freprint%2F20%2F2%2F170&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-Mount-113"><span class="mw-cite-backlink"><a href="#cite_ref-Mount_113-0" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation book">Mount DM (2004). <i>Bioinformatics: Sequence and Genome Analysis</i> (2 ed.). Cold Spring Harbor Laboratory Press. <a href="/index.php?title=International_Standard_Book_Number&amp;action=edit&amp;redlink=1" class="new" title="International Standard Book Number (page does not exist)">ISBN</a>&nbsp;<a href="/index.php/Special:BookSources/0879697121" title="Special:BookSources/0879697121">0879697121</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=book&amp;rft.btitle=Bioinformatics%3A+Sequence+and+Genome+Analysis&amp;rft.edition=2&amp;rft.pub=Cold+Spring+Harbor+Laboratory+Press&amp;rft.date=2004&amp;rft.isbn=0879697121&amp;rft.au=Mount+DM&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span> <span class="cs1-visible-error error citation-comment">Text " Cold Spring Harbor, NY " ignored (<a href="/index.php?title=Help:CS1_errors&amp;action=edit&amp;redlink=1" class="new" title="Help:CS1 errors (page does not exist)">help</a>); </span><span class="cs1-visible-error error citation-comment">Text " location " ignored (<a href="/index.php?title=Help:CS1_errors&amp;action=edit&amp;redlink=1" class="new" title="Help:CS1 errors (page does not exist)">help</a>)</span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-114"><span class="mw-cite-backlink"><a href="#cite_ref-114" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Wray G (2002). <a rel="nofollow" class="external text" href="http://www.pubmedcentral.nih.gov/articlerender.fcgi?tool=pubmed&amp;pubmedid=11806830">"Dating branches on the tree of life using DNA"</a>. <i>Genome Biol</i>. <b>3</b> (1): REVIEWS0001. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1046%2Fj.1525-142X.1999.99010.x">10.1046/j.1525-142X.1999.99010.x</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/11806830">11806830</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Genome+Biol&amp;rft.atitle=Dating+branches+on+the+tree+of+life+using+DNA&amp;rft.volume=3&amp;rft.issue=1&amp;rft.pages=REVIEWS0001&amp;rft.date=2002&amp;rft_id=info%3Adoi%2F10.1046%2Fj.1525-142X.1999.99010.x&amp;rft_id=info%3Apmid%2F11806830&amp;rft.au=Wray+G&amp;rft_id=http%3A%2F%2Fwww.pubmedcentral.nih.gov%2Farticlerender.fcgi%3Ftool%3Dpubmed%26pubmedid%3D11806830&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-115"><span class="mw-cite-backlink"><a href="#cite_ref-115" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><i>Lost Tribes of Israel</i>, <a href="/index.php?title=Nova_(TV_series)&amp;action=edit&amp;redlink=1" class="new" title="Nova (TV series) (page does not exist)">NOVA</a>, PBS airdate: 22 February 2000. Transcript available from <a rel="nofollow" class="external text" href="http://www.pbs.org/wgbh/nova/transcripts/2706israel.html">PBS.org,</a> (last accessed on 4 March 2006)</span>
</li>
<li id="cite_note-116"><span class="mw-cite-backlink"><a href="#cite_ref-116" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text">Kleiman, Yaakov. <a rel="nofollow" class="external text" href="http://www.aish.com/societywork/sciencenature/the_cohanim_-_dna_connection.asp">"The Cohanim/DNA Connection: The fascinating story of how DNA studies confirm an ancient biblical tradition".</a> <i>aish.com</i> (January 13, 2000). Accessed 4 March 2006.</span>
</li>
<li id="cite_note-117"><span class="mw-cite-backlink"><a href="#cite_ref-117" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text">Bhattacharya, Shaoni. <a rel="nofollow" class="external text" href="http://www.newscientist.com/article.ns?id=dn4908">"Killer convicted thanks to relative's DNA".</a> <i>newscientist.com</i> (20 April 2004). Accessed 22 Dec 06</span>
</li>
<li id="cite_note-118"><span class="mw-cite-backlink"><a href="#cite_ref-118" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Dahm R (2005). "Friedrich Miescher and the discovery of DNA". <i>Dev Biol</i>. <b>278</b> (2): 274&amp;ndash, 88. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/15680349">15680349</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Dev+Biol&amp;rft.atitle=Friedrich+Miescher+and+the+discovery+of+DNA&amp;rft.volume=278&amp;rft.issue=2&amp;rft.pages=274%26ndash%2C+88&amp;rft.date=2005&amp;rft_id=info%3Apmid%2F15680349&amp;rft.au=Dahm+R&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-119"><span class="mw-cite-backlink"><a href="#cite_ref-119" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Levene P, (1919). <a rel="nofollow" class="external text" href="http://www.jbc.org/cgi/reprint/40/2/415">"The structure of yeast nucleic acid"</a>. <i>J Biol Chem</i>. <b>40</b> (2): 415&amp;ndash, 24.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=J+Biol+Chem&amp;rft.atitle=The+structure+of+yeast+nucleic+acid&amp;rft.volume=40&amp;rft.issue=2&amp;rft.pages=415%26ndash%2C+24&amp;rft.date=1919&amp;rft.au=Levene+P%2C&amp;rft_id=http%3A%2F%2Fwww.jbc.org%2Fcgi%2Freprint%2F40%2F2%2F415&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-120"><span class="mw-cite-backlink"><a href="#cite_ref-120" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Astbury W, (1947). "Nucleic acid". <i>Symp. SOC. Exp. Bbl</i>. <b>1</b> (66).</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Symp.+SOC.+Exp.+Bbl&amp;rft.atitle=Nucleic+acid&amp;rft.volume=1&amp;rft.issue=66&amp;rft.date=1947&amp;rft.au=Astbury+W%2C&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-121"><span class="mw-cite-backlink"><a href="#cite_ref-121" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Lorenz MG, Wackernagel W (1994). <a rel="nofollow" class="external text" href="http://www.pubmedcentral.nih.gov/articlerender.fcgi?tool=pubmed&amp;pubmedid=7968924">"Bacterial gene transfer by natural genetic transformation in the environment"</a>. <i>Microbiol. Rev</i>. <b>58</b> (3): 563–602. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/7968924">7968924</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Microbiol.+Rev.&amp;rft.atitle=Bacterial+gene+transfer+by+natural+genetic+transformation+in+the+environment&amp;rft.volume=58&amp;rft.issue=3&amp;rft.pages=563-602&amp;rft.date=1994&amp;rft_id=info%3Apmid%2F7968924&amp;rft.au=Lorenz+MG%2C+Wackernagel+W&amp;rft_id=http%3A%2F%2Fwww.pubmedcentral.nih.gov%2Farticlerender.fcgi%3Ftool%3Dpubmed%26pubmedid%3D7968924&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-122"><span class="mw-cite-backlink"><a href="#cite_ref-122" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Avery O, MacLeod C, McCarty M (1944). <a rel="nofollow" class="external text" href="http://www.jem.org/cgi/reprint/149/2/297">"Studies on the chemical nature of the substance inducing transformation of pneumococcal types. Inductions of transformation by a desoxyribonucleic acid fraction isolated from pneumococcus type III"</a>. <i>J Exp Med</i>. <b>79</b> (2): 137&amp;ndash, 158.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=J+Exp+Med&amp;rft.atitle=Studies+on+the+chemical+nature+of+the+substance+inducing+transformation+of+pneumococcal+types.+Inductions+of+transformation+by+a+desoxyribonucleic+acid+fraction+isolated+from+pneumococcus+type+III&amp;rft.volume=79&amp;rft.issue=2&amp;rft.pages=137%26ndash%2C+158&amp;rft.date=1944&amp;rft.au=Avery+O%2C+MacLeod+C%2C+McCarty+M&amp;rft_id=http%3A%2F%2Fwww.jem.org%2Fcgi%2Freprint%2F149%2F2%2F297&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-123"><span class="mw-cite-backlink"><a href="#cite_ref-123" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Hershey A, Chase M (1952). <a rel="nofollow" class="external text" href="http://www.jgp.org/cgi/reprint/36/1/39.pdf">"Independent functions of viral protein and nucleic acid in growth of bacteriophage"</a> <span class="cs1-format">(PDF)</span>. <i>J Gen Physiol</i>. <b>36</b> (1): 39&amp;ndash, 56. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/12981234">12981234</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=J+Gen+Physiol&amp;rft.atitle=Independent+functions+of+viral+protein+and+nucleic+acid+in+growth+of+bacteriophage&amp;rft.volume=36&amp;rft.issue=1&amp;rft.pages=39%26ndash%2C+56&amp;rft.date=1952&amp;rft_id=info%3Apmid%2F12981234&amp;rft.au=Hershey+A%2C+Chase+M&amp;rft_id=http%3A%2F%2Fwww.jgp.org%2Fcgi%2Freprint%2F36%2F1%2F39.pdf&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-FWPUB-124"><span class="mw-cite-backlink">↑ <sup><a href="#cite_ref-FWPUB_124-0"><span class="cite-accessibility-label">Jump up to: </span>124.0</a></sup> <sup><a href="#cite_ref-FWPUB_124-1">124.1</a></sup></span> <span class="reference-text"><cite class="citation journal">Watson J.D. and Crick F.H.C. (1953). <a rel="nofollow" class="external text" href="http://www.nature.com/nature/dna50/watsoncrick.pdf">"A Structure for Deoxyribose Nucleic Acid"</a> <span class="cs1-format">(PDF)</span>. <i>Nature</i>. <b>171</b>: 737&amp;ndash, 738. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1038%2F171737a0">10.1038/171737a0</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/13054692">13054692</a><span class="reference-accessdate">. Retrieved <span class="nowrap">13 Feb</span> 2007</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Nature&amp;rft.atitle=A+Structure+for+Deoxyribose+Nucleic+Acid&amp;rft.volume=171&amp;rft.pages=737%26ndash%2C+738&amp;rft.date=1953&amp;rft_id=info%3Adoi%2F10.1038%2F171737a0&amp;rft_id=info%3Apmid%2F13054692&amp;rft.au=Watson+J.D.+and+Crick+F.H.C.&amp;rft_id=http%3A%2F%2Fwww.nature.com%2Fnature%2Fdna50%2Fwatsoncrick.pdf&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"> </span>
</li>
<li id="cite_note-NatureDNA50-125"><span class="mw-cite-backlink"><a href="#cite_ref-NatureDNA50_125-0" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text">Nature Archives <a rel="nofollow" class="external text" href="http://www.nature.com/nature/dna50/archive.html">Double Helix of DNA: 50 Years</a></span>
</li>
<li id="cite_note-NatFranGos-126"><span class="mw-cite-backlink"><a href="#cite_ref-NatFranGos_126-0" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal"><a rel="nofollow" class="external text" href="http://www.nature.com/nature/dna50/franklingosling.pdf">"Molecular Configuration in Sodium Thymonucleate. Franklin R. and Gosling R.G."</a> <span class="cs1-format">(PDF)</span>. <i>Nature</i>. <b>171</b>: 740&amp;ndash, 741. 1953. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1038%2F171740a0">10.1038/171740a0</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/13054694">13054694</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Nature&amp;rft.atitle=Molecular+Configuration+in+Sodium+Thymonucleate.+Franklin+R.+and+Gosling+R.G.&amp;rft.volume=171&amp;rft.pages=740%26ndash%2C+741&amp;rft.date=1953&amp;rft_id=info%3Adoi%2F10.1038%2F171740a0&amp;rft_id=info%3Apmid%2F13054694&amp;rft_id=http%3A%2F%2Fwww.nature.com%2Fnature%2Fdna50%2Ffranklingosling.pdf&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-127"><span class="mw-cite-backlink"><a href="#cite_ref-127" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><a rel="nofollow" class="external text" href="http://osulibrary.oregonstate.edu/specialcollections/coll/pauling/dna/pictures/franklin-typeBphoto.html">Original X-ray diffraction image</a></span>
</li>
<li id="cite_note-NatWilk-128"><span class="mw-cite-backlink"><a href="#cite_ref-NatWilk_128-0" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Wilkins M.H.F., A.R. Stokes A.R. &amp; Wilson, H.R. (1953). <a rel="nofollow" class="external text" href="http://www.nature.com/nature/dna50/wilkins.pdf">"Molecular Structure of Deoxypentose Nucleic Acids"</a> <span class="cs1-format">(PDF)</span>. <i>Nature</i>. <b>171</b>: 738&amp;ndash, 740. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1038%2F171738a0">10.1038/171738a0</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/13054693">13054693</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Nature&amp;rft.atitle=Molecular+Structure+of+Deoxypentose+Nucleic+Acids&amp;rft.volume=171&amp;rft.pages=738%26ndash%2C+740&amp;rft.date=1953&amp;rft_id=info%3Adoi%2F10.1038%2F171738a0&amp;rft_id=info%3Apmid%2F13054693&amp;rft.au=Wilkins+M.H.F.%2C+A.R.+Stokes+A.R.+%26+Wilson%2C+H.R.&amp;rft_id=http%3A%2F%2Fwww.nature.com%2Fnature%2Fdna50%2Fwilkins.pdf&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><span class="citation-comment" style="display:none; color:#33aa33; margin-left:0.3em">CS1 maint: Multiple names: authors list (<a href="/index.php?title=Category:CS1_maint:_Multiple_names:_authors_list&amp;action=edit&amp;redlink=1" class="new" title="Category:CS1 maint: Multiple names: authors list (page does not exist)">link</a>) </span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-129"><span class="mw-cite-backlink"><a href="#cite_ref-129" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><a rel="nofollow" class="external text" href="http://nobelprize.org/nobel_prizes/medicine/laureates/1962/">The Nobel Prize in Physiology or Medicine 1962</a> Nobelprize .org Accessed 22 Dec 06</span>
</li>
<li id="cite_note-130"><span class="mw-cite-backlink"><a href="#cite_ref-130" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Brenda Maddox (23 January 2003). <a rel="nofollow" class="external text" href="http://www.biomath.nyu.edu/index/course/hw_articles/nature4.pdf">"The double helix and the 'wronged heroine<span class="cs1-kern-right">'</span>"</a> <span class="cs1-format">(PDF)</span>. <i>Nature</i>. <b>421</b>: 407–408. <a href="/index.php/Digital_object_identifier" title="Digital object identifier">doi</a>:<a rel="nofollow" class="external text" href="//doi.org/10.1038%2Fnature01399">10.1038/nature01399</a>. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/12540909">12540909</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Nature&amp;rft.atitle=The+double+helix+and+the+%27wronged+heroine%27&amp;rft.volume=421&amp;rft.pages=407-408&amp;rft.date=2003-01-23&amp;rft_id=info%3Adoi%2F10.1038%2Fnature01399&amp;rft_id=info%3Apmid%2F12540909&amp;rft.au=Brenda+Maddox&amp;rft_id=http%3A%2F%2Fwww.biomath.nyu.edu%2Findex%2Fcourse%2Fhw_articles%2Fnature4.pdf&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-131"><span class="mw-cite-backlink"><a href="#cite_ref-131" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text">Crick, F.H.C. <a rel="nofollow" class="external text" href="http://genome.wellcome.ac.uk/assets/wtx030893.pdf">On degenerate templates and the adaptor hypothesis (PDF).</a> genome.wellcome.ac.uk (Lecture, 1955). Accessed 22 Dec 2006</span>
</li>
<li id="cite_note-132"><span class="mw-cite-backlink"><a href="#cite_ref-132" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><cite class="citation journal">Meselson M, Stahl F (1958). "The replication of DNA in <i>Escherichia coli</i>". <i>Proc Natl Acad Sci U S A</i>. <b>44</b> (7): 671&amp;ndash, 82. <a href="/index.php?title=PubMed_Identifier&amp;action=edit&amp;redlink=1" class="new" title="PubMed Identifier (page does not exist)">PMID</a>&nbsp;<a rel="nofollow" class="external text" href="//www.ncbi.nlm.nih.gov/pubmed/16590258">16590258</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Proc+Natl+Acad+Sci+U+S+A&amp;rft.atitle=The+replication+of+DNA+in+Escherichia+coli&amp;rft.volume=44&amp;rft.issue=7&amp;rft.pages=671%26ndash%2C+82&amp;rft.date=1958&amp;rft_id=info%3Apmid%2F16590258&amp;rft.au=Meselson+M%2C+Stahl+F&amp;rfr_id=info%3Asid%2Fwww.wikidoc.org%3ADNA" class="Z3988"></span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1594688"></span>
</li>
<li id="cite_note-133"><span class="mw-cite-backlink"><a href="#cite_ref-133" aria-label="Jump up" title="Jump up">↑</a></span> <span class="reference-text"><a rel="nofollow" class="external text" href="http://nobelprize.org/nobel_prizes/medicine/laureates/1968/">The Nobel Prize in Physiology or Medicine 1968</a> Nobelprize.org Accessed 22 Dec 06</span>
</li>
</ol></div>
<h2><span class="mw-headline" id="Further_reading">Further reading</span></h2>
<ul><li>Clayton, Julie. (Ed.). <i>50 Years of DNA</i>, Palgrave MacMillan Press, 2003. ISBN 978-1-40-391479-8</li>
<li>Judson, Horace Freeland. <i>The Eighth Day of Creation: Makers of the Revolution in Biology</i>, Cold Spring Harbor Laboratory Press, 1996. ISBN 978-0-87-969478-4</li>
<li>Olby, Robert. <i>The Path to The Double Helix: Discovery of DNA</i>, first published in October 1974 by MacMillan, with foreword by Francis Crick; ISBN 978-0-48-668117-7; the definitive DNA textbook, revised in 1994, with a 9 page postscript.</li>
<li>Ridley, Matt. <i>Francis Crick: Discoverer of the Genetic Code (Eminent Lives)</i> HarperCollins Publishers; 192 pp, ISBN 978-0-06-082333-7 2006</li>
<li>Rose, Steven. <i>The Chemistry of Life</i>, Penguin, ISBN 978-0-14-027273-4.</li>
<li>Watson, James D. and Francis H.C. Crick. <a rel="nofollow" class="external text" href="http://www.nature.com/nature/dna50/watsoncrick.pdf">A structure for Deoxyribose Nucleic Acid</a> (PDF). <i><a href="/index.php/Nature_(journal)" title="Nature (journal)">Nature</a></i> 171, 737–738, 25 April 1953.</li>
<li>Watson, James D. <i>DNA: The Secret of Life</i> ISBN 978-0-375-41546-3.</li>
<li>Watson, James D. <i>The Double Helix: A Personal Account of the Discovery of the Structure of DNA (Norton Critical Editions)</i>. ISBN 978-0-393-95075-5</li>
<li>Watson, James D. "Avoid boring people and other lessons from a life in science" (2007) New York: Random House. ISBN 978-0-375-41284-4</li>
<li>Calladine, Chris R.; Drew, Horace R.; Luisi, Ben F. and Travers, Andrew A. <i>Understanding DNA</i>, Elsevier Academic Press, 2003. ISBN 978-0-12155089-9</li></ul>
<h2><span class="mw-headline" id="External_links">External links</span></h2>
<ul><li><a rel="nofollow" class="external text" href="http://www.tjmitchell.com/stuart/dna.html">The Secret Life of DNA - DNA Music compositions</a></li>
<li><a rel="nofollow" class="external text" href="http://orpheus.ucsd.edu/speccoll/testing/html/mss0660a.html#abstract">The Register of Francis Crick Personal Papers 1938 - 2007</a> at Mandeville Special Collections Library, Geisel Library, University of California, San Diego</li>
<li><a rel="nofollow" class="external text" href="http://www.dnai.org/">DNA Interactive</a> This site from the Dolan DNA Learning Center included dozens of animations as well as interviews with James Watson and others (requires Adobe Flash)</li>
<li><a rel="nofollow" class="external text" href="http://www.dnaftb.org/dnaftb/">DNA from the Beginning</a> Another DNA Learning Center site on DNA, genes, and heredity from Mendel to the human genome project.</li>
<li><a rel="nofollow" class="external text" href="http://www.ncbe.reading.ac.uk/DNA50/">Double Helix 1953–2003</a> National Centre for Biotechnology Education</li>
<li><a rel="nofollow" class="external text" href="http://www.nature.com/nature/dna50/archive.html">Double helix: 50 years of DNA</a>, <i><a href="/index.php/Nature_(journal)" title="Nature (journal)">Nature</a></i></li>
<li><a rel="nofollow" class="external text" href="http://mason.gmu.edu/~emoody/rfranklin.html">Rosalind Franklin's contributions to the study of DNA</a></li>
<li><a rel="nofollow" class="external text" href="http://www.genome.gov/10506367">U.S. National DNA Day</a> — watch videos and participate in real-time chat with top scientists</li>
<li><a rel="nofollow" class="external text" href="http://www.genome.gov/10506718">Genetic Education Modules for Teachers</a> — <i>DNA from the Beginning</i> Study Guide</li>
<li><a rel="nofollow" class="external text" href="http://www.bbc.co.uk/bbcfour/audiointerviews/profilepages/crickwatson1.shtml">Listen to Francis Crick and James Watson talking on the BBC in 1962, 1972, and 1974</a></li>
<li><a href="/index.php/Protein_Data_Bank" title="Protein Data Bank">PDB</a> Molecule of the Month <i><a rel="nofollow" class="external text" href="https://pdb101.rcsb.org/motm/pdb23_1">DNA</a></i></li>
<li><a rel="nofollow" class="external text" href="http://www.fidelitysystems.com/Unlinked_DNA.html">DNA under electron microscope</a></li>
<li><a href="/index.php?title=Template:Dmoz&amp;action=edit&amp;redlink=1" class="new" title="Template:Dmoz (page does not exist)">Template:Dmoz</a></li>
<li><a rel="nofollow" class="external text" href="http://biostudio.com/c_%20education%20mac.htm">DNA coiling to form chromosomes</a></li>
<li><a rel="nofollow" class="external text" href="http://pipe.scs.fsu.edu/displar.html">DISPLAR: DNA binding site prediction on protein</a></li>
<li><a rel="nofollow" class="external text" href="http://www.dnalc.org/">Dolan DNA Learning Center</a></li>
<li>Olby, R. (2003) <a rel="nofollow" class="external text" href="http://chem-faculty.ucsd.edu/joseph/CHEM13/DNA1.pdf">"Quiet debut for the double helix"</a> <i>Nature</i> <b>421</b> (January 23): 402–405.</li>
<li><a rel="nofollow" class="external text" href="http://www.blackwellpublishing.com/trun/artwork/Animations/cloningexp/cloningexp.html">Basic animated guide to DNA cloning</a></li>
<li><a rel="nofollow" class="external text" href="http://nobelprize.org/educational_games/medicine/dna_double_helix/">DNA the Double Helix Game</a> From the official Nobel Prize web site</li>
<li><a rel="nofollow" class="external text" href="http://ca.youtube.com/watch?v=iyb7fwduuGM">DNA Lab, demonstrates how to extract DNA from wheat using readily available equipment and supplies.</a></li></ul>
<p><a href="/index.php?title=Template:Nucleic_acids&amp;action=edit&amp;redlink=1" class="new" title="Template:Nucleic acids (page does not exist)">Template:Nucleic acids</a>
<a href="/index.php?title=Template:Featured_article&amp;action=edit&amp;redlink=1" class="new" title="Template:Featured article (page does not exist)">Template:Featured article</a>
</p><p><a href="/index.php?title=Template:Link_FA&amp;action=edit&amp;redlink=1" class="new" title="Template:Link FA (page does not exist)">Template:Link FA</a>
<a href="/index.php?title=Template:Link_FA&amp;action=edit&amp;redlink=1" class="new" title="Template:Link FA (page does not exist)">Template:Link FA</a>
<a href="/index.php?title=Template:Link_FA&amp;action=edit&amp;redlink=1" class="new" title="Template:Link FA (page does not exist)">Template:Link FA</a>
<a href="/index.php?title=Template:Link_FA&amp;action=edit&amp;redlink=1" class="new" title="Template:Link FA (page does not exist)">Template:Link FA</a>
</p><p><a href="/index.php?title=Ar:%D8%AD%D9%85%D8%B6_%D9%86%D9%88%D9%88%D9%8A_%D8%B1%D9%8A%D8%A8%D9%8A_%D9%85%D9%86%D9%82%D9%88%D8%B5_%D8%A7%D9%84%D8%A3%D9%83%D8%B3%D8%AC%D9%8A%D9%86&amp;action=edit&amp;redlink=1" class="new" title="Ar:حمض نووي ريبي منقوص الأكسجين (page does not exist)">ar:حمض نووي ريبي منقوص الأكسجين</a>
<a href="/index.php?title=Az:Dezoksiribonuklein_tur%C5%9Fusu&amp;action=edit&amp;redlink=1" class="new" title="Az:Dezoksiribonuklein turşusu (page does not exist)">az:Dezoksiribonuklein turşusu</a>
<a href="/index.php?title=Bn:%E0%A6%A1%E0%A6%BF%E0%A6%8F%E0%A6%A8%E0%A6%8F&amp;action=edit&amp;redlink=1" class="new" title="Bn:ডিএনএ (page does not exist)">bn:ডিএনএ</a>
<a href="/index.php?title=Zh-min-nan:DNA&amp;action=edit&amp;redlink=1" class="new" title="Zh-min-nan:DNA (page does not exist)">zh-min-nan:DNA</a>
<a href="/index.php?title=Bs:Dezoksiribonukleinska_kiselina&amp;action=edit&amp;redlink=1" class="new" title="Bs:Dezoksiribonukleinska kiselina (page does not exist)">bs:Dezoksiribonukleinska kiselina</a>
<a href="/index.php?title=Br:Trenkenn_dezoksiribonukleek&amp;action=edit&amp;redlink=1" class="new" title="Br:Trenkenn dezoksiribonukleek (page does not exist)">br:Trenkenn dezoksiribonukleek</a>
<a href="/index.php?title=Bg:%D0%94%D0%9D%D0%9A&amp;action=edit&amp;redlink=1" class="new" title="Bg:ДНК (page does not exist)">bg:ДНК</a>
<a href="/index.php?title=Ca:%C3%80cid_desoxiribonucleic&amp;action=edit&amp;redlink=1" class="new" title="Ca:Àcid desoxiribonucleic (page does not exist)">ca:Àcid desoxiribonucleic</a>
<a href="/index.php?title=Cs:DNA&amp;action=edit&amp;redlink=1" class="new" title="Cs:DNA (page does not exist)">cs:DNA</a>
<a href="/index.php?title=Cy:DNA&amp;action=edit&amp;redlink=1" class="new" title="Cy:DNA (page does not exist)">cy:DNA</a>
<a href="/index.php?title=Da:Dna&amp;action=edit&amp;redlink=1" class="new" title="Da:Dna (page does not exist)">da:Dna</a>
<a href="/index.php?title=Pdc:DNA&amp;action=edit&amp;redlink=1" class="new" title="Pdc:DNA (page does not exist)">pdc:DNA</a>
<a href="/index.php?title=De:Desoxyribonukleins%C3%A4ure&amp;action=edit&amp;redlink=1" class="new" title="De:Desoxyribonukleinsäure (page does not exist)">de:Desoxyribonukleinsäure</a>
<a href="/index.php?title=Et:Desoks%C3%BCribonukleiinhape&amp;action=edit&amp;redlink=1" class="new" title="Et:Desoksüribonukleiinhape (page does not exist)">et:Desoksüribonukleiinhape</a>
<a href="/index.php?title=El:DNA&amp;action=edit&amp;redlink=1" class="new" title="El:DNA (page does not exist)">el:DNA</a>
<a href="/index.php?title=Eo:DNA&amp;action=edit&amp;redlink=1" class="new" title="Eo:DNA (page does not exist)">eo:DNA</a>
<a href="/index.php?title=Eu:Azido_desoxirribonukleiko&amp;action=edit&amp;redlink=1" class="new" title="Eu:Azido desoxirribonukleiko (page does not exist)">eu:Azido desoxirribonukleiko</a>
<a href="/index.php?title=Fa:%D8%AF%DB%8C%E2%80%8C%D8%A7%D9%86%E2%80%8C%D8%A7%DB%8C&amp;action=edit&amp;redlink=1" class="new" title="Fa:دی‌ان‌ای (page does not exist)">fa:دی‌ان‌ای</a>
<a href="/index.php?title=Fo:DNA&amp;action=edit&amp;redlink=1" class="new" title="Fo:DNA (page does not exist)">fo:DNA</a>
<a href="/index.php?title=Ga:ADN&amp;action=edit&amp;redlink=1" class="new" title="Ga:ADN (page does not exist)">ga:ADN</a>
<a href="/index.php?title=Gl:ADN&amp;action=edit&amp;redlink=1" class="new" title="Gl:ADN (page does not exist)">gl:ADN</a>
<a href="/index.php?title=Ko:DNA&amp;action=edit&amp;redlink=1" class="new" title="Ko:DNA (page does not exist)">ko:DNA</a>
<a href="/index.php?title=Hr:Deoksiribonukleinska_kiselina&amp;action=edit&amp;redlink=1" class="new" title="Hr:Deoksiribonukleinska kiselina (page does not exist)">hr:Deoksiribonukleinska kiselina</a>
<a href="/index.php?title=Id:Asam_deoksiribonukleat&amp;action=edit&amp;redlink=1" class="new" title="Id:Asam deoksiribonukleat (page does not exist)">id:Asam deoksiribonukleat</a>
<a href="/index.php?title=Is:DNA&amp;action=edit&amp;redlink=1" class="new" title="Is:DNA (page does not exist)">is:DNA</a>
<a href="/index.php?title=It:DNA&amp;action=edit&amp;redlink=1" class="new" title="It:DNA (page does not exist)">it:DNA</a>
<a href="/index.php?title=He:DNA&amp;action=edit&amp;redlink=1" class="new" title="He:DNA (page does not exist)">he:DNA</a>
<a href="/index.php?title=Pam:DNA&amp;action=edit&amp;redlink=1" class="new" title="Pam:DNA (page does not exist)">pam:DNA</a>
<a href="/index.php?title=Ka:%E1%83%93%E1%83%94%E1%83%96%E1%83%9D%E1%83%A5%E1%83%A1%E1%83%98%E1%83%A0%E1%83%98%E1%83%91%E1%83%9D%E1%83%9C%E1%83%A3%E1%83%99%E1%83%9A%E1%83%94%E1%83%98%E1%83%9C%E1%83%98%E1%83%A1_%E1%83%9B%E1%83%9F%E1%83%90%E1%83%95%E1%83%90&amp;action=edit&amp;redlink=1" class="new" title="Ka:დეზოქსირიბონუკლეინის მჟავა (page does not exist)">ka:დეზოქსირიბონუკლეინის მჟავა</a>
<a href="/index.php?title=Ht:ADN&amp;action=edit&amp;redlink=1" class="new" title="Ht:ADN (page does not exist)">ht:ADN</a>
<a href="/index.php?title=Ku:DNA&amp;action=edit&amp;redlink=1" class="new" title="Ku:DNA (page does not exist)">ku:DNA</a>
<a href="/index.php?title=La:Acidum_desoxyribonucleicum&amp;action=edit&amp;redlink=1" class="new" title="La:Acidum desoxyribonucleicum (page does not exist)">la:Acidum desoxyribonucleicum</a>
<a href="/index.php?title=Lv:Dezoksiribonukle%C4%ABnsk%C4%81be&amp;action=edit&amp;redlink=1" class="new" title="Lv:Dezoksiribonukleīnskābe (page does not exist)">lv:Dezoksiribonukleīnskābe</a>
<a href="/index.php?title=Lt:Deoksiribonukleor%C5%ABg%C5%A1tis&amp;action=edit&amp;redlink=1" class="new" title="Lt:Deoksiribonukleorūgštis (page does not exist)">lt:Deoksiribonukleorūgštis</a>
<a href="/index.php?title=Hu:Dezoxiribonukleinsav&amp;action=edit&amp;redlink=1" class="new" title="Hu:Dezoxiribonukleinsav (page does not exist)">hu:Dezoxiribonukleinsav</a>
<a href="/index.php?title=Mk:%D0%94%D0%9D%D0%9A&amp;action=edit&amp;redlink=1" class="new" title="Mk:ДНК (page does not exist)">mk:ДНК</a>
<a href="/index.php?title=Ml:%E0%B4%A1%E0%B4%BF.%E0%B4%8E%E0%B4%A8%E0%B5%8D%E2%80%8D.%E0%B4%8E.&amp;action=edit&amp;redlink=1" class="new" title="Ml:ഡി.എന്‍.എ. (page does not exist)">ml:ഡി.എന്‍.എ.</a>
<a href="/index.php?title=Ms:DNA&amp;action=edit&amp;redlink=1" class="new" title="Ms:DNA (page does not exist)">ms:DNA</a>
<a href="/index.php?title=Mn:%D0%94%D0%9D%D0%A5&amp;action=edit&amp;redlink=1" class="new" title="Mn:ДНХ (page does not exist)">mn:ДНХ</a>
<a href="/index.php?title=Nl:DNA&amp;action=edit&amp;redlink=1" class="new" title="Nl:DNA (page does not exist)">nl:DNA</a>
<a href="/index.php?title=No:DNA&amp;action=edit&amp;redlink=1" class="new" title="No:DNA (page does not exist)">no:DNA</a>
<a href="/index.php?title=Nn:Deoksyribonukleinsyre&amp;action=edit&amp;redlink=1" class="new" title="Nn:Deoksyribonukleinsyre (page does not exist)">nn:Deoksyribonukleinsyre</a>
<a href="/index.php?title=Nov:DNA&amp;action=edit&amp;redlink=1" class="new" title="Nov:DNA (page does not exist)">nov:DNA</a>
<a href="/index.php?title=Oc:Acid_desoxiribonucle%C3%AFc&amp;action=edit&amp;redlink=1" class="new" title="Oc:Acid desoxiribonucleïc (page does not exist)">oc:Acid desoxiribonucleïc</a>
<a href="/index.php?title=Om:DNA&amp;action=edit&amp;redlink=1" class="new" title="Om:DNA (page does not exist)">om:DNA</a>
<a href="/index.php?title=Ps:%DA%89%D9%8A_%D8%A7%D9%86_%D8%A7%DB%90_(DNA)&amp;action=edit&amp;redlink=1" class="new" title="Ps:ډي ان اې (DNA) (page does not exist)">ps:ډي ان اې (DNA)</a>
<a href="/index.php?title=Pms:DNA&amp;action=edit&amp;redlink=1" class="new" title="Pms:DNA (page does not exist)">pms:DNA</a>
<a href="/index.php?title=Sq:ADN&amp;action=edit&amp;redlink=1" class="new" title="Sq:ADN (page does not exist)">sq:ADN</a>
<a href="/index.php?title=Simple:DNA&amp;action=edit&amp;redlink=1" class="new" title="Simple:DNA (page does not exist)">simple:DNA</a>
<a href="/index.php?title=Sk:Deoxyribonukleov%C3%A1_kyselina&amp;action=edit&amp;redlink=1" class="new" title="Sk:Deoxyribonukleová kyselina (page does not exist)">sk:Deoxyribonukleová kyselina</a>
<a href="/index.php?title=Sl:Deoksiribonukleinska_kislina&amp;action=edit&amp;redlink=1" class="new" title="Sl:Deoksiribonukleinska kislina (page does not exist)">sl:Deoksiribonukleinska kislina</a>
<a href="/index.php?title=Sr:%D0%94%D0%9D%D0%9A&amp;action=edit&amp;redlink=1" class="new" title="Sr:ДНК (page does not exist)">sr:ДНК</a>
<a href="/index.php?title=Sh:DNK&amp;action=edit&amp;redlink=1" class="new" title="Sh:DNK (page does not exist)">sh:DNK</a>
<a href="/index.php?title=Su:DNA&amp;action=edit&amp;redlink=1" class="new" title="Su:DNA (page does not exist)">su:DNA</a>
<a href="/index.php?title=Fi:DNA&amp;action=edit&amp;redlink=1" class="new" title="Fi:DNA (page does not exist)">fi:DNA</a>
<a href="/index.php?title=Sv:DNA&amp;action=edit&amp;redlink=1" class="new" title="Sv:DNA (page does not exist)">sv:DNA</a>
<a href="/index.php?title=Tl:DNA&amp;action=edit&amp;redlink=1" class="new" title="Tl:DNA (page does not exist)">tl:DNA</a>
<a href="/index.php?title=Ta:%E0%AE%86%E0%AE%95%E0%AF%8D%E0%AE%9A%E0%AE%BF%E0%AE%9C%E0%AE%A9%E0%AE%B1%E0%AF%8D%E0%AE%B1_%E0%AE%B0%E0%AF%88%E0%AE%AA%E0%AF%8B_%E0%AE%95%E0%AE%B0%E0%AF%81_%E0%AE%85%E0%AE%AE%E0%AE%BF%E0%AE%B2%E0%AE%AE%E0%AF%8D&amp;action=edit&amp;redlink=1" class="new" title="Ta:ஆக்சிஜனற்ற ரைபோ கரு அமிலம் (page does not exist)">ta:ஆக்சிஜனற்ற ரைபோ கரு அமிலம்</a>
<a href="/index.php?title=Te:%E0%B0%A1%E0%B1%80%E0%B0%86%E0%B0%95%E0%B1%8D%E0%B0%B8%E0%B1%80%E0%B0%B0%E0%B1%88%E0%B0%AC%E0%B1%8B_%E0%B0%95%E0%B1%87%E0%B0%82%E0%B0%A6%E0%B1%8D%E0%B0%B0%E0%B0%95_%E0%B0%86%E0%B0%AE%E0%B1%8D%E0%B0%B2%E0%B0%82&amp;action=edit&amp;redlink=1" class="new" title="Te:డీఆక్సీరైబో కేంద్రక ఆమ్లం (page does not exist)">te:డీఆక్సీరైబో కేంద్రక ఆమ్లం</a>
<a href="/index.php?title=Th:%E0%B8%94%E0%B8%B5%E0%B9%80%E0%B8%AD%E0%B9%87%E0%B8%99%E0%B9%80%E0%B8%AD&amp;action=edit&amp;redlink=1" class="new" title="Th:ดีเอ็นเอ (page does not exist)">th:ดีเอ็นเอ</a>
<a href="/index.php?title=Uk:%D0%94%D0%B5%D0%B7%D0%BE%D0%BA%D1%81%D0%B8%D1%80%D0%B8%D0%B1%D0%BE%D0%BD%D1%83%D0%BA%D0%BB%D0%B5%D1%97%D0%BD%D0%BE%D0%B2%D0%B0_%D0%BA%D0%B8%D1%81%D0%BB%D0%BE%D1%82%D0%B0&amp;action=edit&amp;redlink=1" class="new" title="Uk:Дезоксирибонуклеїнова кислота (page does not exist)">uk:Дезоксирибонуклеїнова кислота</a>
<a href="/index.php?title=Ur:%DA%88%DB%8C_%D8%A7%DB%8C%D9%86_%D8%A7%DB%92&amp;action=edit&amp;redlink=1" class="new" title="Ur:ڈی این اے (page does not exist)">ur:ڈی این اے</a>
<a href="/index.php?title=Vls:DNA&amp;action=edit&amp;redlink=1" class="new" title="Vls:DNA (page does not exist)">vls:DNA</a>
<a href="/index.php?title=Yi:%D7%93%D7%99_%D7%A2%D7%9F_%D7%90%D7%99%D7%99&amp;action=edit&amp;redlink=1" class="new" title="Yi:די ען איי (page does not exist)">yi:די ען איי</a>
<a href="/index.php?title=Yo:DNA&amp;action=edit&amp;redlink=1" class="new" title="Yo:DNA (page does not exist)">yo:DNA</a>
</p><p><a href="/index.php?title=Template:WH&amp;action=edit&amp;redlink=1" class="new" title="Template:WH (page does not exist)">Template:WH</a>
<a href="/index.php?title=Template:WS&amp;action=edit&amp;redlink=1" class="new" title="Template:WS (page does not exist)">Template:WS</a>
</p>
<!-- 
NewPP limit report
Cached time: 20240821210358
Cache expiry: 1209600
Reduced expiry: false
Complications: [show‐toc, no‐toc‐conversion]
CPU time usage: 0.612 seconds
Real time usage: 1.534 seconds
Preprocessor visited node count: 5405/1000000
Post‐expand include size: 210482/2097152 bytes
Template argument size: 91/2097152 bytes
Highest expansion depth: 8/100
Expensive parser function count: 0/100
Unstrip recursion depth: 1/20
Unstrip post‐expand size: 325486/5000000 bytes
Lua time usage: 0.940/7 seconds
Lua virtual size: 8200192/52428800 bytes
Lua estimated memory usage: 0 bytes
-->
<!--
Transclusion expansion time report (%,ms,calls,template)
100.00% 1345.435      1 -total
 91.02% 1224.599      1 Template:Reflist
 79.26% 1066.429    108 Template:Cite_journal
  6.17%   82.970      4 Template:Cite_book
  4.96%   66.687     18 Template:Further
  0.68%    9.213      1 Template:Cite_web
  0.32%    4.361      1 Template:CMG
  0.24%    3.265      1 Template:Main_other
  0.17%    2.338      1 Template:PDB_Molecule_of_the_Month
  0.16%    2.091      1 Template:Column-width
-->

<!-- Saved in parser cache with key wikidb-wd_:pcache:idhash:19727-0!canonical and timestamp 20240821210357 and revision id 1590464. Rendering was triggered because: page-view
 -->
</div>"""
```
